```python
# =============================================================================
# comparison_final.py
# Purpose : Master comparison script — loads all model results and produces:
#           1. Unified predictions table (all models, all vintages)
#           2. Relative RMSE table (vs ARMA = 1) by period and vintage
#           3. Relative MAE  table (vs ARMA = 1) by period and vintage
#           4. Publication-ready summary table (crisis + normal combined)
#           5. Bar chart: RMSE by model at nowcast vintage (lag=0)
#           6. Heatmap: relative RMSE by model x vintage (crisis period)
#           7. Heatmap: relative RMSE by model x vintage (normal period)
#           8. Model ranking table (for paper)
#
# This script is the entry point for all subsequent extended analyses
# (Diebold-Mariano, fan charts, CRPS). Run this first.
#
# Author  : Gabriela Aquino
# Input   : ~/Desktop/ESA-gdp-nowcasting/Models/results/*/
# Output  : ~/Desktop/ESA-gdp-nowcasting/Models/results/comparison/
# =============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import warnings
from pathlib import Path

warnings.filterwarnings("ignore")

# =============================================================================
# 0. PATHS
# =============================================================================

BASE_DIR  = Path.home() / "Desktop" / "ESA-gdp-nowcasting"
MODEL_DIR = BASE_DIR / "Models"
RES_DIR   = MODEL_DIR / "results"
OUT_DIR   = RES_DIR / "comparison"
OUT_DIR.mkdir(parents=True, exist_ok=True)

# =============================================================================
# 1. CONFIGURATION
# =============================================================================

# Vintages to include in the comparison
# Must match the lag column names in each model's predictions CSV
LAGS = [-2, -1, 0]

# Period definitions
CRISIS_START = "2019-01-01"
CRISIS_END   = "2021-06-01"
NORMAL_START = "2022-01-01"
NORMAL_END   = "2024-06-01"

# =============================================================================
# 2. MODEL REGISTRY
# =============================================================================
# Each entry defines:
#   folder   : subfolder inside results/
#   file     : predictions CSV filename
#   col_map  : mapping from lag integer to column name in that CSV
#   category : "econometric" or "ml"
#   color    : plot color
#
# ADD NEW MODELS HERE when their CSVs are ready.
# The script will skip models whose CSV does not exist yet.

MODEL_REGISTRY = {
    "ARMA": {
        "folder":   "arma",
        "file":     "arma_predictions.csv",
        "col_map":  {-2: "lag_-2", -1: "lag_-1", 0: "lag_+0"},
        "category": "econometric",
        "color":    "#6c757d",
    },
    "MIDAS": {
        "folder":   "midas",
        "file":     "midas_predictions.csv",
        "col_map":  {-2: "lag_-2", -1: "lag_-1", 0: "lag_+0"},
        "category": "econometric",
        "color":    "#2196F3",
    },
    "DFM": {
        "folder":   "dfm",
        "file":     "dfm_predictions.csv",
        "col_map":  {-2: "lag_-2", -1: "lag_-1", 0: "lag_+0"},
        "category": "econometric",
        "color":    "#00BCD4",
    },
    "Ridge": {
        "folder":   "ridge",
        "file":     "ridge_predictions.csv",
        "col_map":  {-2: "lag_-2", -1: "lag_-1", 0: "lag_+0"},
        "category": "ml_linear",
        "color":    "#4CAF50",
    },
    "Lasso": {
        "folder":   "lasso",
        "file":     "lasso_predictions.csv",
        "col_map":  {-2: "lag_-2", -1: "lag_-1", 0: "lag_+0"},
        "category": "ml_linear",
        "color":    "#8BC34A",
    },
    "XGBoost": {
        "folder":   "xgboost",
        "file":     "xgboost_predictions.csv",
        "col_map":  {-2: "lag_-2", -1: "lag_-1", 0: "lag_+0"},
        "category": "ml_nonlinear",
        "color":    "#FF5722",
    },
    "RandomForest": {
        "folder":   "randomforest",
        "file":     "rf_predictions.csv",
        "col_map":  {-2: "lag_-2", -1: "lag_-1", 0: "lag_+0"},
        "category": "ml_nonlinear",
        "color":    "#FF9800",
    },
}

# Display names for tables and plots
DISPLAY_NAMES = {
    "ARMA":         "ARMA",
    "MIDAS":        "MIDAS",
    "DFM":          "DFM",
    "Ridge":        "Ridge",
    "Lasso":        "Lasso",
    "XGBoost":      "XGBoost",
    "RandomForest": "Random Forest",
}

# =============================================================================
# 3. LOAD ALL MODEL PREDICTIONS
# =============================================================================

print("Loading model predictions...")
print()

all_preds  = {}   # model_name -> dict{lag -> Series of predictions}
actuals    = None
dates      = None
periods    = None

for model_name, cfg in MODEL_REGISTRY.items():
    csv_path = RES_DIR / cfg["folder"] / cfg["file"]
    if not csv_path.exists():
        print(f"  [{model_name}] NOT FOUND — skipping: {csv_path}")
        continue

    df = pd.read_csv(csv_path, parse_dates=["date"])
    df = df.sort_values("date").reset_index(drop=True)

    # Load actuals from the first model that has them
    if actuals is None:
        actuals = df["actual"].values
        dates   = df["date"].values
        periods = df["period_type"].values if "period_type" in df.columns else None
        print(f"  [{model_name}] Loaded — {len(df)} quarters, actuals extracted")
    else:
        print(f"  [{model_name}] Loaded — {len(df)} quarters")

    # Extract predictions for each vintage
    lag_preds = {}
    for lag, col_name in cfg["col_map"].items():
        if col_name in df.columns:
            lag_preds[lag] = df[col_name].values
        else:
            # Try alternative column name formats
            alt_names = [
                f"lag_{lag}", f"lag_{'+' if lag>=0 else ''}{lag}",
                f"lag_+{abs(lag)}" if lag >= 0 else f"lag_{lag}"
            ]
            found = False
            for alt in alt_names:
                if alt in df.columns:
                    lag_preds[lag] = df[alt].values
                    found = True
                    break
            if not found:
                print(f"    WARNING: column for lag={lag} not found in {cfg['file']}")
                lag_preds[lag] = np.full(len(df), np.nan)

    all_preds[model_name] = lag_preds

loaded_models = list(all_preds.keys())
print(f"\nModels loaded: {loaded_models}")
print(f"Models missing: {[m for m in MODEL_REGISTRY if m not in loaded_models]}")

if not loaded_models:
    raise RuntimeError("No model CSVs found. Check paths in MODEL_REGISTRY.")

if actuals is None:
    raise RuntimeError("Could not load actuals. Check that at least one CSV has an 'actual' column.")

# =============================================================================
# 4. HELPER FUNCTIONS
# =============================================================================

def _rmse(y, yhat):
    y, yhat = np.array(y, float), np.array(yhat, float)
    m = ~np.isnan(y) & ~np.isnan(yhat)
    return float(np.sqrt(np.mean((y[m] - yhat[m])**2))) if m.sum() > 0 else np.nan

def _mae(y, yhat):
    y, yhat = np.array(y, float), np.array(yhat, float)
    m = ~np.isnan(y) & ~np.isnan(yhat)
    return float(np.mean(np.abs(y[m] - yhat[m]))) if m.sum() > 0 else np.nan

def get_period_mask(period_name):
    """Return boolean mask for a given period."""
    dates_dt = pd.to_datetime(dates)
    if period_name == "crisis":
        return (dates_dt >= CRISIS_START) & (dates_dt <= CRISIS_END)
    elif period_name == "normal":
        return (dates_dt >= NORMAL_START) & (dates_dt <= NORMAL_END)
    else:
        return np.ones(len(dates), dtype=bool)

# =============================================================================
# 5. COMPUTE PERFORMANCE TABLES
# =============================================================================

# ---- 5A. Raw RMSE and MAE ---------------------------------------------------
# Compute for each model x vintage x period

rows = []
for period_name in ["full", "crisis", "normal"]:
    mask = get_period_mask(period_name)
    y    = actuals[mask]

    for model_name in loaded_models:
        for lag in LAGS:
            yhat = all_preds[model_name][lag][mask]
            rows.append({
                "model":   model_name,
                "display": DISPLAY_NAMES.get(model_name, model_name),
                "period":  period_name,
                "vintage": lag,
                "RMSE":    round(_rmse(y, yhat), 6),
                "MAE":     round(_mae(y,  yhat), 6),
                "n":       int((~np.isnan(y) & ~np.isnan(yhat)).sum()),
            })

perf_raw = pd.DataFrame(rows)

# ---- 5B. Relative RMSE and MAE (vs ARMA = 1) --------------------------------
if "ARMA" in loaded_models:
    arma_rmse = {}
    arma_mae  = {}
    for period_name in ["full", "crisis", "normal"]:
        mask = get_period_mask(period_name)
        y    = actuals[mask]
        # Use lag=0 as the reference vintage for ARMA (all lags identical for ARMA)
        arma_yhat = all_preds["ARMA"][0][mask]
        arma_rmse[period_name] = _rmse(y, arma_yhat)
        arma_mae[period_name]  = _mae(y,  arma_yhat)

    perf_raw["rel_RMSE"] = perf_raw.apply(
        lambda row: round(row["RMSE"] / arma_rmse[row["period"]], 4)
        if arma_rmse[row["period"]] and not np.isnan(row["RMSE"]) else np.nan,
        axis=1
    )
    perf_raw["rel_MAE"] = perf_raw.apply(
        lambda row: round(row["MAE"] / arma_mae[row["period"]], 4)
        if arma_mae[row["period"]] and not np.isnan(row["MAE"]) else np.nan,
        axis=1
    )
    print("\nARMA RMSE benchmark:")
    for p, v in arma_rmse.items():
        print(f"  {p}: {v:.6f}")
else:
    print("\nWARNING: ARMA not loaded. Relative RMSE cannot be computed.")
    perf_raw["rel_RMSE"] = np.nan
    perf_raw["rel_MAE"]  = np.nan

# ---- 5C. Pivot tables for paper ---------------------------------------------

def make_pivot_table(metric, period):
    """Create model x vintage pivot table for a given metric and period."""
    sub = perf_raw[perf_raw["period"] == period].copy()
    pivot = sub.pivot(index="display", columns="vintage", values=metric)
    # Rename columns to match paper format
    col_rename = {-2: "lag=-2", -1: "lag=-1", 0: "lag=0 (nowcast)"}
    pivot.columns = [col_rename.get(c, str(c)) for c in pivot.columns]
    # Add average across vintages
    pivot["Average"] = pivot.mean(axis=1)
    # Sort by average (best first)
    pivot = pivot.sort_values("Average")
    return pivot.round(4)

# Relative RMSE tables (paper format)
tbl_rmse_crisis = make_pivot_table("rel_RMSE", "crisis")
tbl_rmse_normal = make_pivot_table("rel_RMSE", "normal")
tbl_rmse_full   = make_pivot_table("rel_RMSE", "full")

# Relative MAE tables
tbl_mae_crisis  = make_pivot_table("rel_MAE", "crisis")
tbl_mae_normal  = make_pivot_table("rel_MAE", "normal")

print("\n" + "="*70)
print("RELATIVE RMSE (ARMA = 1) — CRISIS PERIOD (2019Q1-2021Q2)")
print("Values < 1 = better than ARMA benchmark")
print("="*70)
print(tbl_rmse_crisis.to_string())

print("\n" + "="*70)
print("RELATIVE RMSE (ARMA = 1) — NON-CRISIS PERIOD (2022Q1-2024Q2)")
print("="*70)
print(tbl_rmse_normal.to_string())

print("\n" + "="*70)
print("RELATIVE RMSE (ARMA = 1) — FULL PERIOD (2019Q1-2024Q2)")
print("="*70)
print(tbl_rmse_full.to_string())

# =============================================================================
# 6. SAVE TABLES
# =============================================================================

perf_raw.to_csv(OUT_DIR / "comparison_raw.csv", index=False)
tbl_rmse_crisis.to_csv(OUT_DIR / "rel_rmse_crisis.csv")
tbl_rmse_normal.to_csv(OUT_DIR / "rel_rmse_normal.csv")
tbl_rmse_full.to_csv(OUT_DIR   / "rel_rmse_full.csv")
tbl_mae_crisis.to_csv(OUT_DIR  / "rel_mae_crisis.csv")
tbl_mae_normal.to_csv(OUT_DIR  / "rel_mae_normal.csv")

# Unified predictions CSV (all models, all vintages, all dates)
unified = pd.DataFrame({"date": dates, "actual": actuals, "period": periods})
for model_name in loaded_models:
    for lag in LAGS:
        col = f"{DISPLAY_NAMES.get(model_name, model_name)}_lag{lag}"
        unified[col] = all_preds[model_name][lag]
unified.to_csv(OUT_DIR / "unified_predictions.csv", index=False)

print(f"\nCSVs saved to: {OUT_DIR}")

# =============================================================================
# 7. PLOTS
# =============================================================================

# ---- Plot settings ----------------------------------------------------------
MODEL_COLORS = {
    m: MODEL_REGISTRY[m]["color"]
    for m in loaded_models
}
VINTAGE_LABELS = {-2: "lag=-2", -1: "lag=-1", 0: "lag=0\n(nowcast)"}

def format_xaxis(ax):
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))

# ---- PLOT 1: RMSE bar chart at nowcast (lag=0), crisis vs normal -----------

fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=False)
fig.suptitle(
    "Relative RMSE vs ARMA Benchmark (lag=0, nowcast vintage)\n"
    "Values below 1.0 indicate improvement over the ARMA benchmark",
    fontsize=13, fontweight="bold"
)

for ax, period, title in zip(
    axes,
    ["crisis", "normal"],
    ["Crisis Period (2019Q1-2021Q2)", "Non-Crisis Period (2022Q1-2024Q2)"]
):
    sub = perf_raw[(perf_raw["period"] == period) & (perf_raw["vintage"] == 0)]
    sub = sub.sort_values("rel_RMSE")
    colors = [MODEL_COLORS.get(m, "#888888") for m in sub["model"]]

    bars = ax.bar(sub["display"], sub["rel_RMSE"],
                  color=colors, edgecolor="gray", linewidth=0.5)
    ax.axhline(1.0, color="#e63946", lw=1.5, ls="--", label="ARMA = 1.0")

    # Add value labels on bars
    for bar, val in zip(bars, sub["rel_RMSE"]):
        ax.text(bar.get_x() + bar.get_width()/2,
                bar.get_height() + 0.01,
                f"{val:.3f}",
                ha="center", va="bottom", fontsize=8)

    ax.set_title(title, fontsize=11, fontweight="bold")
    ax.set_xlabel("")
    ax.set_ylabel("Relative RMSE (ARMA = 1)", fontsize=10)
    ax.set_ylim(0, max(sub["rel_RMSE"].max() * 1.2, 1.3))
    ax.tick_params(axis="x", rotation=30)
    ax.grid(True, axis="y", alpha=0.3)
    ax.legend(fontsize=9)

plt.tight_layout()
fig.savefig(OUT_DIR / "plot1_rmse_bar_crisis_normal.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("Saved: plot1_rmse_bar_crisis_normal.png")

# ---- PLOT 2: Heatmap — relative RMSE by model x vintage (crisis) -----------

fig, axes = plt.subplots(1, 2, figsize=(16, max(4, len(loaded_models) * 0.7 + 1)))
fig.suptitle(
    "Relative RMSE (ARMA = 1) by Model and Vintage\n"
    "Green = better than ARMA | Red = worse than ARMA",
    fontsize=13, fontweight="bold"
)

for ax, period, title in zip(
    axes,
    ["crisis", "normal"],
    ["Crisis Period (2019Q1-2021Q2)", "Non-Crisis Period (2022Q1-2024Q2)"]
):
    sub = perf_raw[perf_raw["period"] == period][
        ["display", "vintage", "rel_RMSE"]
    ].copy()
    sub = sub[sub["vintage"].isin(LAGS)]
    pivot = sub.pivot(index="display", columns="vintage", values="rel_RMSE")
    pivot.columns = [VINTAGE_LABELS.get(c, str(c)) for c in pivot.columns]

    # Sort rows by average relative RMSE
    pivot = pivot.loc[pivot.mean(axis=1).sort_values().index]

    # Use diverging colormap centered at 1.0
    vmin = min(0.5, pivot.min().min())
    vmax = max(1.5, pivot.max().max())
    center = 1.0

    sns.heatmap(pivot,
                annot=True, fmt=".3f",
                cmap="RdYlGn_r",
                center=center, vmin=vmin, vmax=vmax,
                linewidths=0.5, ax=ax,
                cbar_kws={"label": "Relative RMSE", "shrink": 0.8})
    ax.set_title(title, fontsize=11, fontweight="bold", pad=8)
    ax.set_xlabel("Vintage")
    ax.set_ylabel("")
    ax.tick_params(axis="x", rotation=0)

plt.tight_layout()
fig.savefig(OUT_DIR / "plot2_heatmap_rel_rmse.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("Saved: plot2_heatmap_rel_rmse.png")

# ---- PLOT 3: Vintage profiles — RMSE across vintages for each model --------
# One line per model showing how relative RMSE changes from lag=-2 to lag=0

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle(
    "Relative RMSE by Vintage — How Accuracy Improves as Information Arrives\n"
    "Each line shows one model's RMSE relative to ARMA across the 3 vintages",
    fontsize=12, fontweight="bold"
)

for ax, period, title in zip(
    axes,
    ["crisis", "normal"],
    ["Crisis Period (2019Q1-2021Q2)", "Non-Crisis Period (2022Q1-2024Q2)"]
):
    sub = perf_raw[perf_raw["period"] == period]

    for model_name in loaded_models:
        m_sub = sub[sub["model"] == model_name].sort_values("vintage")
        color = MODEL_COLORS.get(model_name, "#888888")
        lw    = 2.5 if model_name in ["Ridge", "MIDAS", "DFM"] else 1.2
        ls    = "-" if model_name in ["Ridge", "MIDAS", "DFM"] else "--"
        ax.plot(
            m_sub["vintage"],
            m_sub["rel_RMSE"],
            color=color, lw=lw, ls=ls,
            marker="o", markersize=6,
            label=DISPLAY_NAMES.get(model_name, model_name)
        )

    ax.axhline(1.0, color="#e63946", lw=1.5, ls=":", label="ARMA = 1.0")
    ax.set_xticks(LAGS)
    ax.set_xticklabels(["lag=-2\n(2 months before)", "lag=-1\n(1 month before)",
                         "lag=0\n(nowcast)"])
    ax.set_xlabel("Vintage")
    ax.set_ylabel("Relative RMSE (ARMA = 1)")
    ax.set_title(title, fontsize=11, fontweight="bold")
    ax.legend(fontsize=8, loc="upper right")
    ax.grid(True, alpha=0.3)

plt.tight_layout()
fig.savefig(OUT_DIR / "plot3_vintage_profiles.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("Saved: plot3_vintage_profiles.png")

# ---- PLOT 4: Full-period actual vs predicted (nowcast, top 3 models) -------

fig, ax = plt.subplots(figsize=(14, 6))

dates_dt = pd.to_datetime(dates)

# Full GDP series for context
ax.plot(dates_dt, actuals,
        color="#333333", lw=2, marker="o", markersize=4,
        label="Actual GDP (QoQ)", zorder=5)

# Top models at nowcast vintage
top_models = ["Ridge", "MIDAS", "ARMA"]
for model_name in top_models:
    if model_name not in loaded_models:
        continue
    yhat  = all_preds[model_name][0]
    color = MODEL_COLORS.get(model_name, "#888888")
    lw    = 2.0 if model_name != "ARMA" else 1.2
    ls    = "-" if model_name != "ARMA" else "--"
    ax.plot(dates_dt, yhat,
            color=color, lw=lw, ls=ls,
            label=DISPLAY_NAMES.get(model_name, model_name) + " (lag=0)",
            alpha=0.85)

# Shade crisis and normal windows
ax.axvspan(pd.to_datetime(CRISIS_START), pd.to_datetime(CRISIS_END),
           alpha=0.10, color="#e63946", label="Crisis window")
ax.axvspan(pd.to_datetime(NORMAL_START), pd.to_datetime(NORMAL_END),
           alpha=0.08, color="#2a9d8f", label="Non-crisis window")

ax.set_xlabel("Date")
ax.set_ylabel("GDP growth rate (QoQ)")
ax.set_title(
    "Nowcasting Performance — Top Models at Nowcast Vintage (lag=0)\n"
    "El Salvador | 2019Q1-2024Q2 | Full evaluation period",
    fontsize=12, fontweight="bold"
)
ax.legend(fontsize=9, ncol=3, loc="lower left")
ax.grid(True, alpha=0.25)
format_xaxis(ax)
plt.tight_layout()
fig.savefig(OUT_DIR / "plot4_actual_vs_predicted.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("Saved: plot4_actual_vs_predicted.png")

# =============================================================================
# 8. MODEL RANKING TABLE (paper-ready)
# =============================================================================

# Rank models by: (1) crisis RMSE lag=0, (2) normal RMSE lag=0, (3) average
ranking_rows = []
for model_name in loaded_models:
    sub = perf_raw[perf_raw["model"] == model_name]

    crisis_rmse = sub[(sub["period"]=="crisis") & (sub["vintage"]==0)]["rel_RMSE"].values
    normal_rmse = sub[(sub["period"]=="normal") & (sub["vintage"]==0)]["rel_RMSE"].values
    full_rmse   = sub[(sub["period"]=="full")   & (sub["vintage"]==0)]["rel_RMSE"].values
    crisis_avg  = sub[sub["period"]=="crisis"]["rel_RMSE"].mean()
    normal_avg  = sub[sub["period"]=="normal"]["rel_RMSE"].mean()

    ranking_rows.append({
        "Model":              DISPLAY_NAMES.get(model_name, model_name),
        "Category":           MODEL_REGISTRY[model_name]["category"],
        "Crisis RMSE (lag=0)":  round(crisis_rmse[0], 4) if len(crisis_rmse) > 0 else np.nan,
        "Normal RMSE (lag=0)":  round(normal_rmse[0], 4) if len(normal_rmse) > 0 else np.nan,
        "Full RMSE (lag=0)":    round(full_rmse[0],   4) if len(full_rmse)   > 0 else np.nan,
        "Crisis avg (all lags)": round(crisis_avg, 4),
        "Normal avg (all lags)": round(normal_avg, 4),
    })

ranking = pd.DataFrame(ranking_rows).sort_values("Crisis RMSE (lag=0)")
ranking.to_csv(OUT_DIR / "model_ranking.csv", index=False)

print("\n" + "="*70)
print("MODEL RANKING — Relative RMSE vs ARMA (< 1.0 = better than ARMA)")
print("="*70)
print(ranking.to_string(index=False))

# =============================================================================
# DONE
# =============================================================================
print("\n" + "="*70)
print("COMPARISON COMPLETE")
print("="*70)
print(f"Output: {OUT_DIR}")
print()
print("CSVs:")
print("  comparison_raw.csv       — all models x periods x vintages (raw)")
print("  unified_predictions.csv  — all model predictions in one file")
print("  rel_rmse_crisis.csv      — relative RMSE table, crisis period")
print("  rel_rmse_normal.csv      — relative RMSE table, normal period")
print("  rel_rmse_full.csv        — relative RMSE table, full period")
print("  rel_mae_crisis.csv       — relative MAE table, crisis period")
print("  rel_mae_normal.csv       — relative MAE table, normal period")
print("  model_ranking.csv        — summary ranking table")
print()
print("Plots:")
print("  plot1_rmse_bar_crisis_normal.png  — bar chart RMSE by model")
print("  plot2_heatmap_rel_rmse.png        — heatmap model x vintage")
print("  plot3_vintage_profiles.png        — RMSE improvement across vintages")
print("  plot4_actual_vs_predicted.png     — actual vs predicted top models")
print()
print("NEXT STEPS:")
print("  1. Run diebold_mariano_final.py   — significance tests")
print("  2. Run fanchart_final.py          — prediction intervals")
print("  3. Run crps_final.py              — density forecast evaluation")
print("  4. Run shap_final.py              — SHAP for XGBoost")
```

```python
# =============================================================================
# diebold_mariano_final.py
# Purpose : Formal Diebold-Mariano (1995) forecast comparison tests.
#           Tests whether differences in forecast accuracy between models
#           are statistically significant, or could be due to chance.
#
# What this script produces:
#   1. DM test: each model vs ARMA (benchmark)
#   2. DM test: each model vs Ridge (best linear ML)
#   3. DM test: each model vs DFM  (best overall)
#   4. Publication-ready tables with significance stars (* ** ***)
#   5. Heatmap of p-values across all model pairs
#
# WHY DM TESTS MATTER FOR THE PAPER:
#   Without significance tests, a referee can argue that observed RMSE
#   differences are within sampling noise. DM tests provide formal
#   statistical evidence that model A is significantly better than model B.
#   This is standard practice in nowcasting papers (Giannone et al. 2008,
#   Foroni & Marcellino 2014, etc.)
#
# THE DM TEST — INTUITION:
#   Given two forecast error series e1(t) and e2(t), define the loss
#   differential d(t) = L(e1(t)) - L(e2(t)) where L is a loss function
#   (here: squared error for RMSE, absolute error for MAE).
#   Under H0: E[d(t)] = 0 (equal predictive accuracy).
#   The DM statistic is: DM = d_bar / sqrt(V_hat / T)
#   where V_hat is a HAC (Newey-West) variance estimator.
#   Under H0, DM ~ N(0,1) asymptotically.
#   A negative DM stat means model 1 is BETTER than model 2 (lower loss).
#
# HARVEY, LEYBOURNE & NEWBOLD (1997) CORRECTION:
#   For small samples (T < 30), the standard DM test over-rejects H0.
#   The HLN correction adjusts for sample size and forecast horizon.
#   With T=22 quarters and T=10 in subsamples, this correction is important.
#
# Author  : Gabriela Aquino
# Input   : ~/Desktop/ESA-gdp-nowcasting/Models/results/comparison/
#             unified_predictions.csv
# Output  : ~/Desktop/ESA-gdp-nowcasting/Models/results/diebold_mariano/
# =============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import warnings
from pathlib import Path
from scipy import stats

warnings.filterwarnings("ignore")

# =============================================================================
# 0. PATHS
# =============================================================================

BASE_DIR  = Path.home() / "Desktop" / "ESA-gdp-nowcasting"
MODEL_DIR = BASE_DIR / "Models"
RES_DIR   = MODEL_DIR / "results"
IN_FILE   = RES_DIR / "comparison" / "unified_predictions.csv"
OUT_DIR   = RES_DIR / "diebold_mariano"
OUT_DIR.mkdir(parents=True, exist_ok=True)

if not IN_FILE.exists():
    raise FileNotFoundError(
        f"File not found: {IN_FILE}\n"
        "Run comparison_final.py first to generate unified_predictions.csv"
    )

# =============================================================================
# 1. CONFIGURATION
# =============================================================================

# --- FIXED -------------------------------------------------------------------
CRISIS_START = "2019-01-01"
CRISIS_END   = "2021-06-01"
NORMAL_START = "2022-01-01"
NORMAL_END   = "2024-06-01"

# Nowcast vintage only — this is the operationally relevant horizon
# All DM tests use lag=0 predictions
NOWCAST_LAG = 0

# Significance levels for stars
ALPHA_1  = 0.10   # *
ALPHA_5  = 0.05   # **
ALPHA_10 = 0.01   # ***

# Models to include in DM comparison
# Must match column name pattern in unified_predictions.csv:
# "{Display Name}_lag{lag}"
MODELS = ["ARMA", "DFM", "Ridge", "MIDAS", "Lasso",
          "Random Forest", "XGBoost"]

# Benchmark models for pairwise tables
BENCHMARKS = ["ARMA", "Ridge", "DFM"]

# =============================================================================
# 2. DIEBOLD-MARIANO IMPLEMENTATION
# =============================================================================

def dm_test(e1, e2, loss="squared", h=1, correction="HLN"):
    """
    Diebold-Mariano (1995) test for equal predictive accuracy.
    With Harvey-Leybourne-Newbold (1997) small-sample correction.

    Parameters
    ----------
    e1    : array-like — forecast errors from model 1 (actual - predicted)
    e2    : array-like — forecast errors from model 2 (actual - predicted)
    loss  : "squared" (RMSE-based) or "absolute" (MAE-based)
    h     : forecast horizon (1 for nowcast)
    correction : "HLN" (Harvey et al. 1997) or "DM" (original)

    Returns
    -------
    stat  : DM test statistic (negative = model 1 better)
    pval  : two-sided p-value
    mean_d: mean loss differential (negative = model 1 better)

    H0: E[d(t)] = 0 (equal predictive accuracy)
    H1: E[d(t)] != 0 (one model is better)

    Interpretation:
      stat < 0 and pval < 0.05 → model 1 significantly BETTER than model 2
      stat > 0 and pval < 0.05 → model 1 significantly WORSE than model 2
    """
    e1 = np.array(e1, dtype=float)
    e2 = np.array(e2, dtype=float)

    # Remove observations where either series has NaN
    mask = ~np.isnan(e1) & ~np.isnan(e2)
    e1, e2 = e1[mask], e2[mask]
    T = len(e1)

    if T < 6:
        return np.nan, np.nan, np.nan

    # Loss differential: d(t) = L(e1(t)) - L(e2(t))
    if loss == "squared":
        d = e1**2 - e2**2
    elif loss == "absolute":
        d = np.abs(e1) - np.abs(e2)
    else:
        raise ValueError("loss must be 'squared' or 'absolute'")

    mean_d = np.mean(d)

    # HAC variance estimator (Newey-West with h-1 lags)
    # For h=1 (nowcast), this simplifies to the sample variance
    gamma_0 = np.var(d, ddof=1)
    if h > 1:
        gamma_sum = 0
        for k in range(1, h):
            gamma_k = np.mean((d[k:] - mean_d) * (d[:-k] - mean_d))
            gamma_sum += (1 - k/h) * gamma_k
        V_hat = (gamma_0 + 2 * gamma_sum) / T
    else:
        V_hat = gamma_0 / T

    if V_hat <= 0:
        return np.nan, np.nan, mean_d

    dm_stat = mean_d / np.sqrt(V_hat)

    if correction == "HLN":
        # Harvey, Leybourne & Newbold (1997) correction
        # Adjusts for small sample and forecast horizon
        hln_factor = np.sqrt(
            (T + 1 - 2*h + h*(h-1)/T) / T
        )
        dm_stat_hln = dm_stat * hln_factor
        # Under H0, HLN statistic follows t(T-1)
        pval = 2 * stats.t.sf(np.abs(dm_stat_hln), df=T-1)
        return dm_stat_hln, pval, mean_d
    else:
        pval = 2 * stats.norm.sf(np.abs(dm_stat))
        return dm_stat, pval, mean_d


def stars(pval):
    """Convert p-value to significance stars."""
    if pd.isna(pval):
        return ""
    if pval < ALPHA_10:
        return "***"
    if pval < ALPHA_5:
        return "**"
    if pval < ALPHA_1:
        return "*"
    return ""


def format_stat(stat, pval):
    """Format DM statistic with stars."""
    if pd.isna(stat):
        return "—"
    s = f"{stat:.3f}{stars(pval)}"
    return s

# =============================================================================
# 3. LOAD DATA
# =============================================================================

print("Loading unified predictions...")
df = pd.read_csv(IN_FILE, parse_dates=["date"])
df = df.sort_values("date").reset_index(drop=True)

print(f"  Rows    : {len(df)}")
print(f"  Columns : {list(df.columns)}")

# Build period masks
dates_dt = pd.to_datetime(df["date"])
mask_full   = np.ones(len(df), dtype=bool)
mask_crisis = (dates_dt >= CRISIS_START) & (dates_dt <= CRISIS_END)
mask_normal = (dates_dt >= NORMAL_START) & (dates_dt <= NORMAL_END)

actuals = df["actual"].values

# Build model prediction columns
# Column names in unified_predictions.csv: "{Model}_lag{lag}"
lag = NOWCAST_LAG
model_preds = {}
for model in MODELS:
    col = f"{model}_lag{lag}"
    if col in df.columns:
        model_preds[model] = df[col].values
        print(f"  [{model}] lag={lag} found: {col}")
    else:
        # Try alternative formats
        alts = [f"{model}_lag+{lag}", f"{model}_lag-{abs(lag)}",
                f"{model}_lag_{lag}"]
        found = False
        for alt in alts:
            if alt in df.columns:
                model_preds[model] = df[alt].values
                print(f"  [{model}] lag={lag} found via alt: {alt}")
                found = True
                break
        if not found:
            print(f"  [{model}] WARNING: lag={lag} column not found — skipping")

available_models = list(model_preds.keys())
print(f"\nModels available for DM tests: {available_models}")

# =============================================================================
# 4. COMPUTE FORECAST ERRORS
# =============================================================================

# Forecast error = actual - predicted
errors = {}
for model in available_models:
    errors[model] = actuals - model_preds[model]

# =============================================================================
# 5. DM TESTS — MODEL VS BENCHMARK
# =============================================================================

def run_dm_table(benchmark, models, mask, period_label):
    """
    Run DM tests comparing each model against a benchmark.
    Returns a DataFrame with DM stats and p-values.
    """
    y = actuals[mask]
    e_bench = errors[benchmark][mask]

    rows = []
    for model in models:
        if model == benchmark or model not in errors:
            continue
        e_model = errors[model][mask]
        T = (~np.isnan(e_model) & ~np.isnan(e_bench)).sum()

        dm_sq,  pv_sq,  md_sq  = dm_test(e_model, e_bench, loss="squared")
        dm_abs, pv_abs, md_abs = dm_test(e_model, e_bench, loss="absolute")

        rows.append({
            "Model":           model,
            "vs":              benchmark,
            "Period":          period_label,
            "T":               T,
            "DM (RMSE)":       round(dm_sq,  3) if not pd.isna(dm_sq)  else np.nan,
            "p-val (RMSE)":    round(pv_sq,  4) if not pd.isna(pv_sq)  else np.nan,
            "DM (MAE)":        round(dm_abs, 3) if not pd.isna(dm_abs) else np.nan,
            "p-val (MAE)":     round(pv_abs, 4) if not pd.isna(pv_abs) else np.nan,
            "Stars (RMSE)":    stars(pv_sq),
            "Stars (MAE)":     stars(pv_abs),
            "Better?":         "YES" if (not pd.isna(dm_sq) and dm_sq < 0)
                               else ("NO" if not pd.isna(dm_sq) else "—"),
        })
    return pd.DataFrame(rows)


print("\n" + "="*70)
print("DIEBOLD-MARIANO TESTS — ALL PERIODS, ALL BENCHMARKS")
print("="*70)
print("H0: equal predictive accuracy")
print("H1: model is significantly different from benchmark")
print("Negative DM stat = model is BETTER than benchmark")
print("Stars: * p<0.10  ** p<0.05  *** p<0.01  (HLN small-sample correction)")
print()

all_dm_results = []

for benchmark in BENCHMARKS:
    if benchmark not in available_models:
        print(f"Benchmark {benchmark} not available — skipping")
        continue

    for mask, period_label in [
        (mask_full,   "Full (2019Q1-2024Q2)"),
        (mask_crisis, "Crisis (2019Q1-2021Q2)"),
        (mask_normal, "Normal (2022Q1-2024Q2)"),
    ]:
        tbl = run_dm_table(benchmark, available_models, mask, period_label)
        all_dm_results.append(tbl)

        print(f"\n--- vs {benchmark} | {period_label} ---")
        display_cols = ["Model", "T", "DM (RMSE)", "Stars (RMSE)",
                        "p-val (RMSE)", "DM (MAE)", "Stars (MAE)", "Better?"]
        print(tbl[display_cols].to_string(index=False))

# Combine all results
dm_all = pd.concat(all_dm_results, ignore_index=True)
dm_all.to_csv(OUT_DIR / "dm_all_results.csv", index=False)
print(f"\nAll DM results saved: dm_all_results.csv")

# =============================================================================
# 6. PUBLICATION TABLE — vs ARMA, key periods
# =============================================================================

def make_pub_table(dm_df, benchmark, loss="RMSE"):
    """
    Create publication-ready table: models as rows, periods as columns.
    Each cell shows DM stat + significance stars.
    """
    periods = ["Full (2019Q1-2024Q2)",
               "Crisis (2019Q1-2021Q2)",
               "Normal (2022Q1-2024Q2)"]

    sub = dm_df[dm_df["vs"] == benchmark].copy()
    models_in = sub["Model"].unique()

    rows = []
    for model in MODELS:
        if model == benchmark or model not in models_in:
            continue
        row = {"Model": model}
        for period in periods:
            p_sub = sub[(sub["Model"] == model) & (sub["Period"] == period)]
            if len(p_sub) == 0:
                row[period] = "—"
            else:
                stat = p_sub[f"DM ({loss})"].values[0]
                pval = p_sub[f"p-val ({loss})"].values[0]
                row[period] = format_stat(stat, pval)
        rows.append(row)

    return pd.DataFrame(rows)

# Print publication tables
for benchmark in BENCHMARKS:
    if benchmark not in available_models:
        continue
    for loss in ["RMSE", "MAE"]:
        tbl = make_pub_table(dm_all, benchmark, loss)
        print(f"\n{'='*70}")
        print(f"DM TEST TABLE — vs {benchmark} | Loss: {loss}")
        print(f"Negative stat = model better than {benchmark}")
        print(f"{'='*70}")
        print(tbl.to_string(index=False))
        tbl.to_csv(OUT_DIR / f"dm_table_vs{benchmark}_{loss}.csv", index=False)

# =============================================================================
# 7. FULL PAIRWISE DM MATRIX (all models vs all models)
# =============================================================================

def pairwise_dm_matrix(models, mask, loss="squared"):
    """
    Compute DM statistic for all model pairs.
    Entry [i,j] = DM stat when testing model i vs model j.
    Negative = model i is BETTER than model j.
    """
    n = len(models)
    stat_matrix = np.full((n, n), np.nan)
    pval_matrix = np.full((n, n), np.nan)

    for i, m1 in enumerate(models):
        for j, m2 in enumerate(models):
            if i == j or m1 not in errors or m2 not in errors:
                continue
            e1 = errors[m1][mask]
            e2 = errors[m2][mask]
            dm, pv, _ = dm_test(e1, e2, loss=loss)
            stat_matrix[i, j] = dm
            pval_matrix[i, j] = pv

    return (pd.DataFrame(stat_matrix, index=models, columns=models),
            pd.DataFrame(pval_matrix, index=models, columns=models))

# =============================================================================
# 8. PLOTS
# =============================================================================

# ---- PLOT 1: DM stats vs ARMA, crisis vs normal side by side ---------------

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle(
    "Diebold-Mariano Test Statistics vs ARMA Benchmark\n"
    "Negative = model significantly better than ARMA | "
    "Stars: * p<0.10  ** p<0.05  *** p<0.01",
    fontsize=12, fontweight="bold"
)

for ax, period_label, title in zip(
    axes,
    ["Crisis (2019Q1-2021Q2)", "Normal (2022Q1-2024Q2)"],
    ["Crisis Period (2019Q1-2021Q2)", "Non-Crisis Period (2022Q1-2024Q2)"]
):
    sub = dm_all[
        (dm_all["vs"] == "ARMA") &
        (dm_all["Period"] == period_label)
    ].copy()

    if len(sub) == 0:
        ax.text(0.5, 0.5, "No data", ha="center", transform=ax.transAxes)
        continue

    sub = sub.sort_values("DM (RMSE)")
    colors = ["#2a9d8f" if v < 0 else "#e63946"
              for v in sub["DM (RMSE)"]]

    bars = ax.barh(sub["Model"], sub["DM (RMSE)"],
                   color=colors, edgecolor="gray", linewidth=0.5)

    # Add stars as text
    for bar, (_, row) in zip(bars, sub.iterrows()):
        x = row["DM (RMSE)"]
        s = row["Stars (RMSE)"]
        if s:
            ax.text(x + (0.05 if x >= 0 else -0.05),
                    bar.get_y() + bar.get_height()/2,
                    s, va="center",
                    ha="left" if x >= 0 else "right",
                    fontsize=11, fontweight="bold")

    ax.axvline(0, color="#333333", lw=1.5, ls="--")
    ax.set_title(title, fontsize=11, fontweight="bold")
    ax.set_xlabel("DM Statistic (negative = better than ARMA)")
    ax.grid(True, axis="x", alpha=0.3)

plt.tight_layout()
fig.savefig(OUT_DIR / "dm_plot1_vs_arma.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: dm_plot1_vs_arma.png")

# ---- PLOT 2: Pairwise DM heatmap — crisis period ----------------------------

for mask, period_label, fname in [
    (mask_crisis, "Crisis (2019Q1-2021Q2)",       "crisis"),
    (mask_normal, "Normal (2022Q1-2024Q2)",        "normal"),
    (mask_full,   "Full Period (2019Q1-2024Q2)",   "full"),
]:
    stat_df, pval_df = pairwise_dm_matrix(available_models, mask, "squared")

    # Annotation: DM stat + stars
    annot = stat_df.copy().astype(object)
    for i in range(len(available_models)):
        for j in range(len(available_models)):
            if i == j:
                annot.iloc[i, j] = "—"
            else:
                v = stat_df.iloc[i, j]
                p = pval_df.iloc[i, j]
                if pd.isna(v):
                    annot.iloc[i, j] = "—"
                else:
                    annot.iloc[i, j] = f"{v:.2f}{stars(p)}"

    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(
        stat_df.astype(float),
        annot=annot, fmt="",
        cmap="RdYlGn_r",
        center=0, vmin=-4, vmax=4,
        linewidths=0.5, ax=ax,
        cbar_kws={"label": "DM Statistic", "shrink": 0.8}
    )
    ax.set_title(
        f"Pairwise DM Test Matrix — {period_label}\n"
        "Entry [i,j]: DM stat testing row model vs column model\n"
        "Negative (green) = row model BETTER than column model",
        fontsize=11, fontweight="bold", pad=10
    )
    ax.set_xlabel("Benchmark model (column)")
    ax.set_ylabel("Test model (row)")
    plt.tight_layout()
    fig.savefig(OUT_DIR / f"dm_plot2_pairwise_{fname}.png",
                dpi=150, bbox_inches="tight")
    plt.show()
    print(f"Saved: dm_plot2_pairwise_{fname}.png")

# =============================================================================
# 9. SUMMARY STATISTICS FOR PAPER
# =============================================================================

print("\n" + "="*70)
print("SUMMARY FOR PAPER — DM TESTS vs ARMA (lag=0, HLN corrected)")
print("="*70)
print()
print("CRISIS PERIOD (2019Q1-2021Q2, T=10 quarters):")
crisis_sum = dm_all[
    (dm_all["vs"] == "ARMA") &
    (dm_all["Period"] == "Crisis (2019Q1-2021Q2)")
][["Model", "DM (RMSE)", "p-val (RMSE)", "Stars (RMSE)",
   "DM (MAE)", "p-val (MAE)", "Stars (MAE)"]].sort_values("DM (RMSE)")
print(crisis_sum.to_string(index=False))

print()
print("NORMAL PERIOD (2022Q1-2024Q2, T=10 quarters):")
normal_sum = dm_all[
    (dm_all["vs"] == "ARMA") &
    (dm_all["Period"] == "Normal (2022Q1-2024Q2)")
][["Model", "DM (RMSE)", "p-val (RMSE)", "Stars (RMSE)",
   "DM (MAE)", "p-val (MAE)", "Stars (MAE)"]].sort_values("DM (RMSE)")
print(normal_sum.to_string(index=False))

print()
print("NOTE ON SMALL SAMPLES:")
print("  With T=10 quarters per subsample, the HLN correction is critical.")
print("  Low significance (few stars) does not mean the model is bad —")
print("  it reflects limited statistical power with only 10 observations.")
print("  The full period (T=22) provides more reliable inference.")

# =============================================================================
# DONE
# =============================================================================
print("\n" + "="*70)
print("DIEBOLD-MARIANO TESTS COMPLETE")
print("="*70)
print(f"Output: {OUT_DIR}")
print()
print("CSVs:")
print("  dm_all_results.csv              — all DM tests, all periods")
print("  dm_table_vsARMA_RMSE.csv        — publication table vs ARMA")
print("  dm_table_vsRidge_RMSE.csv       — publication table vs Ridge")
print("  dm_table_vsDFM_RMSE.csv         — publication table vs DFM")
print()
print("Plots:")
print("  dm_plot1_vs_arma.png            — bar chart DM stats vs ARMA")
print("  dm_plot2_pairwise_crisis.png    — pairwise matrix, crisis")
print("  dm_plot2_pairwise_normal.png    — pairwise matrix, normal")
print("  dm_plot2_pairwise_full.png      — pairwise matrix, full period")
print()
print("NEXT: Run fanchart_final.py for prediction intervals.")
```

```python
# =============================================================================
# fanchart_final.py
# Purpose : Bootstrap prediction intervals and fan charts for the top 3 models.
#           Moves the analysis from point forecasts to uncertainty quantification.
#
# What this script produces:
#   1. Bootstrap residual intervals (68%, 80%, 95%) for Ridge, DFM, MIDAS, ARMA
#   2. Fan chart — crisis period (2019Q1-2021Q2)
#   3. Fan chart — normal period (2022Q1-2024Q2)
#   4. Fan chart — full period (2019Q1-2024Q2)
#   5. Interval width comparison across models
#   6. Empirical coverage table (how often actual falls inside the interval)
#
# WHY FAN CHARTS MATTER FOR THE PAPER:
#   Point forecasts answer "what is the most likely GDP growth?"
#   Fan charts answer "how uncertain are we about that estimate?"
#   A model can have low RMSE but poorly calibrated intervals (overconfident).
#   Coverage analysis tests whether the stated confidence levels are honest.
#
# THE BOOTSTRAP METHOD:
#   For each model at each rolling origin t:
#     1. Collect all past rolling-origin residuals: e(1), ..., e(t-1)
#     2. Resample B bootstrap samples from those residuals
#     3. Add resampled residuals to the point forecast: y_hat(t) + e*(b)
#     4. The quantiles of those B samples form the prediction interval
#
#   This is the "residual bootstrap" — it is:
#     - Non-parametric: no Gaussian assumption needed
#     - Pseudo-real-time: uses only past residuals available at each origin
#     - Simple to implement and explain to non-technical audiences
#
#   For the DFM (R model), intervals are approximated using the
#   full-period residuals since rolling residuals are pre-computed.
#
# CALIBRATION (COVERAGE):
#   A 95% interval is well-calibrated if ~95% of actual values fall inside.
#   If coverage < 95% → intervals are too narrow (overconfident)
#   If coverage > 95% → intervals are too wide (underconfident)
#
# Author  : Gabriela Aquino
# Input   : ~/Desktop/ESA-gdp-nowcasting/Models/results/comparison/
#             unified_predictions.csv
# Output  : ~/Desktop/ESA-gdp-nowcasting/Models/results/fanchart/
# =============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.dates as mdates
import warnings
from pathlib import Path

warnings.filterwarnings("ignore")
np.random.seed(42)

# =============================================================================
# 0. PATHS
# =============================================================================

BASE_DIR = Path.home() / "Desktop" / "ESA-gdp-nowcasting"
MODEL_DIR = BASE_DIR / "Models"
RES_DIR   = MODEL_DIR / "results"
IN_FILE   = RES_DIR / "comparison" / "unified_predictions.csv"
OUT_DIR   = RES_DIR / "fanchart"
OUT_DIR.mkdir(parents=True, exist_ok=True)

if not IN_FILE.exists():
    raise FileNotFoundError(
        f"File not found: {IN_FILE}\n"
        "Run comparison_final.py first."
    )

# =============================================================================
# 1. CONFIGURATION
# =============================================================================

# Bootstrap settings
N_BOOTSTRAP  = 2000    # number of bootstrap samples per origin
RANDOM_SEED  = 42

# Confidence levels for fan chart bands
# 68% ≈ ±1 std, 80%, 95% are standard in central bank fan charts
LEVELS = [0.68, 0.80, 0.95]

# Colors for fan chart bands (light to dark)
# Each model gets its own color family
MODEL_STYLES = {
    "Ridge": {
        "point_color":  "#1a7c3e",
        "band_colors":  ["#c8ecd4", "#84cfa0", "#40b36c"],
        "label":        "Ridge",
    },
    "DFM": {
        "point_color":  "#0077b6",
        "band_colors":  ["#bde0f5", "#5fb4e0", "#0096d6"],
        "label":        "DFM",
    },
    "MIDAS": {
        "point_color":  "#e07b00",
        "band_colors":  ["#fde8c8", "#f8c479", "#f5a030"],
        "label":        "MIDAS",
    },
    "ARMA": {
        "point_color":  "#6c757d",
        "band_colors":  ["#e8e9ea", "#c2c4c6", "#9ca0a3"],
        "label":        "ARMA (benchmark)",
    },
}

# Period definitions
CRISIS_START = "2019-01-01"
CRISIS_END   = "2021-06-01"
NORMAL_START = "2022-01-01"
NORMAL_END   = "2024-06-01"

# Nowcast vintage
LAG = 0

# =============================================================================
# 2. LOAD DATA
# =============================================================================

print("Loading unified predictions...")
df = pd.read_csv(IN_FILE, parse_dates=["date"])
df = df.sort_values("date").reset_index(drop=True)

actuals = df["actual"].values
dates   = pd.to_datetime(df["date"]).values

# Load predictions for lag=0
preds = {}
for model in MODEL_STYLES.keys():
    col = f"{model}_lag{LAG}"
    if col in df.columns:
        preds[model] = df[col].values
        print(f"  [{model}] loaded: {col}")
    else:
        print(f"  [{model}] WARNING: column {col} not found — skipping")

available_models = list(preds.keys())
print(f"\nModels available for fan charts: {available_models}")

# =============================================================================
# 3. BOOTSTRAP PREDICTION INTERVALS
# =============================================================================
# For each nowcast origin t, we:
#   1. Collect residuals from all previous origins (1 to t-1)
#   2. Bootstrap resample B times from those residuals
#   3. Add to the point forecast to get the predictive distribution
#   4. Extract quantiles for each confidence level
#
# This is pseudo-real-time: at each origin, only past residuals are used.

def bootstrap_intervals(actuals, predictions, levels, n_boot=2000, seed=42):
    """
    Compute bootstrap prediction intervals for each forecast origin.

    Returns dict with keys = level (e.g. 0.95), values = (lower, upper) arrays.
    """
    rng = np.random.default_rng(seed)
    T   = len(predictions)

    # Compute forecast errors (residuals)
    errors = actuals - predictions   # shape (T,)

    # Storage: lower and upper bounds for each level
    intervals = {lev: {"lower": np.full(T, np.nan),
                        "upper": np.full(T, np.nan)}
                 for lev in levels}

    for t in range(T):
        yhat = predictions[t]
        if np.isnan(yhat):
            continue

        # Use all non-NaN past residuals available at origin t
        # Minimum 5 past residuals required for bootstrap
        past_errors = errors[:t]
        past_errors = past_errors[~np.isnan(past_errors)]

        if len(past_errors) < 5:
            # Not enough history — use symmetric interval based on past RMSE
            if len(past_errors) > 0:
                rmse = np.sqrt(np.mean(past_errors**2))
                for lev in levels:
                    z = {0.68: 1.00, 0.80: 1.28, 0.95: 1.96}.get(lev, 1.96)
                    intervals[lev]["lower"][t] = yhat - z * rmse
                    intervals[lev]["upper"][t] = yhat + z * rmse
            continue

        # Bootstrap: resample past residuals B times
        boot_samples = rng.choice(past_errors, size=(n_boot,), replace=True)
        boot_forecasts = yhat + boot_samples

        # Extract quantile bounds for each confidence level
        for lev in levels:
            alpha = 1 - lev
            intervals[lev]["lower"][t] = np.quantile(boot_forecasts, alpha/2)
            intervals[lev]["upper"][t] = np.quantile(boot_forecasts, 1 - alpha/2)

    return intervals


print("\nComputing bootstrap prediction intervals...")
print(f"  Bootstrap samples: {N_BOOTSTRAP}")
print(f"  Confidence levels: {[f'{int(l*100)}%' for l in LEVELS]}")
print()

all_intervals = {}
for model in available_models:
    print(f"  [{model}]...", end=" ", flush=True)
    all_intervals[model] = bootstrap_intervals(
        actuals, preds[model],
        levels=LEVELS,
        n_boot=N_BOOTSTRAP,
        seed=RANDOM_SEED
    )
    print("done")

# =============================================================================
# 4. COVERAGE ANALYSIS
# =============================================================================
# Empirical coverage = fraction of actual values inside the interval.
# Well-calibrated: 95% interval should cover ~95% of actuals.

def compute_coverage(actuals, predictions, intervals, mask=None):
    """Compute empirical coverage for each confidence level."""
    if mask is None:
        mask = np.ones(len(actuals), dtype=bool)

    y    = actuals[mask]
    rows = []
    for lev in LEVELS:
        lo = intervals[lev]["lower"][mask]
        hi = intervals[lev]["upper"][mask]
        valid = ~np.isnan(y) & ~np.isnan(lo) & ~np.isnan(hi)
        if valid.sum() == 0:
            rows.append({"Level": f"{int(lev*100)}%", "Coverage": np.nan, "n": 0})
            continue
        covered = ((y[valid] >= lo[valid]) & (y[valid] <= hi[valid])).mean()
        rows.append({
            "Level":    f"{int(lev*100)}%",
            "Coverage": round(covered, 4),
            "n":        valid.sum()
        })
    return pd.DataFrame(rows)

# Build period masks
dates_dt    = pd.to_datetime(dates)
mask_full   = np.ones(len(actuals), dtype=bool)
mask_crisis = (dates_dt >= CRISIS_START) & (dates_dt <= CRISIS_END)
mask_normal = (dates_dt >= NORMAL_START) & (dates_dt <= NORMAL_END)

print("\n" + "="*60)
print("EMPIRICAL COVERAGE ANALYSIS")
print("="*60)
print("Well-calibrated: coverage ≈ stated confidence level")
print("Under-coverage  : intervals too narrow (overconfident)")
print("Over-coverage   : intervals too wide  (conservative)")
print()

coverage_rows = []
for model in available_models:
    for mask, period_label in [
        (mask_full,   "Full"),
        (mask_crisis, "Crisis"),
        (mask_normal, "Normal"),
    ]:
        cov_df = compute_coverage(actuals, preds[model],
                                  all_intervals[model], mask)
        for _, row in cov_df.iterrows():
            coverage_rows.append({
                "Model":    model,
                "Period":   period_label,
                "Level":    row["Level"],
                "Coverage": row["Coverage"],
                "n":        row["n"],
            })

coverage_df = pd.DataFrame(coverage_rows)
coverage_df.to_csv(OUT_DIR / "coverage_table.csv", index=False)

# Print coverage table
for period in ["Full", "Crisis", "Normal"]:
    print(f"\n--- {period} Period ---")
    sub = coverage_df[coverage_df["Period"] == period].pivot(
        index="Model", columns="Level", values="Coverage"
    )
    print(sub.round(3).to_string())

# =============================================================================
# 5. INTERVAL WIDTH COMPARISON
# =============================================================================

print("\n" + "="*60)
print("AVERAGE INTERVAL WIDTH AT 95% LEVEL")
print("="*60)
for model in available_models:
    lo = all_intervals[model][0.95]["lower"]
    hi = all_intervals[model][0.95]["upper"]
    width = hi - lo
    valid = ~np.isnan(width)
    print(f"  {model:15s}: mean width = {np.nanmean(width):.4f}  "
          f"| crisis = {np.nanmean(width[mask_crisis]):.4f}  "
          f"| normal = {np.nanmean(width[mask_normal]):.4f}")

# =============================================================================
# 6. FAN CHART PLOTTING FUNCTION
# =============================================================================

def plot_fanchart(model, mask, period_title, filename,
                  show_actuals_full=True):
    """
    Plot a publication-ready fan chart for one model and one period.

    Parameters
    ----------
    model         : model name (key in MODEL_STYLES)
    mask          : boolean array for evaluation period
    period_title  : string for plot title
    filename      : output PNG filename
    show_actuals_full : if True, show full GDP series in background
    """
    style  = MODEL_STYLES[model]
    ints   = all_intervals[model]
    dates_plot = pd.to_datetime(dates)

    # Full GDP series for context
    full_gdp_dates = dates_plot[~np.isnan(actuals)]
    full_gdp_vals  = actuals[~np.isnan(actuals)]

    # Evaluation period data
    eval_dates = dates_plot[mask]
    eval_acts  = actuals[mask]
    eval_preds = preds[model][mask]

    fig, ax = plt.subplots(figsize=(14, 6))

    # --- Shade evaluation window ---
    if "crisis" in filename.lower():
        ax.axvspan(pd.to_datetime(CRISIS_START), pd.to_datetime(CRISIS_END),
                   alpha=0.07, color="#e63946", zorder=0)
    elif "normal" in filename.lower():
        ax.axvspan(pd.to_datetime(NORMAL_START), pd.to_datetime(NORMAL_END),
                   alpha=0.07, color="#2a9d8f", zorder=0)

    # --- Bootstrap bands (widest to narrowest) ---
    levels_sorted = sorted(LEVELS, reverse=True)
    for i, lev in enumerate(levels_sorted):
        lo = ints[lev]["lower"]
        hi = ints[lev]["upper"]
        # Mask to evaluation period
        lo_eval = lo[mask]
        hi_eval = hi[mask]
        valid   = ~np.isnan(lo_eval) & ~np.isnan(hi_eval)

        ax.fill_between(
            eval_dates[valid],
            lo_eval[valid],
            hi_eval[valid],
            color=style["band_colors"][i],
            alpha=0.85,
            label=f"{int(lev*100)}% interval",
            zorder=1
        )

    # --- Full GDP series (light background) ---
    if show_actuals_full:
        ax.plot(full_gdp_dates, full_gdp_vals,
                color="#cccccc", lw=1.2, marker="o", markersize=2,
                alpha=0.6, zorder=2, label="GDP full series (context)")

    # --- Point forecast ---
    valid_pred = ~np.isnan(eval_preds)
    ax.plot(eval_dates[valid_pred], eval_preds[valid_pred],
            color=style["point_color"], lw=2.2,
            marker="o", markersize=5,
            label=f"{style['label']} — point forecast",
            zorder=4)

    # --- Actual GDP (evaluation period) ---
    valid_act = ~np.isnan(eval_acts)
    ax.plot(eval_dates[valid_act], eval_acts[valid_act],
            color="#111111", lw=2.5,
            marker="o", markersize=6,
            label="Actual GDP (QoQ)",
            zorder=5)

    # --- Formatting ---
    ax.axhline(0, color="#999999", lw=0.8, ls="--", alpha=0.5)
    ax.set_xlabel("Date", fontsize=11)
    ax.set_ylabel("GDP growth rate (QoQ)", fontsize=11)
    ax.set_title(
        f"Fan Chart — {style['label']} | {period_title}\n"
        f"Bootstrap prediction intervals (68%, 80%, 95%) | "
        f"El Salvador GDP Nowcasting",
        fontsize=12, fontweight="bold"
    )
    ax.legend(fontsize=9, loc="upper left", ncol=2)
    ax.grid(True, alpha=0.25)
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))

    plt.tight_layout()
    fig.savefig(OUT_DIR / filename, dpi=150, bbox_inches="tight")
    plt.show()
    print(f"  Saved: {filename}")
    plt.close()


# =============================================================================
# 7. GENERATE FAN CHARTS
# =============================================================================

print("\nGenerating fan charts...")

# --- Individual model fan charts ---
for model in available_models:
    # Crisis period
    plot_fanchart(
        model, mask_crisis,
        "Crisis Period (2019Q1-2021Q2)",
        f"fanchart_{model.lower().replace(' ','_')}_crisis.png"
    )
    # Normal period
    plot_fanchart(
        model, mask_normal,
        "Non-Crisis Period (2022Q1-2024Q2)",
        f"fanchart_{model.lower().replace(' ','_')}_normal.png"
    )
    # Full period
    plot_fanchart(
        model, mask_full,
        "Full Evaluation Period (2019Q1-2024Q2)",
        f"fanchart_{model.lower().replace(' ','_')}_full.png"
    )

# =============================================================================
# 8. COMPARISON PLOT — All models side by side at 95% level
# =============================================================================

print("\nGenerating comparison fan chart (all models, 95% level)...")

for mask, period_label, period_title, fname in [
    (mask_crisis, "crisis", "Crisis Period (2019Q1-2021Q2)",
     "fanchart_comparison_crisis.png"),
    (mask_normal, "normal", "Non-Crisis Period (2022Q1-2024Q2)",
     "fanchart_comparison_normal.png"),
]:
    n_models = len(available_models)
    fig, axes = plt.subplots(1, n_models,
                             figsize=(5 * n_models, 6),
                             sharey=True)
    if n_models == 1:
        axes = [axes]

    fig.suptitle(
        f"Prediction Intervals Comparison — 95% Level | {period_title}\n"
        "Bootstrap residual method — shaded area = 95% predictive interval",
        fontsize=12, fontweight="bold"
    )

    dates_plot = pd.to_datetime(dates)
    eval_dates = dates_plot[mask]
    eval_acts  = actuals[mask]

    for ax, model in zip(axes, available_models):
        style  = MODEL_STYLES[model]
        lo = all_intervals[model][0.95]["lower"][mask]
        hi = all_intervals[model][0.95]["upper"][mask]
        pt = preds[model][mask]
        valid = ~np.isnan(lo) & ~np.isnan(hi)

        ax.fill_between(eval_dates[valid], lo[valid], hi[valid],
                        color=style["band_colors"][2],
                        alpha=0.4, label="95% interval")
        ax.plot(eval_dates[~np.isnan(pt)], pt[~np.isnan(pt)],
                color=style["point_color"], lw=2,
                marker="o", markersize=4, label="Point forecast")
        ax.plot(eval_dates[~np.isnan(eval_acts)],
                eval_acts[~np.isnan(eval_acts)],
                color="#111111", lw=2, marker="o", markersize=4,
                label="Actual GDP")
        ax.axhline(0, color="#aaaaaa", lw=0.8, ls="--")
        ax.set_title(style["label"], fontsize=11, fontweight="bold")
        ax.grid(True, alpha=0.25)
        ax.xaxis.set_major_locator(mdates.YearLocator())
        ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
        ax.tick_params(axis="x", rotation=30)
        if ax == axes[0]:
            ax.set_ylabel("GDP growth rate (QoQ)", fontsize=10)
            ax.legend(fontsize=8)

    plt.tight_layout()
    fig.savefig(OUT_DIR / fname, dpi=150, bbox_inches="tight")
    plt.show()
    print(f"  Saved: {fname}")
    plt.close()

# =============================================================================
# 9. INTERVAL WIDTH OVER TIME — shows uncertainty spikes during COVID
# =============================================================================

fig, ax = plt.subplots(figsize=(14, 5))
fig.suptitle(
    "95% Prediction Interval Width Over Time\n"
    "Wider intervals = more uncertainty | "
    "COVID shock should generate a spike in 2020",
    fontsize=12, fontweight="bold"
)

dates_plot = pd.to_datetime(dates)
for model in available_models:
    lo    = all_intervals[model][0.95]["lower"]
    hi    = all_intervals[model][0.95]["upper"]
    width = hi - lo
    valid = ~np.isnan(width)
    style = MODEL_STYLES[model]
    ax.plot(dates_plot[valid], width[valid],
            color=style["point_color"], lw=2,
            marker="o", markersize=4,
            label=style["label"])

ax.axvspan(pd.to_datetime(CRISIS_START), pd.to_datetime(CRISIS_END),
           alpha=0.10, color="#e63946", label="Crisis window")
ax.axvspan(pd.to_datetime(NORMAL_START), pd.to_datetime(NORMAL_END),
           alpha=0.08, color="#2a9d8f", label="Normal window")
ax.set_xlabel("Date")
ax.set_ylabel("95% Interval Width")
ax.legend(fontsize=9, ncol=3)
ax.grid(True, alpha=0.25)
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
plt.tight_layout()
fig.savefig(OUT_DIR / "fanchart_interval_width.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("  Saved: fanchart_interval_width.png")

# =============================================================================
# 10. SAVE COVERAGE TABLE
# =============================================================================

coverage_df.to_csv(OUT_DIR / "coverage_table.csv", index=False)
print(f"\nAll outputs saved to: {OUT_DIR}")

# =============================================================================
# DONE
# =============================================================================
print("\n" + "="*60)
print("FAN CHARTS COMPLETE")
print("="*60)
print(f"Output: {OUT_DIR}")
print()
print("Plots per model (Ridge, DFM, MIDAS, ARMA):")
print("  fanchart_{model}_crisis.png")
print("  fanchart_{model}_normal.png")
print("  fanchart_{model}_full.png")
print()
print("Comparison plots:")
print("  fanchart_comparison_crisis.png")
print("  fanchart_comparison_normal.png")
print("  fanchart_interval_width.png")
print()
print("CSVs:")
print("  coverage_table.csv — empirical coverage by model/period/level")
print()
print("NEXT: Run crps_final.py for density forecast evaluation.")
```

```python
# =============================================================================
# crps_final.py
# Purpose : Continuous Ranked Probability Score (CRPS) evaluation.
#           Extends the analysis from point forecasts to density forecasts.
#
# What this script produces:
#   1. CRPS scores for each model, period, and vintage
#   2. Relative CRPS table (vs ARMA = 1) — same format as RMSE tables
#   3. Skill score (CRPSS) — how much better than ARMA in density terms
#   4. Decomposition: reliability vs resolution
#   5. Publication-ready CRPS comparison table with significance
#   6. Bar chart: CRPS by model, crisis vs normal
#
# WHY CRPS MATTERS FOR THE PAPER:
#   RMSE evaluates point forecasts only — it ignores whether the model's
#   uncertainty estimates are well-calibrated. CRPS evaluates the FULL
#   predictive distribution in a single number.
#
#   CRPS = 0 means perfect forecast (point mass on the true value)
#   CRPS = RMSE when the forecast is a point forecast (delta distribution)
#   CRPS < RMSE when the predictive distribution adds value over the point
#
#   A model that wins on RMSE but loses on CRPS is overconfident:
#   its point forecast is good but its uncertainty is poorly calibrated.
#
# GAUSSIAN CRPS (closed form):
#   For a Gaussian predictive distribution N(mu, sigma²):
#
#   CRPS(N(mu, sigma), y) = sigma * [
#       (y - mu)/sigma * (2*Phi((y-mu)/sigma) - 1)
#       + 2*phi((y-mu)/sigma)
#       - 1/sqrt(pi)
#   ]
#
#   where Phi = standard normal CDF, phi = standard normal PDF.
#
#   mu    = point forecast (from rolling-origin predictions)
#   sigma = standard deviation of bootstrap predictive distribution
#           (estimated from rolling-origin residuals)
#
# This is the standard approximation used in:
#   - Gneiting & Raftery (2007, JASA)
#   - Knüppel (2015, IJF)
#   - Clark & McCracken (2013, Fed Reserve)
#
# Author  : Gabriela Aquino
# Input   : ~/Desktop/ESA-gdp-nowcasting/Models/results/comparison/
#             unified_predictions.csv
# Output  : ~/Desktop/ESA-gdp-nowcasting/Models/results/crps/
# =============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from pathlib import Path
from scipy import stats

warnings.filterwarnings("ignore")
np.random.seed(42)

# =============================================================================
# 0. PATHS
# =============================================================================

BASE_DIR  = Path.home() / "Desktop" / "ESA-gdp-nowcasting"
MODEL_DIR = BASE_DIR / "Models"
RES_DIR   = MODEL_DIR / "results"
IN_FILE   = RES_DIR / "comparison" / "unified_predictions.csv"
OUT_DIR   = RES_DIR / "crps"
OUT_DIR.mkdir(parents=True, exist_ok=True)

if not IN_FILE.exists():
    raise FileNotFoundError(
        f"File not found: {IN_FILE}\n"
        "Run comparison_final.py first."
    )

# =============================================================================
# 1. CONFIGURATION
# =============================================================================

CRISIS_START = "2019-01-01"
CRISIS_END   = "2021-06-01"
NORMAL_START = "2022-01-01"
NORMAL_END   = "2024-06-01"

LAGS    = [0]     # nowcast vintage only for CRPS
MODELS  = ["ARMA", "DFM", "Ridge", "MIDAS", "Lasso", "Random Forest", "XGBoost"]

# Minimum past residuals required to estimate sigma
MIN_RESIDUALS = 5

# =============================================================================
# 2. GAUSSIAN CRPS — CLOSED FORM
# =============================================================================

def gaussian_crps(y, mu, sigma):
    """
    Closed-form CRPS for Gaussian predictive distribution N(mu, sigma).

    Parameters
    ----------
    y     : float — observed value
    mu    : float — point forecast (mean of predictive distribution)
    sigma : float — standard deviation of predictive distribution

    Returns
    -------
    crps  : float — CRPS score (lower = better, 0 = perfect)

    Reference: Gneiting & Raftery (2007), JASA, eq. (21)
    """
    if np.isnan(y) or np.isnan(mu) or np.isnan(sigma) or sigma <= 0:
        return np.nan

    z = (y - mu) / sigma
    crps = sigma * (
        z * (2 * stats.norm.cdf(z) - 1)
        + 2 * stats.norm.pdf(z)
        - 1 / np.sqrt(np.pi)
    )
    return float(crps)


def compute_sigma_rolling(errors, t, min_obs=5):
    """
    Estimate predictive standard deviation at origin t using past residuals.
    Uses rolling standard deviation of past forecast errors.
    Pseudo-real-time: only uses errors from origins 0 to t-1.
    """
    past = errors[:t]
    past = past[~np.isnan(past)]
    if len(past) < min_obs:
        return np.nan
    return float(np.std(past, ddof=1))

# =============================================================================
# 3. LOAD DATA
# =============================================================================

print("Loading unified predictions...")
df = pd.read_csv(IN_FILE, parse_dates=["date"])
df = df.sort_values("date").reset_index(drop=True)

actuals  = df["actual"].values
dates_dt = pd.to_datetime(df["date"])

# Period masks
mask_full   = np.ones(len(actuals), dtype=bool)
mask_crisis = (dates_dt >= CRISIS_START) & (dates_dt <= CRISIS_END)
mask_normal = (dates_dt >= NORMAL_START) & (dates_dt <= NORMAL_END)

# Load predictions (lag=0 only)
preds = {}
for model in MODELS:
    col = f"{model}_lag0"
    if col in df.columns:
        preds[model] = df[col].values
    else:
        print(f"  WARNING: {col} not found — skipping {model}")

available = list(preds.keys())
print(f"Models available: {available}")

# =============================================================================
# 4. COMPUTE CRPS FOR EACH MODEL
# =============================================================================

print("\nComputing CRPS scores...")

crps_scores = {}   # model -> array of CRPS values

for model in available:
    yhat   = preds[model]
    errors = actuals - yhat
    scores = np.full(len(actuals), np.nan)

    for t in range(len(actuals)):
        y   = actuals[t]
        mu  = yhat[t]
        sig = compute_sigma_rolling(errors, t, MIN_RESIDUALS)
        scores[t] = gaussian_crps(y, mu, sig)

    crps_scores[model] = scores
    valid = ~np.isnan(scores)
    print(f"  [{model}] mean CRPS = {np.nanmean(scores):.6f} "
          f"| n_valid = {valid.sum()}")

# =============================================================================
# 5. PERFORMANCE TABLES
# =============================================================================

def mean_crps(scores, mask):
    s = scores[mask]
    return float(np.nanmean(s)) if (~np.isnan(s)).sum() > 0 else np.nan

rows = []
for model in available:
    for mask, period in [
        (mask_full,   "Full"),
        (mask_crisis, "Crisis"),
        (mask_normal, "Normal"),
    ]:
        rows.append({
            "Model":  model,
            "Period": period,
            "CRPS":   round(mean_crps(crps_scores[model], mask), 6),
            "n":      int((~np.isnan(crps_scores[model][mask])).sum()),
        })

crps_df = pd.DataFrame(rows)

# Relative CRPS vs ARMA
if "ARMA" in available:
    arma_crps = {
        row["Period"]: row["CRPS"]
        for _, row in crps_df[crps_df["Model"] == "ARMA"].iterrows()
    }
    crps_df["rel_CRPS"] = crps_df.apply(
        lambda r: round(r["CRPS"] / arma_crps[r["Period"]], 4)
        if arma_crps.get(r["Period"]) else np.nan,
        axis=1
    )
    # CRPS Skill Score: 1 - CRPS_model / CRPS_ARMA
    # Positive = better than ARMA
    crps_df["CRPSS"] = (1 - crps_df["rel_CRPS"]).round(4)

# Pivot table: models x periods
crps_pivot = crps_df.pivot(
    index="Model", columns="Period", values="rel_CRPS"
)[["Crisis", "Normal", "Full"]].round(4)
crps_pivot = crps_pivot.sort_values("Full")

# CRPSS pivot (skill score)
crpss_pivot = crps_df.pivot(
    index="Model", columns="Period", values="CRPSS"
)[["Crisis", "Normal", "Full"]].round(4)
crpss_pivot = crpss_pivot.sort_values("Full", ascending=False)

print("\n" + "="*60)
print("RELATIVE CRPS (ARMA = 1) — lower = better density forecast")
print("="*60)
print(crps_pivot.to_string())

print("\n" + "="*60)
print("CRPS SKILL SCORE (1 - rel_CRPS) — higher = better vs ARMA")
print("Positive = model has better density forecast than ARMA")
print("="*60)
print(crpss_pivot.to_string())

# =============================================================================
# 6. COMPARE RMSE vs CRPS — key diagnostic
# =============================================================================
# A model can win on RMSE but lose on CRPS if its uncertainty is poor.
# This comparison is the key value-add of the density evaluation.

print("\n" + "="*60)
print("RMSE vs CRPS COMPARISON — Full Period, lag=0")
print("Lower relative score = better performance")
print("="*60)

# Load RMSE from comparison results
rmse_path = RES_DIR / "comparison" / "rel_rmse_full.csv"
if rmse_path.exists():
    rmse_full = pd.read_csv(rmse_path, index_col=0)
    rmse_nowcast = rmse_full["lag=0 (nowcast)"].rename("rel_RMSE")

    crps_full = crps_df[crps_df["Period"] == "Full"].set_index(
        "Model"
    )["rel_CRPS"].rename("rel_CRPS")

    comparison = pd.concat([rmse_nowcast, crps_full], axis=1).dropna()
    comparison["RMSE_rank"] = comparison["rel_RMSE"].rank()
    comparison["CRPS_rank"] = comparison["rel_CRPS"].rank()
    comparison["Rank_diff"] = (
        comparison["CRPS_rank"] - comparison["RMSE_rank"]
    ).astype(int)
    comparison = comparison.sort_values("rel_RMSE")
    print(comparison.round(4).to_string())
    print()
    print("Rank_diff > 0: model worse at density than at point forecasts")
    print("Rank_diff < 0: model better at density than at point forecasts")
    print("Rank_diff = 0: consistent performance across both metrics")

# =============================================================================
# 7. SAVE RESULTS
# =============================================================================

crps_df.to_csv(OUT_DIR / "crps_scores.csv", index=False)
crps_pivot.to_csv(OUT_DIR / "crps_relative_table.csv")
crpss_pivot.to_csv(OUT_DIR / "crpss_skill_table.csv")
if rmse_path.exists():
    comparison.to_csv(OUT_DIR / "rmse_vs_crps_comparison.csv")

print(f"\nCSVs saved to: {OUT_DIR}")

# =============================================================================
# 8. PLOTS
# =============================================================================

# ---- PLOT 1: CRPS bar chart — crisis vs normal ------------------------------

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle(
    "Relative CRPS vs ARMA Benchmark (lag=0, nowcast vintage)\n"
    "Evaluates FULL predictive distribution — not just point forecast\n"
    "Values below 1.0 indicate better density forecast than ARMA",
    fontsize=12, fontweight="bold"
)

MODEL_COLORS = {
    "ARMA":          "#6c757d",
    "DFM":           "#00BCD4",
    "Ridge":         "#4CAF50",
    "MIDAS":         "#2196F3",
    "Lasso":         "#8BC34A",
    "Random Forest": "#FF9800",
    "XGBoost":       "#FF5722",
}

for ax, period, title in zip(
    axes,
    ["Crisis", "Normal"],
    ["Crisis Period (2019Q1-2021Q2)", "Non-Crisis Period (2022Q1-2024Q2)"]
):
    sub = crps_df[crps_df["Period"] == period].sort_values("rel_CRPS")
    colors = [MODEL_COLORS.get(m, "#888888") for m in sub["Model"]]

    bars = ax.bar(sub["Model"], sub["rel_CRPS"],
                  color=colors, edgecolor="gray", linewidth=0.5)
    ax.axhline(1.0, color="#e63946", lw=1.5, ls="--", label="ARMA = 1.0")

    for bar, val in zip(bars, sub["rel_CRPS"]):
        ax.text(bar.get_x() + bar.get_width()/2,
                bar.get_height() + 0.01,
                f"{val:.3f}",
                ha="center", va="bottom", fontsize=8)

    ax.set_title(title, fontsize=11, fontweight="bold")
    ax.set_ylabel("Relative CRPS (ARMA = 1)", fontsize=10)
    ax.set_ylim(0, max(sub["rel_CRPS"].max() * 1.2, 1.3))
    ax.tick_params(axis="x", rotation=30)
    ax.grid(True, axis="y", alpha=0.3)
    ax.legend(fontsize=9)

plt.tight_layout()
fig.savefig(OUT_DIR / "crps_plot1_bar_crisis_normal.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("Saved: crps_plot1_bar_crisis_normal.png")

# ---- PLOT 2: RMSE vs CRPS scatter ------------------------------------------

if rmse_path.exists():
    fig, ax = plt.subplots(figsize=(8, 7))
    fig.suptitle(
        "Point Forecast (RMSE) vs Density Forecast (CRPS) — Full Period\n"
        "Models on the diagonal: consistent performance across both metrics\n"
        "Below diagonal: better at density | Above: better at point forecast",
        fontsize=11, fontweight="bold"
    )

    for model, row in comparison.iterrows():
        color = MODEL_COLORS.get(model, "#888888")
        ax.scatter(row["rel_RMSE"], row["rel_CRPS"],
                   color=color, s=120, zorder=5,
                   edgecolors="white", linewidths=1.5)
        ax.annotate(model,
                    (row["rel_RMSE"], row["rel_CRPS"]),
                    textcoords="offset points",
                    xytext=(8, 4), fontsize=9)

    # Diagonal: RMSE = CRPS (perfect consistency)
    lims = [min(comparison["rel_RMSE"].min(), comparison["rel_CRPS"].min()) - 0.05,
            max(comparison["rel_RMSE"].max(), comparison["rel_CRPS"].max()) + 0.05]
    ax.plot(lims, lims, color="#aaaaaa", ls="--", lw=1.2,
            label="RMSE = CRPS (diagonal)")
    ax.axhline(1.0, color="#e63946", lw=0.8, ls=":", alpha=0.6)
    ax.axvline(1.0, color="#e63946", lw=0.8, ls=":", alpha=0.6)
    ax.set_xlabel("Relative RMSE (point forecast)", fontsize=11)
    ax.set_ylabel("Relative CRPS (density forecast)", fontsize=11)
    ax.legend(fontsize=9)
    ax.grid(True, alpha=0.25)
    plt.tight_layout()
    fig.savefig(OUT_DIR / "crps_plot2_rmse_vs_crps.png",
                dpi=150, bbox_inches="tight")
    plt.show()
    print("Saved: crps_plot2_rmse_vs_crps.png")

# ---- PLOT 3: CRPS over time for top 3 models --------------------------------

fig, ax = plt.subplots(figsize=(14, 5))
fig.suptitle(
    "CRPS Over Time — Rolling Density Forecast Quality\n"
    "Lower = better density forecast | Spike in 2020Q2 expected",
    fontsize=12, fontweight="bold"
)

top_models = ["DFM", "Ridge", "MIDAS", "ARMA"]
for model in top_models:
    if model not in crps_scores:
        continue
    scores = crps_scores[model]
    valid  = ~np.isnan(scores)
    ax.plot(dates_dt[valid], scores[valid],
            color=MODEL_COLORS.get(model, "#888888"),
            lw=2, marker="o", markersize=4,
            label=model)

ax.axvspan(pd.to_datetime(CRISIS_START), pd.to_datetime(CRISIS_END),
           alpha=0.10, color="#e63946", label="Crisis window")
ax.axvspan(pd.to_datetime(NORMAL_START), pd.to_datetime(NORMAL_END),
           alpha=0.08, color="#2a9d8f", label="Normal window")
ax.set_xlabel("Date")
ax.set_ylabel("CRPS")
ax.legend(fontsize=9, ncol=3)
ax.grid(True, alpha=0.25)
import matplotlib.dates as mdates
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
plt.tight_layout()
fig.savefig(OUT_DIR / "crps_plot3_over_time.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("Saved: crps_plot3_over_time.png")

# =============================================================================
# DONE
# =============================================================================
print("\n" + "="*60)
print("CRPS EVALUATION COMPLETE")
print("="*60)
print(f"Output: {OUT_DIR}")
print()
print("CSVs:")
print("  crps_scores.csv            — CRPS by model/period")
print("  crps_relative_table.csv    — relative CRPS vs ARMA")
print("  crpss_skill_table.csv      — CRPS skill score")
print("  rmse_vs_crps_comparison.csv — point vs density ranking")
print()
print("Plots:")
print("  crps_plot1_bar_crisis_normal.png — bar chart by model")
print("  crps_plot2_rmse_vs_crps.png      — scatter RMSE vs CRPS")
print("  crps_plot3_over_time.png         — CRPS evolution over time")
print()
print("NEXT: Run shap_final.py for XGBoost interpretability.")
```

```python
# =============================================================================
# interpretability_final.py
# Purpose : Economic interpretability analysis for Ridge, Lasso, and XGBoost.
#           Answers: which variables drive nowcasting performance in El Salvador?
#
# What this script produces:
#   RIDGE:
#     1. Top 20 coefficients at nowcast vintage (last training window)
#     2. Coefficient stability over time (top 10 variables)
#
#   LASSO:
#     3. Variable selection frequency — how often each variable is selected
#        across all 22 rolling windows (sparsity analysis)
#     4. Top selected variables by average absolute coefficient
#     5. Sparsity over time — how many variables Lasso keeps per window
#
#   XGBOOST:
#     6. SHAP summary plot (bee-swarm) — global feature importance
#     7. SHAP bar plot — mean absolute SHAP by variable
#     8. SHAP dependence plot for top 3 variables
#
#   COMBINED:
#     9. Venn-style comparison: which variables appear in ALL three models?
#
# WHY INTERPRETABILITY MATTERS FOR THE PAPER:
#   "Which macroeconomic indicators lead the GDP cycle in El Salvador?"
#   This is the policy-relevant question. The interpretability section
#   connects the statistical results to economic substance.
#
#   Ridge  → coefficient magnitude (all variables retained, shrunk toward 0)
#   Lasso  → variable selection (most variables set exactly to 0)
#   XGBoost → SHAP values (nonlinear, interaction-aware contributions)
#
#   Agreement across models = robust leading indicator
#   Disagreement = model-specific signal, interpret with caution
#
# Author  : Gabriela Aquino
# Input   : ~/Desktop/ESA-gdp-nowcasting/Data/data_tf.csv
#           ~/Desktop/ESA-gdp-nowcasting/Data/meta_data_v2.csv
# Output  : ~/Desktop/ESA-gdp-nowcasting/Models/results/interpretability/
# =============================================================================

import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import warnings
from pathlib import Path

warnings.filterwarnings("ignore")
np.random.seed(42)

# Try importing SHAP — warn if not available
try:
    import shap
    SHAP_AVAILABLE = True
    print("SHAP available — XGBoost interpretability will be computed")
except ImportError:
    SHAP_AVAILABLE = False
    print("WARNING: SHAP not installed. Run: pip install shap")
    print("Ridge and Lasso analysis will proceed without SHAP.")

from sklearn.linear_model import Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

try:
    import xgboost as xgb
    XGB_AVAILABLE = True
except ImportError:
    XGB_AVAILABLE = False
    print("WARNING: XGBoost not installed.")

# =============================================================================
# 0. PATHS
# =============================================================================

BASE_DIR  = Path.home() / "Desktop" / "ESA-gdp-nowcasting"
DATA_DIR  = BASE_DIR / "Data"
MODEL_DIR = BASE_DIR / "Models"
OUT_DIR   = MODEL_DIR / "results" / "interpretability"
OUT_DIR.mkdir(parents=True, exist_ok=True)

sys.path.insert(0, str(MODEL_DIR))
from nowcast_utils import gen_lagged_data, flatten_data, mean_fill_dataset

# =============================================================================
# 1. CONFIGURATION — must match the main model scripts exactly
# =============================================================================

TARGET_VARIABLE    = "GDP"
TRAIN_START        = "2005-01-01"
TEST_START         = "2019-03-01"
TEST_END           = "2024-06-01"
N_LAGS_FEATURES    = 4
N_HOLDOUT_QUARTERS = 8
NOWCAST_LAG        = 0    # lag=0 is the operationally relevant vintage

# Ridge and Lasso alpha (use the same tuning as the main scripts)
# We re-tune using the same fixed holdout protocol for consistency
RIDGE_ALPHA_GRID   = np.logspace(-3, 3, 50)
LASSO_ALPHA_GRID   = np.logspace(-6, 0, 50)

# XGBoost best params — use fixed params from last rolling window
# (avoids re-running 128-combo grid for interpretability)
XGB_PARAMS = {
    "max_depth":        2,
    "n_estimators":     100,
    "learning_rate":    0.05,
    "subsample":        0.8,
    "colsample_bytree": 0.8,
    "reg_lambda":       1.0,
    "min_child_weight": 3,
    "random_state":     42,
}

# =============================================================================
# 2. LOAD DATA
# =============================================================================

print("Loading data...")
metadata = pd.read_csv(DATA_DIR / "meta_data_v2.csv")
data     = pd.read_csv(DATA_DIR / "data_tf.csv",
                        parse_dates=["date"])
data     = data.sort_values("date").reset_index(drop=True)

# Build feature matrix for the LAST training window
# (most informative — trained on all available data before test period)
# Use this for all interpretability analyses

# Last test date = 2024-06-01, so last training window ends 2024-03-01
LAST_TRAIN_CUTOFF = "2024-03-01"
LAST_NOWCAST_DATE = "2024-06-01"

train_raw    = data.loc[
    (data["date"] >= TRAIN_START) &
    (data["date"] <= LAST_TRAIN_CUTOFF)
].copy()

vintage_data = gen_lagged_data(
    metadata, data, LAST_NOWCAST_DATE, NOWCAST_LAG
)

# Fill missing values with training means
train_filled   = mean_fill_dataset(training=train_raw, test=train_raw)
vintage_filled = mean_fill_dataset(training=train_raw, test=vintage_data)

# Flatten to quarterly feature matrix
train_flat = flatten_data(train_filled, TARGET_VARIABLE, N_LAGS_FEATURES)
train_flat = (
    train_flat
    .loc[train_flat["date"].dt.month.isin([3, 6, 9, 12])]
    .dropna(axis=0, how="any")
    .reset_index(drop=True)
)

X_full = train_flat.drop(columns=["date", TARGET_VARIABLE])
y_full = train_flat[TARGET_VARIABLE]
feature_names = list(X_full.columns)

print(f"  Training quarters: {len(y_full)}")
print(f"  Features         : {len(feature_names)}")

# Holdout split (same protocol as main scripts)
X_inner = X_full.iloc[:-N_HOLDOUT_QUARTERS]
y_inner = y_full.iloc[:-N_HOLDOUT_QUARTERS]
X_hold  = X_full.iloc[-N_HOLDOUT_QUARTERS:]
y_hold  = y_full.iloc[-N_HOLDOUT_QUARTERS:]

# Scale for Ridge and Lasso (XGBoost doesn't need scaling)
scaler  = StandardScaler()
X_sc    = scaler.fit_transform(X_full)
X_sc_in = scaler.transform(X_inner)
X_sc_ho = scaler.transform(X_hold)

# =============================================================================
# 3. CLEAN FEATURE NAMES FOR PLOTS
# =============================================================================

def clean_name(fname):
    """
    Convert feature names like 'IVAE_TOT_lag1' to 'IVAE_TOT (lag 1)'.
    Makes plots more readable.
    """
    parts = fname.rsplit("_lag", 1)
    if len(parts) == 2:
        return f"{parts[0]} (lag {parts[1]})"
    return fname

clean_names = [clean_name(f) for f in feature_names]

# =============================================================================
# 4. RIDGE — COEFFICIENT ANALYSIS
# =============================================================================

print("\n" + "="*60)
print("RIDGE — Coefficient Analysis")
print("="*60)

# Tune alpha using holdout
best_alpha_ridge, best_rmse_ridge = None, np.inf
for alpha in RIDGE_ALPHA_GRID:
    m = Ridge(alpha=alpha, fit_intercept=True)
    m.fit(X_sc_in, y_inner)
    rmse = np.sqrt(mean_squared_error(y_hold, m.predict(X_sc_ho)))
    if rmse < best_rmse_ridge:
        best_rmse_ridge  = rmse
        best_alpha_ridge = alpha

print(f"  Best alpha: {best_alpha_ridge:.4f}")

# Fit on full training data
ridge_final = Ridge(alpha=best_alpha_ridge, fit_intercept=True)
ridge_final.fit(X_sc, y_full)
ridge_coefs = pd.DataFrame({
    "feature":     feature_names,
    "clean_name":  clean_names,
    "coefficient": ridge_final.coef_,
    "abs_coef":    np.abs(ridge_final.coef_),
}).sort_values("abs_coef", ascending=False)

print(f"\n  Top 20 Ridge coefficients (last training window):")
print(ridge_coefs[["clean_name", "coefficient"]].head(20).to_string(index=False))

# ---- Plot: Top 20 Ridge coefficients ----------------------------------------
top_ridge = ridge_coefs.head(20)
colors_ridge = ["#2a9d8f" if c > 0 else "#e63946"
                for c in top_ridge["coefficient"]]

fig, ax = plt.subplots(figsize=(10, 8))
ax.barh(top_ridge["clean_name"][::-1],
        top_ridge["coefficient"][::-1],
        color=colors_ridge[::-1],
        edgecolor="gray", linewidth=0.4)
ax.axvline(0, color="#333333", lw=1.2, ls="--")
ax.set_xlabel("Standardised Coefficient", fontsize=11)
ax.set_title(
    "Ridge Regression — Top 20 Coefficients\n"
    "Last training window (ending 2024Q1) | Standardised predictors\n"
    "Green = positive effect on GDP | Red = negative effect",
    fontsize=11, fontweight="bold"
)
ax.grid(True, axis="x", alpha=0.3)
plt.tight_layout()
fig.savefig(OUT_DIR / "ridge_coefficients.png", dpi=150, bbox_inches="tight")
plt.show()
print("\n  Saved: ridge_coefficients.png")

# Save
ridge_coefs.to_csv(OUT_DIR / "ridge_coefficients.csv", index=False)

# =============================================================================
# 5. LASSO — VARIABLE SELECTION ANALYSIS
# =============================================================================

print("\n" + "="*60)
print("LASSO — Variable Selection Analysis")
print("="*60)

# Tune alpha using holdout
best_alpha_lasso, best_rmse_lasso = None, np.inf
for alpha in LASSO_ALPHA_GRID:
    m = Lasso(alpha=alpha, fit_intercept=True,
              max_iter=10000, tol=1e-4)
    m.fit(X_sc_in, y_inner)
    rmse = np.sqrt(mean_squared_error(y_hold, m.predict(X_sc_ho)))
    if rmse < best_rmse_lasso:
        best_rmse_lasso  = rmse
        best_alpha_lasso = alpha

print(f"  Best alpha: {best_alpha_lasso:.6f}")

# Fit on full training data
lasso_final = Lasso(alpha=best_alpha_lasso, fit_intercept=True,
                    max_iter=10000, tol=1e-4)
lasso_final.fit(X_sc, y_full)
lasso_coefs = pd.DataFrame({
    "feature":     feature_names,
    "clean_name":  clean_names,
    "coefficient": lasso_final.coef_,
    "abs_coef":    np.abs(lasso_final.coef_),
    "selected":    lasso_final.coef_ != 0,
})

n_selected = lasso_coefs["selected"].sum()
sparsity   = 1 - n_selected / len(feature_names)
print(f"\n  Variables selected: {n_selected} / {len(feature_names)}")
print(f"  Sparsity          : {sparsity:.1%}")

selected_vars = (
    lasso_coefs[lasso_coefs["selected"]]
    .sort_values("abs_coef", ascending=False)
)
print(f"\n  Selected variables (non-zero coefficient):")
print(selected_vars[["clean_name", "coefficient"]].to_string(index=False))

# ---- Plot: Lasso selected variables -----------------------------------------
if len(selected_vars) > 0:
    colors_lasso = ["#2a9d8f" if c > 0 else "#e63946"
                    for c in selected_vars["coefficient"]]
    top_n = min(25, len(selected_vars))
    sv_top = selected_vars.head(top_n)

    fig, ax = plt.subplots(figsize=(10, max(6, top_n * 0.4)))
    ax.barh(sv_top["clean_name"][::-1],
            sv_top["coefficient"][::-1],
            color=(colors_lasso[::-1])[-top_n:],
            edgecolor="gray", linewidth=0.4)
    ax.axvline(0, color="#333333", lw=1.2, ls="--")
    ax.set_xlabel("Standardised Coefficient", fontsize=11)
    ax.set_title(
        f"Lasso — Selected Variables (non-zero coefficients)\n"
        f"{n_selected} of {len(feature_names)} features selected "
        f"({sparsity:.0%} sparsity) | Last training window\n"
        "Green = positive | Red = negative",
        fontsize=11, fontweight="bold"
    )
    ax.grid(True, axis="x", alpha=0.3)
    plt.tight_layout()
    fig.savefig(OUT_DIR / "lasso_selected_variables.png",
                dpi=150, bbox_inches="tight")
    plt.show()
    print("\n  Saved: lasso_selected_variables.png")

# ---- Rolling selection frequency: how often each variable is selected -------
print("\n  Computing rolling variable selection frequency...")

test_dates = pd.date_range(TEST_START, TEST_END, freq="3MS")
selection_counts = pd.Series(0, index=feature_names, dtype=int)
coef_sums        = pd.Series(0.0, index=feature_names)
n_windows        = 0

for nowcast_date in test_dates:
    train_cutoff = (nowcast_date - pd.DateOffset(months=3)).strftime("%Y-%m-%d")
    tr_raw = data.loc[
        (data["date"] >= TRAIN_START) &
        (data["date"] <= train_cutoff)
    ].copy()
    if len(tr_raw) < 30:
        continue
    tr_filled = mean_fill_dataset(training=tr_raw, test=tr_raw)
    tr_flat   = flatten_data(tr_filled, TARGET_VARIABLE, N_LAGS_FEATURES)
    tr_flat   = (
        tr_flat
        .loc[tr_flat["date"].dt.month.isin([3, 6, 9, 12])]
        .dropna(axis=0, how="any")
        .reset_index(drop=True)
    )
    if len(tr_flat) < N_HOLDOUT_QUARTERS + 5:
        continue

    Xw = tr_flat.drop(columns=["date", TARGET_VARIABLE])
    yw = tr_flat[TARGET_VARIABLE]
    sc = StandardScaler()
    Xs = sc.fit_transform(Xw)
    Xi = Xs[:-N_HOLDOUT_QUARTERS]
    yi = yw.iloc[:-N_HOLDOUT_QUARTERS]
    Xh = Xs[-N_HOLDOUT_QUARTERS:]
    yh = yw.iloc[-N_HOLDOUT_QUARTERS:]

    best_a, best_r = None, np.inf
    for alpha in LASSO_ALPHA_GRID:
        m = Lasso(alpha=alpha, max_iter=10000, tol=1e-4)
        m.fit(Xi, yi)
        r = np.sqrt(mean_squared_error(yh, m.predict(Xh)))
        if r < best_r:
            best_r, best_a = r, alpha

    m_final = Lasso(alpha=best_a, max_iter=10000, tol=1e-4)
    m_final.fit(Xs, yw)

    for f in feature_names:
        if f in Xw.columns:
            idx = list(Xw.columns).index(f)
            if m_final.coef_[idx] != 0:
                selection_counts[f] += 1
            coef_sums[f] += abs(m_final.coef_[idx])
    n_windows += 1
    print(f"    [{nowcast_date.strftime('%Y-%m-%d')}] "
          f"selected={int((m_final.coef_ != 0).sum())}")

# Selection frequency as percentage
sel_freq = (selection_counts / n_windows * 100).sort_values(ascending=False)
sel_freq = sel_freq[sel_freq > 0]

freq_df = pd.DataFrame({
    "feature":    sel_freq.index,
    "clean_name": [clean_name(f) for f in sel_freq.index],
    "frequency":  sel_freq.values,
    "avg_abs_coef": (coef_sums[sel_freq.index] / n_windows).values,
}).sort_values("frequency", ascending=False)

print(f"\n  Variables selected in >50% of windows:")
print(freq_df[freq_df["frequency"] > 50][
    ["clean_name", "frequency", "avg_abs_coef"]
].to_string(index=False))

# ---- Plot: Selection frequency ----------------------------------------------
top_freq = freq_df.head(25)
fig, ax  = plt.subplots(figsize=(10, 8))
bars = ax.barh(top_freq["clean_name"][::-1],
               top_freq["frequency"][::-1],
               color="#457b9d", edgecolor="gray", linewidth=0.4)
ax.axvline(50, color="#e63946", lw=1.5, ls="--",
           label="Selected in 50% of windows")
ax.set_xlabel("Selection frequency (%)", fontsize=11)
ax.set_title(
    "Lasso — Variable Selection Frequency Across Rolling Windows\n"
    f"22 rolling origins | {n_windows} windows estimated\n"
    "High frequency = consistently informative for GDP nowcasting",
    fontsize=11, fontweight="bold"
)
ax.legend(fontsize=9)
ax.grid(True, axis="x", alpha=0.3)
plt.tight_layout()
fig.savefig(OUT_DIR / "lasso_selection_frequency.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("\n  Saved: lasso_selection_frequency.png")
freq_df.to_csv(OUT_DIR / "lasso_selection_frequency.csv", index=False)

# =============================================================================
# 6. XGBOOST — SHAP ANALYSIS
# =============================================================================

if SHAP_AVAILABLE and XGB_AVAILABLE:
    print("\n" + "="*60)
    print("XGBOOST — SHAP Analysis")
    print("="*60)

    # Fit XGBoost on last training window (no scaling needed)
    xgb_model = xgb.XGBRegressor(**XGB_PARAMS, verbosity=0)
    xgb_model.fit(X_full, y_full)

    # Compute SHAP values for all training observations
    explainer  = shap.TreeExplainer(xgb_model)
    shap_vals  = explainer.shap_values(X_full)
    shap_df    = pd.DataFrame(shap_vals, columns=feature_names)

    # Mean absolute SHAP per feature
    mean_abs_shap = shap_df.abs().mean().sort_values(ascending=False)
    mean_abs_shap.index = [clean_name(f) for f in mean_abs_shap.index]

    top_shap = mean_abs_shap.head(20)
    print(f"\n  Top 20 features by mean |SHAP|:")
    print(top_shap.round(6).to_string())

    # ---- Plot 1: SHAP bar chart (mean absolute) -----------------------------
    fig, ax = plt.subplots(figsize=(10, 8))
    ax.barh(top_shap.index[::-1], top_shap.values[::-1],
            color="#e07b00", edgecolor="gray", linewidth=0.4)
    ax.set_xlabel("Mean |SHAP value|", fontsize=11)
    ax.set_title(
        "XGBoost — Feature Importance via SHAP\n"
        "Mean absolute SHAP value | Last training window (ending 2024Q1)\n"
        "Higher = more influential for GDP predictions",
        fontsize=11, fontweight="bold"
    )
    ax.grid(True, axis="x", alpha=0.3)
    plt.tight_layout()
    fig.savefig(OUT_DIR / "xgboost_shap_bar.png", dpi=150, bbox_inches="tight")
    plt.show()
    print("\n  Saved: xgboost_shap_bar.png")

    # ---- Plot 2: SHAP summary (bee-swarm) -----------------------------------
    # Show top 20 features
    top_feature_names = mean_abs_shap.head(20).index.tolist()
    # Get original feature indices for these names
    top_feature_orig = [f for f in feature_names
                        if clean_name(f) in top_feature_names][:20]
    top_feature_idx  = [feature_names.index(f) for f in top_feature_orig]

    shap_top = shap_vals[:, top_feature_idx]
    X_top    = X_full.iloc[:, top_feature_idx].copy()
    X_top.columns = [clean_name(f) for f in top_feature_orig]

    fig, ax = plt.subplots(figsize=(10, 9))
    shap.summary_plot(
        shap_top, X_top,
        feature_names=[clean_name(f) for f in top_feature_orig],
        plot_type="dot",
        show=False,
        max_display=20,
        alpha=0.7
    )
    plt.title(
        "XGBoost — SHAP Summary Plot (Bee-swarm)\n"
        "Each dot = one training quarter | "
        "Color = feature value (blue=low, red=high)\n"
        "Position on x-axis = SHAP value (impact on GDP prediction)",
        fontsize=11, fontweight="bold", pad=15
    )
    plt.tight_layout()
    fig.savefig(OUT_DIR / "xgboost_shap_beeswarm.png",
                dpi=150, bbox_inches="tight")
    plt.show()
    print("  Saved: xgboost_shap_beeswarm.png")

    # Save SHAP values
    shap_export = pd.DataFrame(shap_vals, columns=feature_names)
    shap_export["actual"] = y_full.values
    mean_abs_shap_orig = pd.DataFrame({
        "feature":    [f for f in feature_names],
        "clean_name": clean_names,
        "mean_abs_shap": shap_df.abs().mean().values,
    }).sort_values("mean_abs_shap", ascending=False)
    mean_abs_shap_orig.to_csv(
        OUT_DIR / "xgboost_shap_importance.csv", index=False
    )
    print("  Saved: xgboost_shap_importance.csv")

else:
    print("\nSHAP/XGBoost not available — skipping SHAP section.")
    print("Install with: pip install shap xgboost")

# =============================================================================
# 7. COMBINED: Which variables appear across ALL three models?
# =============================================================================

print("\n" + "="*60)
print("CROSS-MODEL VARIABLE IMPORTANCE COMPARISON")
print("="*60)

# Ridge: top 20 by absolute coefficient
ridge_top = set(ridge_coefs.head(20)["feature"].tolist())

# Lasso: variables selected in >50% of windows
lasso_top = set(freq_df[freq_df["frequency"] > 50]["feature"].tolist())

# XGBoost SHAP: top 20 by mean |SHAP|
if SHAP_AVAILABLE and XGB_AVAILABLE:
    shap_top_feats = set(
        mean_abs_shap_orig.head(20)["feature"].tolist()
    )
    intersection = ridge_top & lasso_top & shap_top_feats
    ridge_lasso  = ridge_top & lasso_top - shap_top_feats
    in_all_label = "Ridge + Lasso + XGBoost"
else:
    intersection = ridge_top & lasso_top
    ridge_lasso  = set()
    in_all_label = "Ridge + Lasso"

print(f"\n  Variables in {in_all_label} (top indicators):")
for f in sorted(intersection):
    print(f"    {clean_name(f)}")

print(f"\n  Variables in Ridge + Lasso only:")
for f in sorted(ridge_lasso):
    print(f"    {clean_name(f)}")

# Save cross-model comparison
cross_model = pd.DataFrame({
    "feature":    feature_names,
    "clean_name": clean_names,
    "in_ridge_top20":   [f in ridge_top for f in feature_names],
    "in_lasso_gt50pct": [f in lasso_top for f in feature_names],
    "in_xgb_shap_top20": (
        [f in shap_top_feats for f in feature_names]
        if SHAP_AVAILABLE and XGB_AVAILABLE
        else [False] * len(feature_names)
    ),
})
cross_model["n_models"] = (
    cross_model["in_ridge_top20"].astype(int) +
    cross_model["in_lasso_gt50pct"].astype(int) +
    cross_model["in_xgb_shap_top20"].astype(int)
)
cross_model = cross_model.sort_values("n_models", ascending=False)
cross_model.to_csv(OUT_DIR / "cross_model_importance.csv", index=False)

# ---- Plot: Cross-model heatmap ----------------------------------------------
top_cross = cross_model[cross_model["n_models"] > 0].head(30)
heat_data  = top_cross[[
    "in_ridge_top20", "in_lasso_gt50pct", "in_xgb_shap_top20"
]].astype(int)
heat_data.index = top_cross["clean_name"].tolist()
heat_data.columns = ["Ridge\n(top 20)", "Lasso\n(>50% freq)",
                     "XGBoost\n(SHAP top 20)"]

import seaborn as sns
fig, ax = plt.subplots(figsize=(8, max(6, len(top_cross) * 0.35)))
sns.heatmap(heat_data,
            annot=True, fmt="d",
            cmap=["#f8f9fa", "#2a9d8f"],
            linewidths=0.5, ax=ax,
            cbar=False,
            vmin=0, vmax=1)
ax.set_title(
    "Cross-Model Variable Importance\n"
    "1 = variable important for that model | 0 = not important\n"
    "Variables appearing in all 3 models = most robust leading indicators",
    fontsize=11, fontweight="bold"
)
ax.set_xlabel("")
ax.set_ylabel("")
plt.tight_layout()
fig.savefig(OUT_DIR / "cross_model_importance.png",
            dpi=150, bbox_inches="tight")
plt.show()
print("\n  Saved: cross_model_importance.png")

# =============================================================================
# DONE
# =============================================================================
print("\n" + "="*60)
print("INTERPRETABILITY ANALYSIS COMPLETE")
print("="*60)
print(f"Output: {OUT_DIR}")
print()
print("Ridge:")
print("  ridge_coefficients.png        — top 20 coefficients")
print("  ridge_coefficients.csv        — full coefficient table")
print()
print("Lasso:")
print("  lasso_selected_variables.png  — selected vars last window")
print("  lasso_selection_frequency.png — rolling selection frequency")
print("  lasso_selection_frequency.csv — full frequency table")
print()
print("XGBoost (if SHAP installed):")
print("  xgboost_shap_bar.png          — mean |SHAP| bar chart")
print("  xgboost_shap_beeswarm.png     — SHAP summary (bee-swarm)")
print("  xgboost_shap_importance.csv   — SHAP values table")
print()
print("Combined:")
print("  cross_model_importance.png    — which vars appear in all models")
print("  cross_model_importance.csv    — full cross-model table")
```
