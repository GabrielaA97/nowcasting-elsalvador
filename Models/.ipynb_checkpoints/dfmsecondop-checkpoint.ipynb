```python
# =============================================================================
# dfm_postprocess.py
# Post-process an existing dfm_predictions.csv to align it with the
# canonical test window and period partition defined in nowcast_utils.py.
#
# This script DOES NOT re-estimate the DFM. It only:
#   1. Filters predictions to the canonical test window (2020Q1–2024Q2).
#   2. Reclassifies period_type using classify_crisis_periods from utils
#      (no more "other" — raises if any date falls outside crisis/normal).
#   3. Regenerates performance tables (full + stratified) using rmse/mae
#      from utils.
#
# Inputs:
#   dfm_predictions.csv  (existing — the 22-window file from the 4-day run)
#   nowcast_utils.py     (canonical constants and metrics)
#
# Outputs:
#   dfm_predictions_v2.csv
#   dfm_performance_full_v2.csv
#   dfm_performance_stratified_v2.csv
#
# Author : Gabriela Aquino
# =============================================================================

import sys
from pathlib import Path
import pandas as pd
import numpy as np

# --- Paths ------------------------------------------------------------------
BASE_DIR  = Path.home() / "Desktop" / "ESA-gdp-nowcasting"
MODEL_DIR = BASE_DIR / "Models"
IN_DIR    = MODEL_DIR / "results" / "dfm"
OUT_DIR   = IN_DIR            # overwrite-compatible: writes *_v2.csv alongside

sys.path.insert(0, str(MODEL_DIR))

from nowcast_utils import (
    LAGS, TEST_START, TEST_END,
    classify_crisis_periods,
    rmse, mae,
)

PATH_IN = IN_DIR / "dfm_predictions.csv"
if not PATH_IN.exists():
    raise FileNotFoundError(f"\n  File not found: {PATH_IN}")

# --- Load -------------------------------------------------------------------
print("Loading existing DFM predictions...")
df = pd.read_csv(PATH_IN, parse_dates=["date"])
print(f"  Rows in  : {len(df)}")
print(f"  Date range in  : {df['date'].min().date()} to {df['date'].max().date()}")

# --- Filter to canonical test window ---------------------------------------
test_start_dt = pd.to_datetime(TEST_START)
test_end_dt   = pd.to_datetime(TEST_END)

before = len(df)
df = df.loc[
    (df["date"] >= test_start_dt) & (df["date"] <= test_end_dt)
].reset_index(drop=True)
after = len(df)
print(f"\nFiltered to canonical test window: {TEST_START} → {TEST_END}")
print(f"  Windows kept    : {after}")
print(f"  Windows dropped : {before - after}  (pre-2020Q1 not part of test)")

expected = 18
if after != expected:
    print(f"  WARNING: expected {expected} windows, got {after}")

# --- Reclassify period_type using canonical utils --------------------------
print("\nReclassifying period_type with canonical crisis/normal windows...")
old_labels = df["period_type"].tolist()
new_labels = classify_crisis_periods(df["date"].tolist()).tolist()
df["period_type"] = new_labels

# Report any relabels
changed = [
    (d.date(), old, new)
    for d, old, new in zip(df["date"], old_labels, new_labels)
    if old != new
]
if changed:
    print(f"  Relabelled {len(changed)} rows:")
    for d, old, new in changed:
        print(f"    {d}  {old!r:10s} -> {new!r}")
else:
    print("  No relabels needed.")

print(f"\nFinal period counts:")
print(df["period_type"].value_counts().to_string())

# --- Recompute performance tables ------------------------------------------
actuals = df["actual"].tolist()
lag_cols = {-2: "lag_-2", -1: "lag_-1", 0: "lag_+0"}

def _n_pred(series):
    return int(series.notna().sum())

# Full test
full_rows = []
for lag in LAGS:
    col = lag_cols[lag]
    full_rows.append({
        "Vintage": lag,
        "RMSE":    round(rmse(actuals, df[col].tolist()), 6),
        "MAE":     round(mae(actuals,  df[col].tolist()), 6),
        "n":       _n_pred(df[col]),
    })
perf_full = pd.DataFrame(full_rows)

# Stratified
strat_rows = []
for period, label in [
    ("crisis", "Crisis (2020Q1-2021Q2)"),
    ("normal", "Normal (2021Q3-2024Q2)"),
]:
    sub = df.loc[df["period_type"] == period]
    y   = sub["actual"].tolist()
    for lag in LAGS:
        col  = lag_cols[lag]
        yhat = sub[col].tolist()
        strat_rows.append({
            "Period":  label,
            "Vintage": lag,
            "n":       sum(1 for v in yhat if not pd.isna(v)),
            "RMSE":    round(rmse(y, yhat), 6),
            "MAE":     round(mae(y,  yhat), 6),
        })
perf_strat = pd.DataFrame(strat_rows)

# --- Print ------------------------------------------------------------------
print("\n" + "=" * 64)
print("FULL TEST PERIOD — DFM (2020Q1–2024Q2, post-processed)")
print("=" * 64)
print(perf_full.to_string(index=False))

print("\n" + "=" * 64)
print("STRATIFIED — Crisis vs Normal")
print("=" * 64)
print(perf_strat.to_string(index=False))

# --- Save -------------------------------------------------------------------
OUT_PRED  = OUT_DIR / "dfm_predictions_v2.csv"
OUT_FULL  = OUT_DIR / "dfm_performance_full_v2.csv"
OUT_STRAT = OUT_DIR / "dfm_performance_stratified_v2.csv"

df.to_csv(OUT_PRED,   index=False)
perf_full.to_csv(OUT_FULL,   index=False)
perf_strat.to_csv(OUT_STRAT, index=False)

print(f"\nSaved:")
print(f"  {OUT_PRED}")
print(f"  {OUT_FULL}")
print(f"  {OUT_STRAT}")
```
