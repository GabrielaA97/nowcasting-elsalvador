```python
# =============================================================================
# ARMA MODEL -- Monthly IVAE Nowcasting (El Salvador)
# =============================================================================
# Univariate benchmark model. Uses only the history of IVAE_TOT m-o-m
# growth rate to predict the current month's value.
#
# Role in the paper:
#   The ARMA serves as the univariate benchmark -- the simplest possible
#   model. If a multivariate model (Ridge, XGBoost, DFM, etc.) cannot
#   beat ARMA, it has no value added.
#
# Evaluation:
#   - Two windows: Crisis (2019M3-2021M6) and Post-Pandemic (2021M9-2024M6)
#   - Three vintages: lag -2, -1, 0
#     lag -2 = predict this month using data available 2 months ago
#     lag -1 = predict this month using data available 1 month ago
#     lag  0 = predict this month using all data available this month
#
# Metrics: RMSE, MAE, R2
# =============================================================================

# -- Setup: working directory -------------------------------------------------
import os
os.chdir("/Users/gabrielaaquino/Desktop/real-time-gdp-nowcasting-sv/Models")

# -- Imports ------------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings("ignore")

import pmdarima as pm
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.stats.diagnostic import acorr_ljungbox
from scipy import stats

from utils import (
    CONFIG,
    load_data,
    gen_lagged_data,
    compute_metrics,
    get_monthly_test_dates,
    print_performance_table,
)

plt.rcParams["figure.figsize"] = [12, 5]

# =============================================================================
# 1. CONFIGURATION
# =============================================================================

TARGET   = CONFIG["target_variable"]    # IVAE_TOT
LAGS_USE = CONFIG["lags"]               # [-2, -1, 0]

print("=" * 55)
print("  ARMA -- Monthly IVAE Nowcasting")
print("=" * 55)
print(f"Target   : {TARGET} (m-o-m growth rate)")
print(f"Vintages : {LAGS_USE}")
print(f"Window 1 : {CONFIG['window_1']['name']}")
print(f"Window 2 : {CONFIG['window_2']['name']}")

# =============================================================================
# 2. LOAD DATA
# =============================================================================

data, metadata = load_data()

panel = (
    data.loc[data.date >= CONFIG["train_start_date"]]
    .reset_index(drop=True)
)

# Quick look at target
print(f"\nIVAE_TOT last 6 months:")
print(data[["date", TARGET]].dropna().tail(6).to_string(index=False))
```

```python




# =============================================================================
# 4. RESIDUAL DIAGNOSTICS
# =============================================================================

print("\n-- Residual Diagnostics (training fit) --")
residuals = auto_model.resid()

jb_stat, jb_p = stats.jarque_bera(residuals)
print(f"Jarque-Bera    : stat={jb_stat:.4f}  p={jb_p:.4f}  "
      f"({'normal' if jb_p > 0.05 else 'non-normal'})")

lb = acorr_ljungbox(residuals, lags=[6, 12], return_df=True)
print(f"Ljung-Box(6)   : p={lb['lb_pvalue'].iloc[0]:.4f}  "
      f"({'no autocorr' if lb['lb_pvalue'].iloc[0] > 0.05 else 'autocorr detected'})")
print(f"Ljung-Box(12)  : p={lb['lb_pvalue'].iloc[1]:.4f}  "
      f"({'no autocorr' if lb['lb_pvalue'].iloc[1] > 0.05 else 'autocorr detected'})")

from statsmodels.stats.stattools import durbin_watson
dw = durbin_watson(residuals)
print(f"Durbin-Watson  : {dw:.4f}  (2=no autocorr)")

# =============================================================================
# 5. PREDICTION FUNCTION
# =============================================================================

def predict_arma(data, metadata, target_date, lag, ar_order, ma_order):
    """
    Predict IVAE_TOT for target_date using data available at 'lag' offset.

    Parameters
    ----------
    target_date : str  e.g. "2020-04-01"
    lag         : int  -2, -1, or 0
                  -2 = use data available 2 months before target_date
                  -1 = use data available 1 month before target_date
                   0 = use all data available at target_date

    Returns
    -------
    float : predicted IVAE_TOT m-o-m growth rate, or np.nan if failed
    """
    lagged = gen_lagged_data(metadata, data, target_date, lag)
    series = lagged[TARGET].dropna().values

    # Need enough history for the model
    if len(series) < max(ar_order + ma_order + 5, 10):
        return np.nan

    try:
        fit  = ARIMA(series, order=(ar_order, 0, ma_order)).fit(
            method_kwargs={"maxiter": 500}
        )
        return float(fit.forecast(steps=1)[0])
    except Exception:
        return np.nan

# =============================================================================
# 6. NOWCAST LOOP -- MONTHLY, TWO WINDOWS
# =============================================================================

def run_window(window_cfg, window_label):
    """
    Run monthly nowcast loop for one evaluation window.

    For each month in the window:
      - Get data available at each lag (-2, -1, 0)
      - Fit ARMA on available IVAE_TOT history
      - Predict one step ahead
      - Compare to realized IVAE_TOT

    Returns
    -------
    perf      : pd.DataFrame  metrics by vintage
    pred_dict : dict          {lag: {date_str: prediction}}
    actuals   : dict          {date_str: actual_value}
    """
    print(f"\n{'='*55}\n  Window: {window_label}\n{'='*55}")

    dates_df     = get_monthly_test_dates(window=window_cfg)
    all_months   = dates_df["date"].tolist()
    pred_dict    = {lag: {} for lag in LAGS_USE}
    actuals_dict = {}

    for month in all_months:
        date_str   = str(month.date())
        actual_row = data.loc[data.date == month, TARGET]

        if actual_row.empty or actual_row.isna().all():
            continue

        actuals_dict[date_str] = float(actual_row.values[0])

        for lag in LAGS_USE:
            pred_dict[lag][date_str] = predict_arma(
                data, metadata, date_str, lag, AR_ORDER, MA_ORDER
            )

        print(f"  {date_str}: "
              f"actual={actuals_dict[date_str]*100:.2f}%  "
              f"lag-2={pred_dict[-2].get(date_str,np.nan)*100:.2f}%  "
              f"lag-1={pred_dict[-1].get(date_str,np.nan)*100:.2f}%  "
              f"lag 0={pred_dict[0].get(date_str,np.nan)*100:.2f}%")

    # -- Metrics by vintage
    rows = []
    for lag in LAGS_USE:
        common = [
            d for d in actuals_dict
            if d in pred_dict[lag] and not np.isnan(pred_dict[lag][d])
        ]
        if not common:
            continue
        a = np.array([actuals_dict[d]   for d in common])
        f = np.array([pred_dict[lag][d] for d in common])
        m = compute_metrics(a, f, label=str(lag))
        m["Vintage"] = lag
        rows.append(m)

    perf = pd.DataFrame(rows)[["Vintage", "RMSE", "MAE", "R2", "n"]]
    print_performance_table(perf.round(4), f"ARMA({AR_ORDER},{MA_ORDER}) -- {window_label}")
    return perf, pred_dict, actuals_dict

# -- Run both windows
perf_w1, preds_w1, actuals_w1 = run_window(CONFIG["window_1"], "Crisis (COVID-19)")
perf_w2, preds_w2, actuals_w2 = run_window(CONFIG["window_2"], "Post-Pandemic")

# =============================================================================
# 7. COMBINED PERFORMANCE TABLE
# =============================================================================

perf_w1["Window"] = "Crisis"
perf_w2["Window"] = "Post-Pandemic"
combined = pd.concat([perf_w1, perf_w2], ignore_index=True)
combined["Model"] = f"ARMA({AR_ORDER},{MA_ORDER})"

print("\n-- Combined Performance --")
print(combined[["Model", "Window", "Vintage", "RMSE", "MAE", "R2", "n"]]
      .to_string(index=False))

# =============================================================================
# 8. VISUALIZATIONS
# =============================================================================

def plot_window(preds, actuals_dict, window_label):
    dates   = sorted(actuals_dict.keys())
    dt      = [pd.Timestamp(d) for d in dates]
    actual  = [actuals_dict[d] * 100 for d in dates]
    pred_0  = [preds[0].get(d,  np.nan) * 100 for d in dates]
    pred_m1 = [preds[-1].get(d, np.nan) * 100 for d in dates]
    pred_m2 = [preds[-2].get(d, np.nan) * 100 for d in dates]

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(13, 9))

    # -- Panel 1: Actual vs predictions
    ax1.plot(dt, actual,  color="black",   lw=1.8, label="Actual IVAE m-o-m", zorder=5)
    ax1.plot(dt, pred_0,  color="#2166ac", lw=1.2, ls="--",  label="Nowcast (lag 0)")
    ax1.plot(dt, pred_m1, color="#f46d43", lw=1.0, ls=":",   label="lag -1", alpha=0.85)
    ax1.plot(dt, pred_m2, color="#abdda4", lw=1.0, ls="-.",  label="lag -2", alpha=0.85)
    ax1.axhline(0, color="gray", lw=0.6, ls="--")
    ax1.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
    ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
    plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45, ha="right")
    ax1.set_title(f"ARMA({AR_ORDER},{MA_ORDER}) | {window_label}",
                  fontsize=12, fontweight="bold")
    ax1.set_ylabel("IVAE m-o-m (%)")
    ax1.legend(fontsize=9)
    ax1.grid(True, alpha=0.3)

    # -- Panel 2: Forecast errors (lag 0)
    valid  = [(t, a, p) for t, a, p in zip(dt, actual, pred_0) if not np.isnan(p)]
    err_dt = [v[0] for v in valid]
    errors = [v[1] - v[2] for v in valid]
    colors = ["#d73027" if e < 0 else "#4dac26" for e in errors]
    ax2.bar(err_dt, errors, color=colors, width=20, alpha=0.85)
    ax2.axhline(0, color="black", lw=0.8)
    ax2.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
    ax2.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
    plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45, ha="right")
    ax2.set_title("Forecast Errors -- Actual minus Predicted (lag 0)", fontsize=11)
    ax2.set_ylabel("Error (pp)")
    ax2.grid(True, alpha=0.3)

    plt.suptitle(
        f"ARMA | El Salvador IVAE Monthly Nowcasting | {window_label}",
        fontsize=13, fontweight="bold"
    )
    plt.tight_layout()
    plt.show()


def plot_metrics(perf_w1, perf_w2):
    """Side-by-side RMSE comparison across vintages and windows."""
    fig, axes = plt.subplots(1, 2, figsize=(12, 4), sharey=False)

    for ax, perf, label in zip(
        axes,
        [perf_w1, perf_w2],
        ["Crisis (COVID-19)", "Post-Pandemic"]
    ):
        x     = np.arange(len(LAGS_USE))
        width = 0.35
        ax.bar(x - width/2, perf["RMSE"] * 100, width,
               label="RMSE", color="#2166ac", alpha=0.85)
        ax.bar(x + width/2, perf["MAE"]  * 100, width,
               label="MAE",  color="#d73027", alpha=0.85)
        ax.set_xticks(x)
        ax.set_xticklabels(["lag -2\n(Month 1)", "lag -1\n(Month 2)", "lag 0\n(Month 3)"])
        ax.set_title(f"ARMA | {label}", fontsize=11, fontweight="bold")
        ax.set_ylabel("Error (pp)")
        ax.legend(fontsize=9)
        ax.grid(True, alpha=0.3, axis="y")

    plt.suptitle("ARMA -- Forecast Error by Vintage", fontsize=13, fontweight="bold")
    plt.tight_layout()
    plt.show()


plot_window(preds_w1, actuals_w1, "Crisis (COVID-19)")
plot_window(preds_w2, actuals_w2, "Post-Pandemic Normalisation")
plot_metrics(perf_w1, perf_w2)

# =============================================================================
# 9. SAVE RESULTS
# =============================================================================

def build_output(preds, actuals_dict, window_name):
    dates = sorted(actuals_dict.keys())
    return pd.DataFrame({
        "date"        : dates,
        "actual"      : [actuals_dict[d]          for d in dates],
        "pred_lag_m2" : [preds[-2].get(d, np.nan) for d in dates],
        "pred_lag_m1" : [preds[-1].get(d, np.nan) for d in dates],
        "pred_lag_0"  : [preds[0].get(d,  np.nan) for d in dates],
        "model"       : f"ARMA({AR_ORDER},{MA_ORDER})",
        "window"      : window_name,
    })

results = pd.concat([
    build_output(preds_w1, actuals_w1, "Crisis"),
    build_output(preds_w2, actuals_w2, "Post-Pandemic"),
], ignore_index=True)

os.makedirs("outputs", exist_ok=True)
results.to_csv("outputs/arma_predictions.csv", index=False)
combined.to_csv("outputs/arma_performance.csv", index=False)

print(f"\nARMA complete.")
print(f"Predictions saved : outputs/arma_predictions.csv ({results.shape[0]} rows)")
print(f"Performance saved : outputs/arma_performance.csv")
print(f"\nNext model: Ridge")
```
