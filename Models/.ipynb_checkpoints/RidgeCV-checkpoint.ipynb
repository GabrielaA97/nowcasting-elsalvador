```python
# =============================================================================
# LASSO REGRESSION -- Monthly IVAE Nowcasting (El Salvador)
# =============================================================================
# Regularized regression with L1 penalty (variable selection).
# Alpha selected via time-series cross-validation (expanding window).
#
# Role in paper:
#   Lasso performs automatic variable selection -- coefficients of
#   irrelevant predictors are shrunk exactly to zero. With 123 predictors,
#   Lasso identifies the sparse set of indicators most predictive of
#   IVAE_TOT m-o-m growth. Complements Ridge (L2) which shrinks all
#   coefficients but never eliminates them.
#
# Key difference from Ridge:
#   Ridge (L2): shrinks all coefficients toward zero, keeps all variables
#   Lasso (L1): sets some coefficients exactly to zero (variable selection)
#   --> Lasso is more interpretable: tells you WHICH variables matter
#
# CV implementation:
#   LassoCV with TimeSeriesSplit (expanding window) -- no data leakage.
#   Alpha grid searched on log scale from 1e-4 to 1e2.
# =============================================================================

# -- Setup --------------------------------------------------------------------
import os
os.chdir("/Users/gabrielaaquino/Desktop/real-time-gdp-nowcasting-sv/Models")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import warnings
warnings.filterwarnings("ignore")

from sklearn.linear_model import LassoCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import TimeSeriesSplit
from collections import Counter

from utils import (
    CONFIG,
    load_data,
    gen_lagged_data,
    flatten_data,
    mean_fill_dataset,
    compute_metrics,
    get_monthly_test_dates,
    print_performance_table,
)

plt.rcParams["figure.figsize"] = [12, 5]

# =============================================================================
# 1. CONFIGURATION
# =============================================================================

TARGET   = CONFIG["target_variable"]   # IVAE_TOT
LAGS_USE = CONFIG["lags"]              # [-2, -1, 0]
N_LAGS   = CONFIG["n_lags"]            # 4 monthly lags per variable

# Alpha grid on log scale -- Lasso needs finer grid than Ridge
ALPHAS = np.logspace(-4, 2, 50)       # 50 values from 0.0001 to 100

# TimeSeriesSplit: expanding window -- no data leakage
TS_CV = TimeSeriesSplit(n_splits=5)

print("=" * 55)
print("  LASSO -- Monthly IVAE Nowcasting")
print("=" * 55)
print(f"Target    : {TARGET} (m-o-m growth rate)")
print(f"Vintages  : {LAGS_USE}")
print(f"Feat lags : {N_LAGS} months per variable")
print(f"Alpha grid: log-scale [{ALPHAS[0]:.4f} ... {ALPHAS[-1]:.1f}] ({len(ALPHAS)} values)")
print(f"CV        : TimeSeriesSplit(n_splits=5) -- expanding window")

# =============================================================================
# 2. LOAD DATA
# =============================================================================

data, metadata = load_data()

# =============================================================================
# 3. NOWCAST LOOP -- TWO WINDOWS
# =============================================================================

def run_window(window_cfg, window_label):
    """
    Run monthly nowcast loop for one evaluation window.

    For each month:
      - Train Lasso with TimeSeriesSplit CV (one model per month)
      - Apply to 3 ragged-edge vintages [-2, -1, 0]
      - Track: alpha selected, number of non-zero coefficients
    """
    print(f"\n{'='*55}\n  Window: {window_label}\n{'='*55}")

    dates_df     = get_monthly_test_dates(window=window_cfg)
    all_months   = dates_df["date"].tolist()
    pred_dict    = {lag: {} for lag in LAGS_USE}
    actuals_dict = {}
    alphas_used  = []
    n_nonzero    = []

    for month in all_months:
        date_str     = str(month.date())
        train_cutoff = str((month - pd.DateOffset(months=1)).date())

        actual_row = data.loc[data.date == month, TARGET]
        if actual_row.empty or actual_row.isna().all():
            continue
        actuals_dict[date_str] = float(actual_row.values[0])

        # -- Build training set
        try:
            train = data.loc[
                (data.date >= pd.Timestamp(CONFIG["train_start_date"])) &
                (data.date <= pd.Timestamp(train_cutoff))
            ].copy().reset_index(drop=True)

            train_flat = mean_fill_dataset(train, train)
            train_flat = flatten_data(train_flat, TARGET, N_LAGS)
            train_flat = (train_flat
                          .dropna(subset=[TARGET])
                          .dropna(axis=0, how="any")
                          .reset_index(drop=True))

            if len(train_flat) < 10:
                for lag in LAGS_USE:
                    pred_dict[lag][date_str] = np.nan
                continue

            X_train = train_flat.drop(["date", TARGET], axis=1)
            y_train = train_flat[TARGET]

            # -- Fit LassoCV with TimeSeriesSplit
            # LassoCV searches alpha internally -- more efficient than manual grid
            scaler  = StandardScaler()
            X_scaled = scaler.fit_transform(X_train)

            lasso_cv = LassoCV(
                alphas    = ALPHAS,
                cv        = TS_CV,
                max_iter  = 5000,    # increase for convergence with many features
                n_jobs    = -1       # use all CPU cores
            )
            lasso_cv.fit(X_scaled, y_train)

            alphas_used.append(lasso_cv.alpha_)
            n_nonzero.append(np.sum(lasso_cv.coef_ != 0))

        except Exception:
            for lag in LAGS_USE:
                pred_dict[lag][date_str] = np.nan
            continue

        # -- Predict for each vintage with the SAME fitted model
        for lag in LAGS_USE:
            try:
                lagged        = gen_lagged_data(metadata, data, date_str, lag)
                lagged_filled = mean_fill_dataset(train, lagged)
                lagged_flat   = flatten_data(lagged_filled, TARGET, N_LAGS)
                target_row    = lagged_flat[
                    lagged_flat.date == pd.Timestamp(date_str)
                ]

                if target_row.empty:
                    pred_dict[lag][date_str] = np.nan
                    continue

                X_test = target_row.drop(["date", TARGET], axis=1)

                # Align columns
                for col in set(X_train.columns) - set(X_test.columns):
                    X_test[col] = 0.0
                X_test = X_test[X_train.columns]

                X_test_scaled = scaler.transform(X_test)
                pred_dict[lag][date_str] = float(lasso_cv.predict(X_test_scaled)[0])

            except Exception:
                pred_dict[lag][date_str] = np.nan

        print(f"  {date_str}: "
              f"actual={actuals_dict[date_str]*100:+.2f}%  "
              f"lag-2={pred_dict[-2].get(date_str, np.nan)*100:+.2f}%  "
              f"lag-1={pred_dict[-1].get(date_str, np.nan)*100:+.2f}%  "
              f"lag 0={pred_dict[0].get(date_str, np.nan)*100:+.2f}%  "
              f"alpha={lasso_cv.alpha_:.4f}  "
              f"nonzero={np.sum(lasso_cv.coef_ != 0)}")

    # -- Alpha and sparsity summary
    if alphas_used:
        alpha_counts = Counter([round(a, 6) for a in alphas_used])
        print(f"\n  Alpha most frequent : {alpha_counts.most_common(1)[0][0]}")
        print(f"  Alpha mean          : {np.mean(alphas_used):.4f}")
        print(f"  Non-zero coefs mean : {np.mean(n_nonzero):.1f} / {len(X_train.columns)}")
        print(f"  Sparsity            : {(1 - np.mean(n_nonzero)/len(X_train.columns))*100:.1f}% of features zeroed out")

    # -- Metrics
    rows = []
    for lag in LAGS_USE:
        common = [
            d for d in actuals_dict
            if d in pred_dict[lag] and pd.notna(pred_dict[lag][d])
        ]
        if not common:
            continue
        a = np.array([actuals_dict[d]   for d in common])
        f = np.array([pred_dict[lag][d] for d in common])
        m = compute_metrics(a, f, label=str(lag))
        m["Vintage"] = lag
        rows.append(m)

    perf = pd.DataFrame(rows)[["Vintage", "RMSE", "MAE", "R2", "n"]]
    print_performance_table(perf.round(4), f"Lasso -- {window_label}")
    return perf, pred_dict, actuals_dict, lasso_cv, scaler, X_train


print("\n>>> Running Window 1: Crisis (COVID-19)")
perf_w1, preds_w1, actuals_w1, lasso_last, scaler_last, X_train_last = run_window(
    CONFIG["window_1"], "Crisis (COVID-19)"
)

print("\n>>> Running Window 2: Post-Pandemic")
perf_w2, preds_w2, actuals_w2, lasso_last, scaler_last, X_train_last = run_window(
    CONFIG["window_2"], "Post-Pandemic Normalisation"
)

# =============================================================================
# 4. COMBINED PERFORMANCE TABLE
# =============================================================================

perf_w1["Window"] = "Crisis"
perf_w2["Window"] = "Post-Pandemic"
combined = pd.concat([perf_w1, perf_w2], ignore_index=True)
combined["Model"] = "Lasso"

print("\n-- Combined Performance --")
print(combined[["Model", "Window", "Vintage", "RMSE", "MAE", "R2", "n"]]
      .to_string(index=False))

# =============================================================================
# 5. VARIABLE SELECTION ANALYSIS
# =============================================================================
# Key advantage of Lasso: tells you WHICH variables matter

print("\n--- Lasso Variable Selection (last fitted model) ---")
try:
    coef_series = pd.Series(
        lasso_cv.coef_,
        index=X_train_last.columns
    )

    selected    = coef_series[coef_series != 0].sort_values(key=abs, ascending=False)
    zeroed_out  = coef_series[coef_series == 0]

    print(f"Total features    : {len(coef_series)}")
    print(f"Selected (nonzero): {len(selected)}")
    print(f"Zeroed out        : {len(zeroed_out)}")
    print(f"\nTop 15 selected variables:")
    print(selected.head(15).round(4).to_string())

    # Plot selected variables
    top_selected = selected.head(20).sort_values(key=abs)
    colors = ["#d73027" if v < 0 else "#2166ac" for v in top_selected]

    fig, ax = plt.subplots(figsize=(10, 7))
    ax.barh(top_selected.index, top_selected.values, color=colors, alpha=0.85)
    ax.axvline(0, color="black", lw=0.8)
    ax.set_title("Lasso -- Selected Variables and Coefficients\n"
                 "(Blue = positive effect, Red = negative effect)",
                 fontsize=12, fontweight="bold")
    ax.set_xlabel("Coefficient")
    ax.grid(True, alpha=0.3, axis="x")
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"  Variable selection analysis failed: {e}")

# =============================================================================
# 6. VISUALIZATIONS
# =============================================================================

def plot_window(preds, actuals_dict, window_label):
    dates   = sorted(actuals_dict.keys())
    dt      = [pd.Timestamp(d) for d in dates]
    actual  = [actuals_dict[d] * 100 for d in dates]
    pred_0  = [preds[0].get(d,  np.nan) * 100 for d in dates]
    pred_m1 = [preds[-1].get(d, np.nan) * 100 for d in dates]
    pred_m2 = [preds[-2].get(d, np.nan) * 100 for d in dates]

    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(13, 9))

    ax1.plot(dt, actual,  color="black",   lw=1.8, label="Actual IVAE m-o-m")
    ax1.plot(dt, pred_0,  color="#2166ac", lw=1.2, ls="--", label="Nowcast (lag 0)")
    ax1.plot(dt, pred_m1, color="#f46d43", lw=1.0, ls=":",  label="lag -1", alpha=0.85)
    ax1.plot(dt, pred_m2, color="#abdda4", lw=1.0, ls="-.", label="lag -2", alpha=0.85)
    ax1.axhline(0, color="gray", lw=0.6, ls="--")
    ax1.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
    ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
    plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45, ha="right")
    ax1.set_title(f"Lasso | {window_label}", fontsize=12, fontweight="bold")
    ax1.set_ylabel("IVAE m-o-m (%)")
    ax1.legend(fontsize=9)
    ax1.grid(True, alpha=0.3)

    valid  = [(t, a, p) for t, a, p in zip(dt, actual, pred_0)
              if not np.isnan(p)]
    if valid:
        err_dt = [v[0] for v in valid]
        errors = [v[1] - v[2] for v in valid]
        colors_bar = ["#d73027" if e < 0 else "#4dac26" for e in errors]
        ax2.bar(err_dt, errors, color=colors_bar, width=20, alpha=0.85)
    ax2.axhline(0, color="black", lw=0.8)
    ax2.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m"))
    ax2.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
    plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45, ha="right")
    ax2.set_title("Forecast Errors (Actual - Predicted, lag 0)", fontsize=11)
    ax2.set_ylabel("Error (pp)")
    ax2.grid(True, alpha=0.3)

    plt.suptitle(f"Lasso | El Salvador IVAE | {window_label}",
                 fontsize=13, fontweight="bold")
    plt.tight_layout()
    plt.show()


plot_window(preds_w1, actuals_w1, "Crisis (COVID-19)")
plot_window(preds_w2, actuals_w2, "Post-Pandemic Normalisation")

# =============================================================================
# 7. SAVE RESULTS
# =============================================================================

def build_output(preds, actuals_dict, window_name):
    dates = sorted(actuals_dict.keys())
    return pd.DataFrame({
        "date"        : dates,
        "actual"      : [actuals_dict[d]          for d in dates],
        "pred_lag_m2" : [preds[-2].get(d, np.nan) for d in dates],
        "pred_lag_m1" : [preds[-1].get(d, np.nan) for d in dates],
        "pred_lag_0"  : [preds[0].get(d,  np.nan) for d in dates],
        "model"       : "Lasso",
        "window"      : window_name,
    })

results = pd.concat([
    build_output(preds_w1, actuals_w1, "Crisis"),
    build_output(preds_w2, actuals_w2, "Post-Pandemic"),
], ignore_index=True)

os.makedirs("outputs", exist_ok=True)
results.to_csv("outputs/lasso_predictions.csv", index=False)
combined.to_csv("outputs/lasso_performance.csv", index=False)

print(f"\nLasso complete.")
print(f"Predictions : outputs/lasso_predictions.csv ({len(results)} rows)")
print(f"Performance : outputs/lasso_performance.csv")
print(f"\nNext: Decision Tree")

```
