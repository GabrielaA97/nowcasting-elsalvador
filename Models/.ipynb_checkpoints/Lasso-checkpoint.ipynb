```python
# =============================================================================
# lasso_final.py
# Lasso Regression nowcasting pipeline for El Salvador GDP.
# Alpha selected via fixed holdout (last 8 quarters of training window).
# OLS baseline included for reference.
#
# Author : Gabriela Aquino
# =============================================================================
#
# TUNABLE PARAMETERS (safe to modify — see Section 1):
#   N_LAGS_FEATURES     : monthly lags per predictor (baseline = 4)
#   ALPHA_LOG_MIN/MAX   : log-spaced alpha grid (Lasso needs smaller alphas
#                         than Ridge)
#   N_HOLDOUT_QUARTERS  : size of fixed holdout for alpha selection (default 8)
#
# SHARED CONFIG (do not modify — comes from nowcast_utils.py):
#   LAGS, TEST_START, TEST_END, CRISIS_START/END, NORMAL_START/END, etc.
# =============================================================================

import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import warnings
from pathlib import Path

from sklearn.linear_model import Lasso, Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

warnings.filterwarnings("ignore")

# =============================================================================
# 0. PATHS AND SHARED-CONFIG IMPORT
# =============================================================================
BASE_DIR  = Path.home() / "Desktop" / "ESA-gdp-nowcasting"
DATA_DIR  = BASE_DIR / "Data"
MODEL_DIR = BASE_DIR / "Models"
OUT_DIR   = MODEL_DIR / "results" / "lasso"
OUT_DIR.mkdir(parents=True, exist_ok=True)
sys.path.insert(0, str(MODEL_DIR))

from nowcast_utils import (
    LAGS, TARGET_VARIABLE,
    TEST_START, TEST_END,
    CRISIS_START, CRISIS_END,
    NORMAL_START, NORMAL_END,
    RANDOM_SEED,
    gen_lagged_data, flatten_data, mean_fill_dataset,
    classify_crisis_periods,
)

PATH_DATA = DATA_DIR / "data_tf.csv"
PATH_META = DATA_DIR / "meta_data_v2.csv"
for p in [PATH_DATA, PATH_META]:
    if not p.exists():
        raise FileNotFoundError(f"\n  File not found: {p}")

# =============================================================================
# 1. TUNABLE PARAMETERS (Lasso-specific)
# =============================================================================
TRAIN_START: str = "2005-01-01"     # earliest date used for training

# Number of monthly lags per predictor (same as Ridge for comparability)
# Try: 2 (light), 4 (baseline), 6 (extended)
N_LAGS_FEATURES: int = 4

# Alpha grid: Lasso needs smaller alphas than Ridge (L1 is stronger).
# If sparsity is always 0%, raise ALPHA_LOG_MIN.
# If sparsity is always 100%, lower ALPHA_LOG_MAX.
ALPHA_LOG_MIN: int      = -6   # 1e-6
ALPHA_LOG_MAX: int      =  0   # 1e0  (1.0)
ALPHA_N_CANDIDATES: int = 40
ALPHA_GRID = np.logspace(ALPHA_LOG_MIN, ALPHA_LOG_MAX, ALPHA_N_CANDIDATES)

# Fixed holdout: last N quarters of training used for alpha selection.
# 8 quarters = 2 years of validation. Increase to 12 for more stability.
N_HOLDOUT_QUARTERS: int = 8

# Lasso solver
LASSO_MAX_ITER: int = 10000
LASSO_TOL:      float = 1e-4

np.random.seed(RANDOM_SEED)

# =============================================================================
# 2. LOAD DATA
# =============================================================================
print("Loading data...")
data     = pd.read_csv(PATH_DATA, parse_dates=["date"]).sort_values("date").reset_index(drop=True)
metadata = pd.read_csv(PATH_META)

print(f"  Rows       : {data.shape[0]}")
print(f"  Columns    : {data.shape[1]}")
print(f"  Date range : {data['date'].min().date()} to {data['date'].max().date()}")
print(f"  GDP obs    : {data[TARGET_VARIABLE].notna().sum()} quarterly observations")

test_dates = pd.date_range(TEST_START, TEST_END, freq="3MS").strftime("%Y-%m-%d").tolist()

# =============================================================================
# 3. ROLLING-ORIGIN EVALUATION
# =============================================================================
print("\nRunning rolling-origin evaluation...")
print(f"  Test window     : {TEST_START} to {TEST_END}")
print(f"  Vintages        : {LAGS}")
print(f"  Lag depth       : {N_LAGS_FEATURES} monthly lags per predictor")
print(f"  Alpha grid      : {ALPHA_N_CANDIDATES} candidates in [1e{ALPHA_LOG_MIN}, 1e{ALPHA_LOG_MAX}]")
print(f"  Alpha selection : fixed holdout ({N_HOLDOUT_QUARTERS} quarters)\n")

pred_dict               = {lag: [] for lag in LAGS}
pred_dict_ols           = {lag: [] for lag in LAGS}
dates_out               = []
actuals_out             = []
lambdas_out             = []
sparsity_out            = []
feature_importance_last = None
train_cutoff_last       = None

for nowcast_date in test_dates:

    # --- Actual GDP value ----------------------------------------------------
    actual_rows = data.loc[
        data["date"].astype(str).str.startswith(nowcast_date[:7]),
        TARGET_VARIABLE
    ]
    if actual_rows.empty or actual_rows.isna().all():
        print(f"  [{nowcast_date}] No actual GDP - skipping.")
        for lag in LAGS:
            pred_dict[lag].append(np.nan)
            pred_dict_ols[lag].append(np.nan)
        dates_out.append(nowcast_date)
        actuals_out.append(np.nan)
        lambdas_out.append(np.nan)
        sparsity_out.append(np.nan)
        continue

    dates_out.append(nowcast_date)
    actuals_out.append(float(actual_rows.values[0]))

    # --- Training window -----------------------------------------------------
    train_cutoff      = (pd.to_datetime(nowcast_date) - pd.DateOffset(months=3)).strftime("%Y-%m-%d")
    train_cutoff_last = train_cutoff

    train_raw = data.loc[
        (data["date"] >= TRAIN_START) & (data["date"] <= train_cutoff)
    ].copy()

    n_q = train_raw[TARGET_VARIABLE].notna().sum()

    # --- Build feature matrix ------------------------------------------------
    train_filled = mean_fill_dataset(training=train_raw, test=train_raw)
    train_flat   = flatten_data(train_filled, TARGET_VARIABLE, N_LAGS_FEATURES)
    train_flat   = (
        train_flat
        .loc[train_flat["date"].dt.month.isin([3, 6, 9, 12])]
        .dropna(axis=0, how="any")
        .reset_index(drop=True)
    )

    if len(train_flat) < N_HOLDOUT_QUARTERS + 5:
        print(f"  [{nowcast_date}] Only {len(train_flat)} rows - skipping.")
        for lag in LAGS:
            pred_dict[lag].append(np.nan)
            pred_dict_ols[lag].append(np.nan)
        lambdas_out.append(np.nan)
        sparsity_out.append(np.nan)
        continue

    X_train = train_flat.drop(columns=["date", TARGET_VARIABLE])
    y_train = train_flat[TARGET_VARIABLE]

    # --- Alpha selection via fixed holdout -----------------------------------
    # Scaler fitted on inner-train only — holdout never influences scaling.
    X_inner = X_train.iloc[:-N_HOLDOUT_QUARTERS]
    y_inner = y_train.iloc[:-N_HOLDOUT_QUARTERS]
    X_hold  = X_train.iloc[-N_HOLDOUT_QUARTERS:]
    y_hold  = y_train.iloc[-N_HOLDOUT_QUARTERS:]

    scaler_inner = StandardScaler().fit(X_inner)
    X_inner_s    = scaler_inner.transform(X_inner)
    X_hold_s     = scaler_inner.transform(X_hold)

    # Scaler for final model: fit on full training window (all pre-cutoff)
    scaler  = StandardScaler().fit(X_train)
    X_all_s = scaler.transform(X_train)

    # Search alpha grid
    best_alpha, best_rmse = None, np.inf
    for alpha_candidate in ALPHA_GRID:
        lasso_cv = Lasso(
            alpha=alpha_candidate,
            fit_intercept=True,
            max_iter=LASSO_MAX_ITER,
            tol=LASSO_TOL,
            random_state=RANDOM_SEED,
        )
        lasso_cv.fit(X_inner_s, y_inner)
        val_rmse = np.sqrt(mean_squared_error(y_hold, lasso_cv.predict(X_hold_s)))
        if val_rmse < best_rmse:
            best_rmse  = val_rmse
            best_alpha = alpha_candidate

    lambdas_out.append(best_alpha)

    # --- Refit Lasso on full training with best alpha ------------------------
    final_lasso = Lasso(
        alpha=best_alpha,
        fit_intercept=True,
        max_iter=LASSO_MAX_ITER,
        tol=LASSO_TOL,
        random_state=RANDOM_SEED,
    )
    final_lasso.fit(X_all_s, y_train)

    # Sparsity = fraction of zero coefficients
    n_zero   = int(np.sum(final_lasso.coef_ == 0))
    n_total  = len(final_lasso.coef_)
    sparsity = round(100.0 * n_zero / n_total, 1)
    sparsity_out.append(sparsity)

    # --- OLS baseline --------------------------------------------------------
    ols_model = Ridge(alpha=1e-10, fit_intercept=True)
    ols_model.fit(X_all_s, y_train)

    # --- Diagnostics print ---------------------------------------------------
    min_warn = "HIT_MIN" if best_alpha <= ALPHA_GRID[1]  else "ok"
    max_warn = "HIT_MAX" if best_alpha >= ALPHA_GRID[-2] else "ok"
    n_nonzero = n_total - n_zero
    print(f"  [{nowcast_date}]  alpha={best_alpha:.2e}  n_train={n_q}  "
          f"features={n_total}  nonzero={n_nonzero}  sparsity={sparsity}%  "
          f"grid_min={min_warn}  grid_max={max_warn}")

    # --- Predict for each vintage -------------------------------------------
    for lag in LAGS:
        vintage_data = gen_lagged_data(metadata, data, nowcast_date, lag)
        vintage_data = mean_fill_dataset(training=train_raw, test=vintage_data)
        vintage_flat = flatten_data(vintage_data, TARGET_VARIABLE, N_LAGS_FEATURES)

        target_row = vintage_flat.loc[
            vintage_flat["date"].astype(str).str.startswith(nowcast_date[:7])
        ]
        if target_row.empty:
            pred_dict[lag].append(np.nan)
            pred_dict_ols[lag].append(np.nan)
            continue

        X_test = (
            target_row
            .drop(columns=["date", TARGET_VARIABLE], errors="ignore")
            .reindex(columns=X_train.columns, fill_value=0.0)
        )
        X_test_s = scaler.transform(X_test)

        pred_dict[lag].append(float(final_lasso.predict(X_test_s)[0]))
        pred_dict_ols[lag].append(float(ols_model.predict(X_test_s)[0]))

    feature_importance_last = pd.DataFrame({
        "feature":     X_train.columns,
        "coefficient": final_lasso.coef_,
    }).assign(abs_coef=lambda df: df["coefficient"].abs())

# =============================================================================
# 4. PERFORMANCE METRICS
# =============================================================================

def _rmse(y, yhat):
    y, yhat = np.array(y, float), np.array(yhat, float)
    m = ~np.isnan(y) & ~np.isnan(yhat)
    return float(np.sqrt(np.mean((y[m] - yhat[m])**2))) if m.sum() > 0 else np.nan

def _mae(y, yhat):
    y, yhat = np.array(y, float), np.array(yhat, float)
    m = ~np.isnan(y) & ~np.isnan(yhat)
    return float(np.mean(np.abs(y[m] - yhat[m]))) if m.sum() > 0 else np.nan

# Period labels come from nowcast_utils (single source of truth)
period_labels = classify_crisis_periods(dates_out).tolist()

# --- Full test period --------------------------------------------------------
full_rows = []
for lag in LAGS:
    full_rows.append({
        "Vintage":  lag,
        "RMSE":     round(_rmse(actuals_out, pred_dict[lag]),     6),
        "MAE":      round(_mae(actuals_out,  pred_dict[lag]),     6),
        "RMSE_OLS": round(_rmse(actuals_out, pred_dict_ols[lag]), 6),
        "n":        sum(1 for v in pred_dict[lag] if not np.isnan(v)),
    })
perf_full = pd.DataFrame(full_rows)

# --- Stratified: Crisis / Normal --------------------------------------------
strat_rows = []
for period, label in [
    ("crisis", "Crisis (2020Q1-2021Q2)"),
    ("normal", "Normal (2021Q3-2024Q2)"),
]:
    idx = [i for i, p in enumerate(period_labels) if p == period]
    y   = [actuals_out[i] for i in idx]
    for lag in LAGS:
        yhat     = [pred_dict[lag][i]     for i in idx]
        yhat_ols = [pred_dict_ols[lag][i] for i in idx]
        strat_rows.append({
            "Period":   label,
            "Vintage":  lag,
            "n":        len([v for v in yhat if not np.isnan(v)]),
            "RMSE":     round(_rmse(y, yhat),     6),
            "MAE":      round(_mae(y,  yhat),     6),
            "RMSE_OLS": round(_rmse(y, yhat_ols), 6),
        })
perf_strat = pd.DataFrame(strat_rows)

# --- Print results -----------------------------------------------------------
print("\n" + "=" * 64)
print("FULL TEST PERIOD — Lasso vs OLS")
print("=" * 64)
print(perf_full.to_string(index=False))

print("\n" + "=" * 64)
print("STRATIFIED — Crisis vs Normal")
print("=" * 64)
print(perf_strat.to_string(index=False))

# --- Regularisation benefit summary -----------------------------------------
nowcast_lasso = perf_full.loc[perf_full["Vintage"] == 0, "RMSE"].values[0]
nowcast_ols   = perf_full.loc[perf_full["Vintage"] == 0, "RMSE_OLS"].values[0]
benefit       = (nowcast_ols - nowcast_lasso) / nowcast_ols * 100
avg_sparsity  = np.nanmean(sparsity_out)

print(f"\nRegularisation summary (nowcast, lag=0):")
print(f"  Lasso RMSE   : {nowcast_lasso:.6f}")
print(f"  OLS RMSE     : {nowcast_ols:.6f}")
if benefit > 0:
    print(f"  Lasso is {benefit:.1f}% better than OLS")
else:
    print(f"  OLS is {abs(benefit):.1f}% better than Lasso")
print(f"  Avg sparsity : {avg_sparsity:.1f}% of features zeroed out per window")

# =============================================================================
# 5. SAVE CSVs
# =============================================================================
pred_table = pd.DataFrame({
    "date":           dates_out,
    "actual":         actuals_out,
    **{f"lag_{'-' if lag < 0 else '+'}{abs(lag)}": pred_dict[lag] for lag in LAGS},
    "period_type":    period_labels,
    "selected_alpha": lambdas_out,
    "sparsity_pct":   sparsity_out,
})

pred_table.to_csv(OUT_DIR  / "lasso_predictions.csv",            index=False)
perf_full.to_csv(OUT_DIR   / "lasso_performance_full.csv",       index=False)
perf_strat.to_csv(OUT_DIR  / "lasso_performance_stratified.csv", index=False)

if feature_importance_last is not None:
    (feature_importance_last
     .sort_values("abs_coef", ascending=False)
     .to_csv(OUT_DIR / "lasso_feature_importance.csv", index=False))

print(f"\nCSVs saved to: {OUT_DIR}")

# =============================================================================
# 6. PLOTTING STYLE (matches ARMA for visual consistency)
# =============================================================================
STYLE = {
    "c_crisis":         "#B6213B",
    "c_normal":         "#2A8F93",
    "c_actual":         "#2E2E2E",
    "zero_line_colour": "#555555",
    "shade_crisis":     0.11,
    "shade_normal":     0.08,
}
plt.rcParams.update({
    "font.family":       "serif",
    "axes.titlesize":    13,
    "axes.labelsize":    10.5,
    "xtick.labelsize":   9,
    "ytick.labelsize":   9,
    "legend.fontsize":   9,
    "axes.grid":         True,
    "grid.alpha":        0.20,
    "grid.color":        "#888888",
    "grid.linewidth":    0.5,
    "axes.edgecolor":    "#333333",
    "axes.linewidth":    0.7,
    "axes.spines.top":   False,
    "axes.spines.right": False,
    "figure.dpi":        120,
})

# Vintage styling — only 3 vintages now
LAG_COLORS = {-2: "#a8dadc", -1: "#457b9d", 0: "#1d3557"}
LAG_LW     = {-2: 1.2, -1: 1.2, 0: 2.4}
LAG_LS     = {-2: "--", -1: "--", 0: "-"}
LAG_LABELS = {-2: "lag=-2 (forecast)",
              -1: "lag=-1 (early nowcast)",
               0: "lag= 0 (nowcast)"}


def _format_xaxis(ax):
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))


def _shade_regimes(ax):
    ax.axvspan(pd.to_datetime(CRISIS_START), pd.to_datetime(CRISIS_END),
               alpha=STYLE["shade_crisis"], color=STYLE["c_crisis"],
               edgecolor="none", zorder=0)
    ax.axvspan(pd.to_datetime(NORMAL_START), pd.to_datetime(NORMAL_END),
               alpha=STYLE["shade_normal"], color=STYLE["c_normal"],
               edgecolor="none", zorder=0)


full_series = (
    data[["date", TARGET_VARIABLE]]
    .dropna(subset=[TARGET_VARIABLE])
    .sort_values("date")
    .reset_index(drop=True)
)

# =============================================================================
# 7. PLOT A — Crisis period (2020Q1–2021Q2)
# =============================================================================
crisis_rows = pred_table[pred_table["period_type"] == "crisis"].copy()
crisis_rows["date"] = pd.to_datetime(crisis_rows["date"])

fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(full_series["date"], full_series[TARGET_VARIABLE],
        color="#444444", lw=1.3, marker="o", ms=3, alpha=0.65,
        label="Actual GDP (2005–2024)", zorder=3)
ax.axvspan(pd.to_datetime(CRISIS_START), pd.to_datetime(CRISIS_END),
           alpha=0.13, color=STYLE["c_crisis"], zorder=0,
           label="Crisis window (2020Q1–2021Q2)")
ax.axhline(0, color=STYLE["zero_line_colour"], lw=0.8, alpha=0.4, zorder=1)

for lag in LAGS:
    col = f"lag_{'-' if lag < 0 else '+'}{abs(lag)}"
    ax.plot(crisis_rows["date"], crisis_rows[col],
            color=LAG_COLORS[lag], lw=LAG_LW[lag], ls=LAG_LS[lag],
            marker="o", ms=4.5, label=LAG_LABELS[lag], zorder=5)

ax.set_xlabel("Date")
ax.set_ylabel("GDP growth rate (QoQ)")
ax.set_title("Lasso Regression — Crisis Period (2020Q1–2021Q2)\n"
             "Nowcasting Vintages | El Salvador",
             fontweight="bold")
ax.legend(fontsize=8, ncol=3, loc="lower left")
_format_xaxis(ax)
plt.tight_layout()
fig.savefig(OUT_DIR / "lasso_plot_crisis.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: lasso_plot_crisis.png")

# =============================================================================
# 8. PLOT B — Normal period (2021Q3–2024Q2)
# =============================================================================
normal_rows = pred_table[pred_table["period_type"] == "normal"].copy()
normal_rows["date"] = pd.to_datetime(normal_rows["date"])

fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(full_series["date"], full_series[TARGET_VARIABLE],
        color="#444444", lw=1.3, marker="o", ms=3, alpha=0.65,
        label="Actual GDP (2005–2024)", zorder=3)
ax.axvspan(pd.to_datetime(NORMAL_START), pd.to_datetime(NORMAL_END),
           alpha=0.12, color=STYLE["c_normal"], zorder=0,
           label="Normal window (2021Q3–2024Q2)")
ax.axvspan(pd.to_datetime(CRISIS_START), pd.to_datetime(CRISIS_END),
           alpha=0.07, color=STYLE["c_crisis"], zorder=0,
           label="Crisis period (reference)")
ax.axhline(0, color=STYLE["zero_line_colour"], lw=0.8, alpha=0.4, zorder=1)

for lag in LAGS:
    col = f"lag_{'-' if lag < 0 else '+'}{abs(lag)}"
    ax.plot(normal_rows["date"], normal_rows[col],
            color=LAG_COLORS[lag], lw=LAG_LW[lag], ls=LAG_LS[lag],
            marker="o", ms=4.5, label=LAG_LABELS[lag], zorder=5)

ax.set_xlabel("Date")
ax.set_ylabel("GDP growth rate (QoQ)")
ax.set_title("Lasso Regression — Normal Period (2021Q3–2024Q2)\n"
             "Nowcasting Vintages | El Salvador",
             fontweight="bold")
ax.legend(fontsize=8, ncol=3, loc="lower left")
_format_xaxis(ax)
plt.tight_layout()
fig.savefig(OUT_DIR / "lasso_plot_normal.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: lasso_plot_normal.png")

# =============================================================================
# 9. PLOT C — RMSE heatmap
# =============================================================================
pivot = perf_strat.pivot(index="Period", columns="Vintage", values="RMSE")
fig, ax = plt.subplots(figsize=(9, 3))
sns.heatmap(pivot, annot=True, fmt=".5f", cmap="YlOrRd",
            linewidths=0.6, ax=ax, cbar_kws={"label": "RMSE", "shrink": 0.8})
ax.set_title("Lasso Regression — RMSE by Period and Vintage",
             fontweight="bold", pad=10)
ax.set_xlabel("Vintage (months relative to quarter-end)")
ax.set_ylabel("")
plt.tight_layout()
fig.savefig(OUT_DIR / "lasso_plot_heatmap.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: lasso_plot_heatmap.png")

# =============================================================================
# 10. PLOT D — Feature importance (non-zero coefficients)
# =============================================================================
if feature_importance_last is not None:
    nonzero_feats = feature_importance_last[
        feature_importance_last["coefficient"] != 0
    ].sort_values("abs_coef", ascending=False)

    if len(nonzero_feats) == 0:
        print("All coefficients are zero in last window - alpha may be too large.")
    else:
        top_n     = min(15, len(nonzero_feats))
        top_feats = nonzero_feats.head(top_n)
        bar_cols  = ["#2a9d8f" if c >= 0 else "#e76f51"
                     for c in top_feats["coefficient"]]
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.barh(top_feats["feature"][::-1],
                top_feats["coefficient"][::-1],
                color=bar_cols[::-1], edgecolor="gray", linewidth=0.4)
        ax.axvline(0, color="black", lw=0.8)
        ax.set_xlabel("Lasso coefficient (standardised features)")
        ax.set_title(f"Top {top_n} Non-Zero Features — Lasso\n"
                     f"Last training window ending {train_cutoff_last}",
                     fontweight="bold")
        plt.tight_layout()
        fig.savefig(OUT_DIR / "lasso_plot_features.png",
                    dpi=150, bbox_inches="tight")
        plt.show()
        print("Saved: lasso_plot_features.png")
        print(f"  Non-zero features in last window: {len(nonzero_feats)} / {len(feature_importance_last)}")

# =============================================================================
# 11. PLOT E — Alpha path + sparsity over time
# =============================================================================
valid = [(d, a, s) for d, a, s in zip(dates_out, lambdas_out, sparsity_out)
         if a is not None and not np.isnan(a)]

if valid:
    v_dates    = pd.to_datetime([x[0] for x in valid])
    v_alphas   = [x[1] for x in valid]
    v_sparsity = [x[2] for x in valid]

    fig, axes = plt.subplots(2, 1, figsize=(13, 8))
    fig.suptitle("Lasso Regression — Regularisation Path and Sparsity",
                 fontweight="bold")

    # Panel A: alpha
    ax = axes[0]
    _shade_regimes(ax)
    ax.plot(v_dates, v_alphas, color="#1d3557", lw=1.8,
            marker="o", ms=5, label="Selected alpha")
    ax.axhline(ALPHA_GRID[0],  color="#e63946", lw=0.8, ls="--",
               label=f"Grid min = {ALPHA_GRID[0]:.0e}")
    ax.axhline(ALPHA_GRID[-1], color="#f4a261", lw=0.8, ls="--",
               label=f"Grid max = {ALPHA_GRID[-1]:.0e}")
    ax.set_yscale("log")
    ax.set_ylabel("Selected alpha (log)")
    ax.set_title(f"CV-selected alpha over time (holdout = last {N_HOLDOUT_QUARTERS} quarters)",
                 fontsize=10)
    ax.legend(fontsize=8)
    _format_xaxis(ax)

    # Panel B: sparsity
    ax = axes[1]
    _shade_regimes(ax)
    ax.fill_between(v_dates, v_sparsity, alpha=0.3, color=STYLE["c_normal"])
    ax.plot(v_dates, v_sparsity, color=STYLE["c_normal"], lw=1.8,
            marker="o", ms=5, label="Sparsity %")
    ax.set_ylabel("Features zeroed out (%)")
    ax.set_ylim(0, 105)
    ax.set_title("Sparsity over time — higher % = fewer features retained",
                 fontsize=10)
    ax.legend(fontsize=8)
    _format_xaxis(ax)

    plt.tight_layout()
    fig.savefig(OUT_DIR / "lasso_plot_alpha_sparsity.png",
                dpi=150, bbox_inches="tight")
    plt.show()
    print("Saved: lasso_plot_alpha_sparsity.png")

# =============================================================================
# DONE
# =============================================================================
print("\n" + "=" * 64)
print("LASSO NOWCASTING COMPLETE")
print("=" * 64)
print(f"Output : {OUT_DIR}")
```

```python
# =============================================================================
# Lasso.py
# Lasso Regression nowcasting pipeline for El Salvador GDP.
# Alpha selected via fixed holdout (last 8 quarters of training window).
# Primary benchmark is ARMA (see ARMA.ipynb); no OLS comparison here
# because p >> n makes OLS numerically ill-defined.
#
# Author : Gabriela Aquino
# =============================================================================
#
# TUNABLE PARAMETERS (safe to modify — see Section 1):
#   N_LAGS_FEATURES     : monthly lags per predictor (baseline = 4)
#   ALPHA_LOG_MIN/MAX   : log-spaced alpha grid
#   N_HOLDOUT_QUARTERS  : size of fixed holdout for alpha selection (default 8)
#
# SHARED CONFIG (do not modify — comes from nowcast_utils.py):
#   LAGS, TRAIN_START, TEST_START, TEST_END,
#   CRISIS_START/END, NORMAL_START/END, etc.
# =============================================================================

import sys
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from pathlib import Path

from sklearn.linear_model import Lasso
from sklearn.preprocessing import StandardScaler

# Selective warning filters.
# We WANT to see sklearn ConvergenceWarning (diagnoses alpha / iter issues),
# and our own custom warnings from gen_lagged_data (missing metadata).
# Silence only noisy FutureWarning and UserWarning from third parties.
warnings.filterwarnings("default")
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning)

# =============================================================================
# 0. PATHS AND SHARED-CONFIG IMPORT
# =============================================================================
BASE_DIR  = Path.home() / "Desktop" / "ESA-gdp-nowcasting"
DATA_DIR  = BASE_DIR / "Data"
MODEL_DIR = BASE_DIR / "Models"
OUT_DIR   = MODEL_DIR / "results" / "lasso"
OUT_DIR.mkdir(parents=True, exist_ok=True)
sys.path.insert(0, str(MODEL_DIR))

from nowcast_utils import (
    LAGS, TARGET_VARIABLE,
    TRAIN_START, TEST_START, TEST_END,
    CRISIS_START, CRISIS_END,
    NORMAL_START, NORMAL_END,
    RANDOM_SEED,
    gen_lagged_data, flatten_data, mean_fill_dataset,
    classify_crisis_periods,
    rmse, mae,
)

PATH_DATA = DATA_DIR / "data_tf.csv"
PATH_META = DATA_DIR / "meta_data_v2.csv"
for p in [PATH_DATA, PATH_META]:
    if not p.exists():
        raise FileNotFoundError(f"\n  File not found: {p}")

# =============================================================================
# 1. TUNABLE PARAMETERS (Lasso-specific)
# =============================================================================
# Number of monthly lags per predictor (same as Ridge for comparability)
# Try: 2 (light), 4 (baseline), 6 (extended)
N_LAGS_FEATURES: int = 4

# Alpha grid: Lasso needs smaller alphas than Ridge (L1 is stronger).
# If sparsity is always 0%, raise ALPHA_LOG_MIN.
# If sparsity is always 100%, lower ALPHA_LOG_MAX.
ALPHA_LOG_MIN: int      = -6   # 1e-6
ALPHA_LOG_MAX: int      =  0   # 1e0  (1.0)
ALPHA_N_CANDIDATES: int = 40
ALPHA_GRID = np.logspace(ALPHA_LOG_MIN, ALPHA_LOG_MAX, ALPHA_N_CANDIDATES)

# Fixed holdout: last N quarters of training used for alpha selection.
# 8 quarters = 2 years of validation. Increase to 12 for more stability.
N_HOLDOUT_QUARTERS: int = 8

# Lasso solver — tightened to avoid false sparsity from non-convergence.
LASSO_MAX_ITER: int   = 50000
LASSO_TOL:      float = 1e-5

np.random.seed(RANDOM_SEED)

# =============================================================================
# 2. LOAD DATA
# =============================================================================
print("Loading data...")
data     = pd.read_csv(PATH_DATA, parse_dates=["date"]).sort_values("date").reset_index(drop=True)
metadata = pd.read_csv(PATH_META)

print(f"  Rows       : {data.shape[0]}")
print(f"  Columns    : {data.shape[1]}")
print(f"  Date range : {data['date'].min().date()} to {data['date'].max().date()}")
print(f"  GDP obs    : {data[TARGET_VARIABLE].notna().sum()} quarterly observations")

test_dates = pd.date_range(TEST_START, TEST_END, freq="3MS").strftime("%Y-%m-%d").tolist()

# =============================================================================
# 3. ROLLING-ORIGIN EVALUATION
# =============================================================================
print("\nRunning rolling-origin evaluation...")
print(f"  Train start     : {TRAIN_START}")
print(f"  Test window     : {TEST_START} to {TEST_END}")
print(f"  Vintages        : {LAGS}")
print(f"  Lag depth       : {N_LAGS_FEATURES} monthly lags per predictor")
print(f"  Alpha grid      : {ALPHA_N_CANDIDATES} candidates in [1e{ALPHA_LOG_MIN}, 1e{ALPHA_LOG_MAX}]")
print(f"  Alpha selection : fixed holdout ({N_HOLDOUT_QUARTERS} quarters)")
print(f"  Solver          : max_iter={LASSO_MAX_ITER}, tol={LASSO_TOL}\n")

pred_dict          = {lag: [] for lag in LAGS}
dates_out          = []
actuals_out        = []
lambdas_out        = []
sparsity_out       = []
feature_importance_by_window = []   # list of (date, sparsity, df) for later selection
train_cutoff_last  = None

# Rolling selection frequency: for each variable, count how often it was
# selected (coef != 0) across all windows that reached the Lasso refit.
# Also accumulate absolute coefficients to report average magnitude.
selection_counts = {}        # feature_name -> int
coef_abs_sums    = {}        # feature_name -> float (cumulative |coef|)
n_windows_counted = 0

for nowcast_date in test_dates:

    # --- Actual GDP value ----------------------------------------------------
    nowcast_dt  = pd.to_datetime(nowcast_date)
    actual_rows = data.loc[
        (data["date"] == nowcast_dt) & (~pd.isna(data[TARGET_VARIABLE])),
        TARGET_VARIABLE
    ]
    if actual_rows.empty:
        print(f"  [{nowcast_date}] No actual GDP - skipping.")
        for lag in LAGS:
            pred_dict[lag].append(np.nan)
        dates_out.append(nowcast_date)
        actuals_out.append(np.nan)
        lambdas_out.append(np.nan)
        sparsity_out.append(np.nan)
        continue

    dates_out.append(nowcast_date)
    actuals_out.append(float(actual_rows.values[0]))

    # --- Training window -----------------------------------------------------
    train_cutoff      = (nowcast_dt - pd.DateOffset(months=3)).strftime("%Y-%m-%d")
    train_cutoff_last = train_cutoff

    train_raw = data.loc[
        (data["date"] >= TRAIN_START) & (data["date"] <= train_cutoff)
    ].copy()

    n_q = train_raw[TARGET_VARIABLE].notna().sum()

    # --- Build feature matrix ------------------------------------------------
    # Mean-fill with training means (standard in nowcasting literature to
    # handle ragged edges). Then flatten to quarterly rows (months 3/6/9/12).
    train_filled = mean_fill_dataset(training=train_raw, test=train_raw)
    train_flat   = flatten_data(train_filled, TARGET_VARIABLE, N_LAGS_FEATURES)
    train_flat   = (
        train_flat
        .loc[train_flat["date"].dt.month.isin([3, 6, 9, 12])]
        .dropna(axis=0, how="any")
        .reset_index(drop=True)
    )

    if len(train_flat) < N_HOLDOUT_QUARTERS + 5:
        print(f"  [{nowcast_date}] Only {len(train_flat)} rows - skipping.")
        for lag in LAGS:
            pred_dict[lag].append(np.nan)
        lambdas_out.append(np.nan)
        sparsity_out.append(np.nan)
        continue

    X_train = train_flat.drop(columns=["date", TARGET_VARIABLE])
    y_train = train_flat[TARGET_VARIABLE]

    # --- Alpha selection via fixed holdout -----------------------------------
    # Scaler fitted on inner-train only — holdout never influences scaling.
    X_inner = X_train.iloc[:-N_HOLDOUT_QUARTERS]
    y_inner = y_train.iloc[:-N_HOLDOUT_QUARTERS]
    X_hold  = X_train.iloc[-N_HOLDOUT_QUARTERS:]
    y_hold  = y_train.iloc[-N_HOLDOUT_QUARTERS:]

    scaler_inner = StandardScaler().fit(X_inner)
    X_inner_s    = scaler_inner.transform(X_inner)
    X_hold_s     = scaler_inner.transform(X_hold)

    # Scaler for final model: fit on full training window (all pre-cutoff).
    scaler  = StandardScaler().fit(X_train)
    X_all_s = scaler.transform(X_train)

    # Search alpha grid
    best_alpha, best_rmse = None, np.inf
    for alpha_candidate in ALPHA_GRID:
        lasso_cv = Lasso(
            alpha=alpha_candidate,
            fit_intercept=True,
            max_iter=LASSO_MAX_ITER,
            tol=LASSO_TOL,
            random_state=RANDOM_SEED,
        )
        lasso_cv.fit(X_inner_s, y_inner)
        val_rmse = rmse(y_hold, lasso_cv.predict(X_hold_s))
        if val_rmse < best_rmse:
            best_rmse  = val_rmse
            best_alpha = alpha_candidate

    lambdas_out.append(best_alpha)

    # --- Refit Lasso on full training with best alpha ------------------------
    final_lasso = Lasso(
        alpha=best_alpha,
        fit_intercept=True,
        max_iter=LASSO_MAX_ITER,
        tol=LASSO_TOL,
        random_state=RANDOM_SEED,
    )
    final_lasso.fit(X_all_s, y_train)

    # Sparsity = fraction of zero coefficients
    n_zero   = int(np.sum(final_lasso.coef_ == 0))
    n_total  = len(final_lasso.coef_)
    sparsity = round(100.0 * n_zero / n_total, 1)
    sparsity_out.append(sparsity)

    # Accumulate rolling selection frequency.
    # For each feature: +1 if coef != 0 in this window, plus |coef|.
    n_windows_counted += 1
    for feat, coef_val in zip(X_train.columns, final_lasso.coef_):
        selection_counts[feat] = selection_counts.get(feat, 0) + int(coef_val != 0)
        coef_abs_sums[feat]    = coef_abs_sums.get(feat, 0.0) + abs(float(coef_val))

    # --- Diagnostics print ---------------------------------------------------
    min_warn = "HIT_MIN" if best_alpha <= ALPHA_GRID[1]  else "ok"
    max_warn = "HIT_MAX" if best_alpha >= ALPHA_GRID[-2] else "ok"
    n_nonzero = n_total - n_zero
    print(f"  [{nowcast_date}]  alpha={best_alpha:.2e}  n_train={n_q}  "
          f"features={n_total}  nonzero={n_nonzero}  sparsity={sparsity}%  "
          f"grid_min={min_warn}  grid_max={max_warn}")

    # --- Predict for each vintage -------------------------------------------
    for lag in LAGS:
        vintage_data = gen_lagged_data(metadata, data, nowcast_date, lag)
        vintage_data = mean_fill_dataset(training=train_raw, test=vintage_data)
        vintage_flat = flatten_data(vintage_data, TARGET_VARIABLE, N_LAGS_FEATURES)

        target_row = vintage_flat.loc[vintage_flat["date"] == nowcast_dt]
        if target_row.empty:
            pred_dict[lag].append(np.nan)
            continue

        X_test = (
            target_row
            .drop(columns=["date", TARGET_VARIABLE], errors="ignore")
            .reindex(columns=X_train.columns, fill_value=0.0)
        )
        X_test_s = scaler.transform(X_test)

        pred_dict[lag].append(float(final_lasso.predict(X_test_s)[0]))

    feature_importance_by_window.append((
        nowcast_date,
        sparsity,
        pd.DataFrame({
            "feature":     X_train.columns,
            "coefficient": final_lasso.coef_,
        }).assign(abs_coef=lambda df: df["coefficient"].abs())
    ))

# =============================================================================
# 4. PERFORMANCE METRICS
# =============================================================================
# Period labels come from nowcast_utils (single source of truth)
period_labels = classify_crisis_periods(dates_out).tolist()

# --- Full test period --------------------------------------------------------
full_rows = []
for lag in LAGS:
    full_rows.append({
        "Vintage": lag,
        "RMSE":    round(rmse(actuals_out, pred_dict[lag]), 6),
        "MAE":     round(mae(actuals_out,  pred_dict[lag]), 6),
        "n":       sum(1 for v in pred_dict[lag] if not np.isnan(v)),
    })
perf_full = pd.DataFrame(full_rows)

# --- Stratified: Crisis / Normal --------------------------------------------
strat_rows = []
for period, label in [
    ("crisis", "Crisis (2020Q1-2021Q2)"),
    ("normal", "Normal (2021Q3-2024Q2)"),
]:
    idx = [i for i, p in enumerate(period_labels) if p == period]
    y   = [actuals_out[i] for i in idx]
    for lag in LAGS:
        yhat = [pred_dict[lag][i] for i in idx]
        strat_rows.append({
            "Period":  label,
            "Vintage": lag,
            "n":       len([v for v in yhat if not np.isnan(v)]),
            "RMSE":    round(rmse(y, yhat), 6),
            "MAE":     round(mae(y,  yhat), 6),
        })
perf_strat = pd.DataFrame(strat_rows)

# --- Print results -----------------------------------------------------------
print("\n" + "=" * 64)
print("FULL TEST PERIOD — Lasso")
print("=" * 64)
print(perf_full.to_string(index=False))

print("\n" + "=" * 64)
print("STRATIFIED — Crisis vs Normal")
print("=" * 64)
print(perf_strat.to_string(index=False))

# --- Regularisation benefit summary -----------------------------------------
nowcast_lasso = perf_full.loc[perf_full["Vintage"] == 0, "RMSE"].values[0]
avg_sparsity  = np.nanmean(sparsity_out)

print(f"\nRegularisation summary (nowcast, lag=0):")
print(f"  Lasso RMSE   : {nowcast_lasso:.6f}")
print(f"  Avg sparsity : {avg_sparsity:.1f}% of features zeroed out per window")
print(f"  Benchmark    : compare against ARMA (see arma_benchmark.csv)")

# =============================================================================
# 5. SAVE CSVs
# =============================================================================
pred_table = pd.DataFrame({
    "date":           dates_out,
    "actual":         actuals_out,
    **{f"lag_{'-' if lag < 0 else '+'}{abs(lag)}": pred_dict[lag] for lag in LAGS},
    "period_type":    period_labels,
    "selected_alpha": lambdas_out,
    "sparsity_pct":   sparsity_out,
})

pred_table.to_csv(OUT_DIR  / "lasso_predictions.csv",            index=False)
perf_full.to_csv(OUT_DIR   / "lasso_performance_full.csv",       index=False)
perf_strat.to_csv(OUT_DIR  / "lasso_performance_stratified.csv", index=False)

# --- Feature importance: pick a representative (non-trivial) window ---------
# Prefer last pre-crisis window; if none, fallback to window with lowest
# sparsity (most informative coefficient vector).
crisis_start_dt = pd.to_datetime(CRISIS_START)
pre_crisis = [
    (d, s, df) for (d, s, df) in feature_importance_by_window
    if pd.to_datetime(d) < crisis_start_dt
]

if pre_crisis:
    chosen_date, chosen_sparsity, chosen_df = pre_crisis[-1]
    feature_importance_source = f"last pre-crisis window ({chosen_date})"
elif feature_importance_by_window:
    chosen_date, chosen_sparsity, chosen_df = min(
        feature_importance_by_window, key=lambda x: x[1]
    )
    feature_importance_source = f"lowest-sparsity window ({chosen_date}, sparsity={chosen_sparsity}%)"
else:
    chosen_df = None
    feature_importance_source = "no window available"

if chosen_df is not None:
    (chosen_df
     .sort_values("abs_coef", ascending=False)
     .to_csv(OUT_DIR / "lasso_feature_importance.csv", index=False))
    print(f"\nFeature importance from: {feature_importance_source}")

# --- Rolling selection frequency --------------------------------------------
# For each feature, report:
#   freq_pct     : share of windows in which the feature was selected
#   avg_abs_coef : average |coefficient| across ALL windows (including zeros)
# Useful for interpretability (notebook 02): variables with high freq_pct are
# consistently informative; variables with high freq_pct AND high avg_abs_coef
# are both consistently selected and influential when selected.
if n_windows_counted > 0:
    freq_rows = []
    for feat in selection_counts:
        freq_rows.append({
            "feature":      feat,
            "n_selected":   selection_counts[feat],
            "n_windows":    n_windows_counted,
            "freq_pct":     round(100.0 * selection_counts[feat] / n_windows_counted, 2),
            "avg_abs_coef": round(coef_abs_sums[feat]   / n_windows_counted, 8),
        })
    freq_df = pd.DataFrame(freq_rows).sort_values(
        ["freq_pct", "avg_abs_coef"], ascending=[False, False]
    ).reset_index(drop=True)
    freq_df.to_csv(OUT_DIR / "lasso_selection_frequency.csv", index=False)
    print(f"Rolling selection frequency across {n_windows_counted} windows saved: "
          f"lasso_selection_frequency.csv")

print(f"\nCSVs saved to: {OUT_DIR}")

# =============================================================================
# 6. PLOTTING STYLE (matches ARMA for visual consistency)
# =============================================================================
STYLE = {
    "c_crisis":         "#B6213B",
    "c_normal":         "#2A8F93",
    "c_actual":         "#2E2E2E",
    "zero_line_colour": "#555555",
    "shade_crisis":     0.11,
    "shade_normal":     0.08,
    "c_sparsity":       "#1D3557",   # deep navy — separate from regime colours
}
plt.rcParams.update({
    "font.family":       "serif",
    "axes.titlesize":    13,
    "axes.labelsize":    10.5,
    "xtick.labelsize":   9,
    "ytick.labelsize":   9,
    "legend.fontsize":   9,
    "axes.grid":         True,
    "grid.alpha":        0.20,
    "grid.color":        "#888888",
    "grid.linewidth":    0.5,
    "axes.edgecolor":    "#333333",
    "axes.linewidth":    0.7,
    "axes.spines.top":   False,
    "axes.spines.right": False,
    "figure.dpi":        120,
})

# Vintage styling — only 3 vintages now
LAG_COLORS = {-2: "#a8dadc", -1: "#457b9d", 0: "#1d3557"}
LAG_LW     = {-2: 1.2, -1: 1.2, 0: 2.4}
LAG_LS     = {-2: "--", -1: "--", 0: "-"}
LAG_LABELS = {-2: "lag=-2 (forecast)",
              -1: "lag=-1 (early nowcast)",
               0: "lag= 0 (nowcast)"}


def _format_xaxis(ax):
    ax.xaxis.set_major_locator(mdates.YearLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))


def _shade_regimes(ax):
    ax.axvspan(pd.to_datetime(CRISIS_START), pd.to_datetime(CRISIS_END),
               alpha=STYLE["shade_crisis"], color=STYLE["c_crisis"],
               edgecolor="none", zorder=0)
    ax.axvspan(pd.to_datetime(NORMAL_START), pd.to_datetime(NORMAL_END),
               alpha=STYLE["shade_normal"], color=STYLE["c_normal"],
               edgecolor="none", zorder=0)


full_series = (
    data[["date", TARGET_VARIABLE]]
    .dropna(subset=[TARGET_VARIABLE])
    .sort_values("date")
    .reset_index(drop=True)
)

# =============================================================================
# 7. PLOT A — Crisis period (2020Q1–2021Q2)
# =============================================================================
crisis_rows = pred_table[pred_table["period_type"] == "crisis"].copy()
crisis_rows["date"] = pd.to_datetime(crisis_rows["date"])

fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(full_series["date"], full_series[TARGET_VARIABLE],
        color="#444444", lw=1.3, marker="o", ms=3, alpha=0.65,
        label="Actual GDP (2005–2024)", zorder=3)
ax.axvspan(pd.to_datetime(CRISIS_START), pd.to_datetime(CRISIS_END),
           alpha=0.13, color=STYLE["c_crisis"], zorder=0,
           label="Crisis window (2020Q1–2021Q2)")
ax.axhline(0, color=STYLE["zero_line_colour"], lw=0.8, alpha=0.4, zorder=1)

for lag in LAGS:
    col = f"lag_{'-' if lag < 0 else '+'}{abs(lag)}"
    ax.plot(crisis_rows["date"], crisis_rows[col],
            color=LAG_COLORS[lag], lw=LAG_LW[lag], ls=LAG_LS[lag],
            marker="o", ms=4.5, label=LAG_LABELS[lag], zorder=5)

ax.set_xlabel("Date")
ax.set_ylabel("GDP growth rate (QoQ)")
ax.set_title("Lasso Regression — Crisis Period (2020Q1–2021Q2)\n"
             "Nowcasting Vintages | El Salvador",
             fontweight="bold")
ax.legend(fontsize=8, ncol=3, loc="lower left")
_format_xaxis(ax)
plt.tight_layout()
fig.savefig(OUT_DIR / "lasso_plot_crisis.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: lasso_plot_crisis.png")

# =============================================================================
# 8. PLOT B — Normal period (2021Q3–2024Q2)
# =============================================================================
normal_rows = pred_table[pred_table["period_type"] == "normal"].copy()
normal_rows["date"] = pd.to_datetime(normal_rows["date"])

fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(full_series["date"], full_series[TARGET_VARIABLE],
        color="#444444", lw=1.3, marker="o", ms=3, alpha=0.65,
        label="Actual GDP (2005–2024)", zorder=3)
ax.axvspan(pd.to_datetime(NORMAL_START), pd.to_datetime(NORMAL_END),
           alpha=0.12, color=STYLE["c_normal"], zorder=0,
           label="Normal window (2021Q3–2024Q2)")
ax.axvspan(pd.to_datetime(CRISIS_START), pd.to_datetime(CRISIS_END),
           alpha=0.07, color=STYLE["c_crisis"], zorder=0,
           label="Crisis period (reference)")
ax.axhline(0, color=STYLE["zero_line_colour"], lw=0.8, alpha=0.4, zorder=1)

for lag in LAGS:
    col = f"lag_{'-' if lag < 0 else '+'}{abs(lag)}"
    ax.plot(normal_rows["date"], normal_rows[col],
            color=LAG_COLORS[lag], lw=LAG_LW[lag], ls=LAG_LS[lag],
            marker="o", ms=4.5, label=LAG_LABELS[lag], zorder=5)

ax.set_xlabel("Date")
ax.set_ylabel("GDP growth rate (QoQ)")
ax.set_title("Lasso Regression — Normal Period (2021Q3–2024Q2)\n"
             "Nowcasting Vintages | El Salvador",
             fontweight="bold")
ax.legend(fontsize=8, ncol=3, loc="lower left")
_format_xaxis(ax)
plt.tight_layout()
fig.savefig(OUT_DIR / "lasso_plot_normal.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: lasso_plot_normal.png")

# =============================================================================
# 9. PLOT C — RMSE heatmap
# =============================================================================
pivot = perf_strat.pivot(index="Period", columns="Vintage", values="RMSE")
fig, ax = plt.subplots(figsize=(9, 3))
sns.heatmap(pivot, annot=True, fmt=".4f", cmap="YlOrRd",
            linewidths=0.6, ax=ax, cbar_kws={"label": "RMSE", "shrink": 0.8})
ax.set_title("Lasso Regression — RMSE by Period and Vintage",
             fontweight="bold", pad=10)
ax.set_xlabel("Vintage (months relative to quarter-end)")
ax.set_ylabel("")
plt.tight_layout()
fig.savefig(OUT_DIR / "lasso_plot_heatmap.png", dpi=150, bbox_inches="tight")
plt.show()
print("Saved: lasso_plot_heatmap.png")

# =============================================================================
# 10. PLOT D — Feature importance (non-zero coefficients)
# =============================================================================
if chosen_df is not None:
    nonzero_feats = chosen_df[chosen_df["coefficient"] != 0] \
        .sort_values("abs_coef", ascending=False)

    if len(nonzero_feats) == 0:
        print("All coefficients are zero in chosen window - alpha may be too large.")
    else:
        top_n     = min(15, len(nonzero_feats))
        top_feats = nonzero_feats.head(top_n)
        bar_cols  = ["#2a9d8f" if c >= 0 else "#e76f51"
                     for c in top_feats["coefficient"]]
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.barh(top_feats["feature"][::-1],
                top_feats["coefficient"][::-1],
                color=bar_cols[::-1], edgecolor="gray", linewidth=0.4)
        ax.axvline(0, color="black", lw=0.8)
        ax.set_xlabel("Lasso coefficient (standardised features)")
        ax.set_title(f"Top {top_n} Non-Zero Features — Lasso\n"
                     f"Source: {feature_importance_source}",
                     fontweight="bold")
        plt.tight_layout()
        fig.savefig(OUT_DIR / "lasso_plot_features.png",
                    dpi=150, bbox_inches="tight")
        plt.show()
        print("Saved: lasso_plot_features.png")
        print(f"  Non-zero features in chosen window: {len(nonzero_feats)} / {len(chosen_df)}")

# =============================================================================
# 11. PLOT E — Alpha path + sparsity over time
# =============================================================================
valid = [(d, a, s) for d, a, s in zip(dates_out, lambdas_out, sparsity_out)
         if a is not None and not np.isnan(a)]

if valid:
    v_dates    = pd.to_datetime([x[0] for x in valid])
    v_alphas   = [x[1] for x in valid]
    v_sparsity = [x[2] for x in valid]

    fig, axes = plt.subplots(2, 1, figsize=(13, 8))
    fig.suptitle("Lasso Regression — Regularisation Path and Sparsity",
                 fontweight="bold")

    # Panel A: alpha
    ax = axes[0]
    _shade_regimes(ax)
    ax.plot(v_dates, v_alphas, color="#1d3557", lw=1.8,
            marker="o", ms=5, label="Selected alpha")
    ax.axhline(ALPHA_GRID[0],  color="#e63946", lw=0.8, ls="--",
               label=f"Grid min = {ALPHA_GRID[0]:.0e}")
    ax.axhline(ALPHA_GRID[-1], color="#f4a261", lw=0.8, ls="--",
               label=f"Grid max = {ALPHA_GRID[-1]:.0e}")
    ax.set_yscale("log")
    ax.set_ylabel("Selected alpha (log)")
    ax.set_title(f"CV-selected alpha over time (holdout = last {N_HOLDOUT_QUARTERS} quarters)",
                 fontsize=10)
    ax.legend(fontsize=8)
    _format_xaxis(ax)

    # Panel B: sparsity (deep navy — distinct from regime shading)
    ax = axes[1]
    _shade_regimes(ax)
    ax.fill_between(v_dates, v_sparsity, alpha=0.25, color=STYLE["c_sparsity"])
    ax.plot(v_dates, v_sparsity, color=STYLE["c_sparsity"], lw=1.8,
            marker="o", ms=5, label="Sparsity %")
    ax.set_ylabel("Features zeroed out (%)")
    ax.set_ylim(0, 105)
    ax.set_title("Sparsity over time — higher % = fewer features retained",
                 fontsize=10)
    ax.legend(fontsize=8)
    _format_xaxis(ax)

    plt.tight_layout()
    fig.savefig(OUT_DIR / "lasso_plot_alpha_sparsity.png",
                dpi=150, bbox_inches="tight")
    plt.show()
    print("Saved: lasso_plot_alpha_sparsity.png")

# =============================================================================
# DONE
# =============================================================================
print("\n" + "=" * 64)
print("LASSO NOWCASTING COMPLETE")
print("=" * 64)
print(f"Output : {OUT_DIR}")
```
