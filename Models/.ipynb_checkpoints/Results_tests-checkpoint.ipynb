```python
# =============================================================================
# master_evaluation.py
# Purpose : Single script that produces ALL evaluation outputs for the paper:
#           A. Coverage analysis (68%, 80%, 95%) — full, crisis, normal
#           B. Integrated master table (RMSE, MAE, rel. RMSE, CRPS, coverage,
#              interval width, DM stat, ranking)
#           C. Model decision framework (best point / density / calibration)
#           D. Vintage gain table (lag-2 → lag-1 → lag-0 improvement)
#           E. Clean DM significance table
#           F. Interval width table by period
#           G. Consistency table (point vs density vs calibration)
#           H. Naive random-walk benchmark (robustness check)
#           I. CRPS by vintage (extends density evaluation to all 3 lags)
#
# Input  : unified_predictions.csv   (from comparison_final.py)
#          dm_all_results.csv         (from diebold_mariano_final.py)
#          coverage_table.csv         (from fanchart_final.py)
#          crps_scores.csv            (from crps_final.py)
#          comparison_raw.csv         (from comparison_final.py)
#
# Output : ~/Desktop/ESA-gdp-nowcasting/Models/results/master_evaluation/
#           master_table.csv            — full integrated table
#           coverage_analysis.csv       — coverage by model/period/level
#           vintage_gain_table.csv      — lag improvement analysis
#           interval_width_table.csv    — avg widths by model/period
#           dm_clean_table.csv          — clean DM table for paper
#           crps_by_vintage.csv         — CRPS at all 3 lags
#           decision_framework.csv      — model classification
#           consistency_table.csv       — point vs density vs calibration
#           robustness_naive.csv        — naive benchmark comparison
#           + one publication-ready figure per analysis section
#
# Author : Gabriela Aquino
# =============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import warnings
from pathlib import Path
from scipy import stats

warnings.filterwarnings("ignore")
np.random.seed(42)

# =============================================================================
# 0. PATHS AND CONFIGURATION
# =============================================================================

BASE_DIR  = Path.home() / "Desktop" / "ESA-gdp-nowcasting"
RES_DIR   = BASE_DIR / "Models" / "results"
OUT_DIR   = RES_DIR / "master_evaluation"
OUT_DIR.mkdir(parents=True, exist_ok=True)

# Input files
UNIFIED_FILE  = RES_DIR / "comparison"       / "unified_predictions.csv"
DM_FILE       = RES_DIR / "diebold_mariano"  / "dm_all_results.csv"
COVERAGE_FILE = RES_DIR / "fanchart"         / "coverage_table.csv"
CRPS_FILE     = RES_DIR / "crps"             / "crps_scores.csv"
RAW_FILE      = RES_DIR / "comparison"       / "comparison_raw.csv"

# Period definitions — must match all other scripts
CRISIS_START = "2019-01-01"
CRISIS_END   = "2021-06-01"
NORMAL_START = "2022-01-01"
NORMAL_END   = "2024-06-01"

# Models in preferred display order
MODELS_ORDER = ["DFM", "Ridge", "MIDAS", "Lasso", "Random Forest", "XGBoost", "ARMA"]

# Nowcast vintage (operationally most relevant)
LAG_NOWCAST = 0
LAGS        = [-2, -1, 0]

# Bootstrap settings (for interval computation within this script)
N_BOOT      = 2000
CONF_LEVELS = [0.68, 0.80, 0.95]

# Publication figure style
plt.rcParams.update({
    "font.family":      "sans-serif",
    "font.size":        10,
    "axes.spines.top":  False,
    "axes.spines.right":False,
    "axes.grid":        True,
    "grid.alpha":       0.25,
    "grid.linewidth":   0.5,
    "figure.dpi":       150,
})

MODEL_COLORS = {
    "DFM":           "#1D9E75",
    "Ridge":         "#185FA5",
    "MIDAS":         "#BA7517",
    "Lasso":         "#D4537E",
    "Random Forest": "#888780",
    "XGBoost":       "#D85A30",
    "ARMA":          "#5F5E5A",
}

print("=" * 70)
print("MASTER EVALUATION — El Salvador GDP Nowcasting")
print("=" * 70)

# =============================================================================
# 1. LOAD ALL DATA
# =============================================================================

def check_file(path):
    if not path.exists():
        raise FileNotFoundError(
            f"\nFile not found: {path}\n"
            f"Make sure the corresponding script has been run first.\n"
        )

check_file(UNIFIED_FILE)
check_file(RAW_FILE)

print("\nLoading data...")
df_unified = pd.read_csv(UNIFIED_FILE, parse_dates=["date"]).sort_values("date").reset_index(drop=True)
df_raw     = pd.read_csv(RAW_FILE)

# Load optional files (some analyses skip gracefully if missing)
df_dm       = pd.read_csv(DM_FILE)       if DM_FILE.exists()       else None
df_coverage = pd.read_csv(COVERAGE_FILE) if COVERAGE_FILE.exists() else None
df_crps     = pd.read_csv(CRPS_FILE)     if CRPS_FILE.exists()     else None

print(f"  Unified predictions : {len(df_unified)} rows x {len(df_unified.columns)} cols")
print(f"  DM results          : {'loaded' if df_dm       is not None else 'not found — will skip DM section'}")
print(f"  Coverage table      : {'loaded' if df_coverage is not None else 'not found — will recompute'}")
print(f"  CRPS scores         : {'loaded' if df_crps     is not None else 'not found — will recompute'}")

actuals  = df_unified["actual"].values
dates_dt = pd.to_datetime(df_unified["date"])

# Period masks
mask_full   = np.ones(len(actuals), dtype=bool)
mask_crisis = (dates_dt >= CRISIS_START) & (dates_dt <= CRISIS_END)
mask_normal = (dates_dt >= NORMAL_START) & (dates_dt <= NORMAL_END)
MASKS = {"Full": mask_full, "Crisis": mask_crisis, "Normal": mask_normal}

# Load predictions for all models x lags
all_preds = {}  # {model: {lag: array}}
for model in MODELS_ORDER:
    all_preds[model] = {}
    for lag in LAGS:
        col = f"{model}_lag{lag}"
        if col in df_unified.columns:
            all_preds[model][lag] = df_unified[col].values
        else:
            all_preds[model][lag] = np.full(len(actuals), np.nan)

loaded_models = [m for m in MODELS_ORDER if not np.all(np.isnan(all_preds[m][0]))]
print(f"\nModels with predictions at lag=0: {loaded_models}")

# =============================================================================
# 2. METRIC HELPER FUNCTIONS
# =============================================================================

def rmse(y, yhat):
    mask = ~np.isnan(y) & ~np.isnan(yhat)
    if mask.sum() < 2:
        return np.nan
    return float(np.sqrt(np.mean((y[mask] - yhat[mask]) ** 2)))

def mae(y, yhat):
    mask = ~np.isnan(y) & ~np.isnan(yhat)
    if mask.sum() < 2:
        return np.nan
    return float(np.mean(np.abs(y[mask] - yhat[mask])))

def gaussian_crps(y, mu, sigma):
    """Closed-form Gaussian CRPS. Gneiting & Raftery (2007), eq. 21."""
    if np.isnan(y) or np.isnan(mu) or np.isnan(sigma) or sigma <= 0:
        return np.nan
    z = (y - mu) / sigma
    return float(sigma * (z * (2 * stats.norm.cdf(z) - 1)
                          + 2 * stats.norm.pdf(z)
                          - 1 / np.sqrt(np.pi)))

def rolling_sigma(actuals, preds, t, min_obs=5):
    """Estimate sigma from rolling-origin residuals up to time t."""
    errors = (actuals - preds)[:t]
    errors = errors[~np.isnan(errors)]
    if len(errors) < min_obs:
        return np.nan
    return float(np.std(errors, ddof=1))

def bootstrap_intervals(actuals, preds, levels=CONF_LEVELS, n_boot=N_BOOT, seed=42):
    """Non-parametric bootstrap prediction intervals (pseudo-real-time)."""
    rng = np.random.default_rng(seed)
    T   = len(preds)
    errors = actuals - preds

    intervals = {lev: {"lower": np.full(T, np.nan),
                        "upper": np.full(T, np.nan),
                        "width": np.full(T, np.nan)} for lev in levels}

    for t in range(T):
        yhat = preds[t]
        if np.isnan(yhat):
            continue
        past = errors[:t]
        past = past[~np.isnan(past)]
        if len(past) < 5:
            # Fallback: Gaussian with past RMSE
            if len(past) > 0:
                s = np.sqrt(np.mean(past ** 2))
                for lev in levels:
                    z = {0.68: 1.00, 0.80: 1.28, 0.95: 1.96}.get(lev, 1.96)
                    intervals[lev]["lower"][t] = yhat - z * s
                    intervals[lev]["upper"][t] = yhat + z * s
                    intervals[lev]["width"][t] = 2 * z * s
            continue
        boot = rng.choice(past, size=n_boot, replace=True)
        boot_fc = yhat + boot
        for lev in levels:
            alpha = 1 - lev
            lo = float(np.quantile(boot_fc, alpha / 2))
            hi = float(np.quantile(boot_fc, 1 - alpha / 2))
            intervals[lev]["lower"][t] = lo
            intervals[lev]["upper"][t] = hi
            intervals[lev]["width"][t] = hi - lo

    return intervals

def empirical_coverage(actuals, intervals, mask, lev):
    """Fraction of actuals inside the interval at the given confidence level."""
    lo = intervals[lev]["lower"][mask]
    hi = intervals[lev]["upper"][mask]
    y  = actuals[mask]
    valid = ~np.isnan(y) & ~np.isnan(lo) & ~np.isnan(hi)
    if valid.sum() == 0:
        return np.nan, 0
    cov = ((y[valid] >= lo[valid]) & (y[valid] <= hi[valid])).mean()
    return float(cov), int(valid.sum())

# =============================================================================
# 3. COMPUTE BOOTSTRAP INTERVALS FOR ALL MODELS (recompute for full analysis)
# =============================================================================

print("\nComputing bootstrap intervals for all models...")
all_ivals = {}
for model in loaded_models:
    yhat = all_preds[model][LAG_NOWCAST]
    all_ivals[model] = bootstrap_intervals(actuals, yhat)
    print(f"  [{model}] done")

# =============================================================================
# H. NAIVE RANDOM-WALK BENCHMARK (additional robustness check)
# =============================================================================
# RW forecast: yhat(t) = actual(t-1)
# Historical mean: yhat(t) = mean of actuals up to t-1

print("\nBuilding naive benchmarks...")
T = len(actuals)
rw_preds   = np.full(T, np.nan)  # random walk
hm_preds   = np.full(T, np.nan)  # expanding historical mean

for t in range(1, T):
    if not np.isnan(actuals[t - 1]):
        rw_preds[t] = actuals[t - 1]
    past = actuals[:t]
    past = past[~np.isnan(past)]
    if len(past) > 0:
        hm_preds[t] = past.mean()

all_preds["RandomWalk"]  = {lag: rw_preds for lag in LAGS}
all_preds["HistMean"]    = {lag: hm_preds for lag in LAGS}
loaded_models_ext        = loaded_models + ["RandomWalk", "HistMean"]

# =============================================================================
# A. COVERAGE ANALYSIS — SECTION A
# =============================================================================
# Empirical coverage at 68%, 80%, 95% for full / crisis / normal periods.
# Coverage deviation = |empirical - nominal| in percentage points.

print("\n" + "=" * 70)
print("A. COVERAGE ANALYSIS")
print("=" * 70)

cov_rows = []
for model in loaded_models:
    for period_label, mask in MASKS.items():
        for lev in CONF_LEVELS:
            cov, n = empirical_coverage(actuals, all_ivals[model], mask, lev)
            nom    = lev * 100
            dev    = abs(cov * 100 - nom) if not np.isnan(cov) else np.nan

            # Average interval width for this period and level
            w_arr  = all_ivals[model][lev]["width"][mask]
            w_arr  = w_arr[~np.isnan(w_arr)]
            avg_w  = float(w_arr.mean()) if len(w_arr) > 0 else np.nan

            cov_rows.append({
                "Model":              model,
                "Period":             period_label,
                "Nominal level":      f"{int(nom)}%",
                "Nominal (frac)":     lev,
                "Empirical coverage": round(cov, 4) if not np.isnan(cov) else np.nan,
                "Empirical (pct)":    round(cov * 100, 1) if not np.isnan(cov) else np.nan,
                "Coverage deviation (pp)": round(dev, 1) if not np.isnan(dev) else np.nan,
                "Avg interval width": round(avg_w, 5) if not np.isnan(avg_w) else np.nan,
                "n":                  n,
            })

cov_df = pd.DataFrame(cov_rows)
cov_df.to_csv(OUT_DIR / "coverage_analysis.csv", index=False)

# Print clean table — 95% level only
print("\nEmpirical coverage at 95% nominal level:")
print("-" * 60)
sub95 = cov_df[cov_df["Nominal level"] == "95%"].copy()
pivot_cov = sub95.pivot(index="Model", columns="Period",
                         values="Empirical (pct)")[["Full", "Crisis", "Normal"]]
pivot_dev = sub95.pivot(index="Model", columns="Period",
                         values="Coverage deviation (pp)")[["Full", "Crisis", "Normal"]]

print(f"{'Model':<16} {'Full (T=22)':>12} {'Crisis (T=10)':>14} {'Normal (T=10)':>14}")
print("-" * 60)
for model in loaded_models:
    row_cov = pivot_cov.loc[model] if model in pivot_cov.index else {}
    row_dev = pivot_dev.loc[model] if model in pivot_dev.index else {}
    f  = f"{row_cov.get('Full', np.nan):.1f}%" if not pd.isna(row_cov.get('Full', np.nan)) else "—"
    cr = f"{row_cov.get('Crisis', np.nan):.1f}%" if not pd.isna(row_cov.get('Crisis', np.nan)) else "—"
    nm = f"{row_cov.get('Normal', np.nan):.1f}%" if not pd.isna(row_cov.get('Normal', np.nan)) else "—"
    print(f"{model:<16} {f:>12} {cr:>14} {nm:>14}")

print("\nCoverage deviation from nominal (|empirical − 95%|) in pp:")
print("-" * 60)
for model in loaded_models:
    row = pivot_dev.loc[model] if model in pivot_dev.index else {}
    f  = f"{row.get('Full', np.nan):.1f}pp"   if not pd.isna(row.get('Full', np.nan))   else "—"
    cr = f"{row.get('Crisis', np.nan):.1f}pp"  if not pd.isna(row.get('Crisis', np.nan))  else "—"
    nm = f"{row.get('Normal', np.nan):.1f}pp"  if not pd.isna(row.get('Normal', np.nan))  else "—"
    print(f"{model:<16} {f:>12} {cr:>14} {nm:>14}")

print("\nSaved: coverage_analysis.csv")

# =============================================================================
# D. VINTAGE GAIN TABLE
# =============================================================================
# Relative improvement in RMSE from lag=-2 → lag=-1 → lag=0 for each model.
# "Vintage gain" = how much each additional month of monthly data helps.

print("\n" + "=" * 70)
print("D. VINTAGE GAIN TABLE")
print("=" * 70)

vg_rows = []
for model in loaded_models:
    for period_label, mask in MASKS.items():
        y = actuals[mask]
        rmse_vals = {}
        for lag in LAGS:
            yhat = all_preds[model][lag][mask]
            rmse_vals[lag] = rmse(y, yhat)

        # Gains: negative = improvement (lower RMSE)
        gain_m2_m1 = rmse_vals[-1] - rmse_vals[-2]  # lag-2 to lag-1
        gain_m1_0  = rmse_vals[0]  - rmse_vals[-1]  # lag-1 to lag-0
        gain_total = rmse_vals[0]  - rmse_vals[-2]  # lag-2 to lag-0

        # Relative gain (%)
        def rel_pct(base, val):
            if base and not np.isnan(base) and base != 0:
                return round((val / base) * 100, 1)
            return np.nan

        vg_rows.append({
            "Model":                 model,
            "Period":                period_label,
            "RMSE lag=-2":           round(rmse_vals[-2], 5),
            "RMSE lag=-1":           round(rmse_vals[-1], 5),
            "RMSE lag=0 (nowcast)":  round(rmse_vals[0],  5),
            "Gain -2→-1 (pp RMSE)":  round(gain_m2_m1, 5),
            "Gain -1→0  (pp RMSE)":  round(gain_m1_0,  5),
            "Total gain -2→0 (pp)":  round(gain_total, 5),
            "Total gain (%)":        rel_pct(rmse_vals[-2], gain_total),
        })

vg_df = pd.DataFrame(vg_rows)
vg_df.to_csv(OUT_DIR / "vintage_gain_table.csv", index=False)

# Print summary for crisis period
print("\nVintage gain in RMSE — Crisis period (2019Q1–2021Q2):")
print("Negative = improvement (lower RMSE with more information)")
print("-" * 70)
print(f"{'Model':<16} {'RMSE lag=-2':>12} {'RMSE lag=-1':>12} {'RMSE lag=0':>12} {'Total gain':>12}")
print("-" * 70)
sub_c = vg_df[vg_df["Period"] == "Crisis"]
for _, row in sub_c.iterrows():
    print(f"{row['Model']:<16} {row['RMSE lag=-2']:>12.4f} {row['RMSE lag=-1']:>12.4f} "
          f"{row['RMSE lag=0 (nowcast)']:>12.4f} {row['Total gain (%)']:>11.1f}%")

print("\nVintage gain in RMSE — Normal period (2022Q1–2024Q2):")
print("-" * 70)
print(f"{'Model':<16} {'RMSE lag=-2':>12} {'RMSE lag=-1':>12} {'RMSE lag=0':>12} {'Total gain':>12}")
print("-" * 70)
sub_n = vg_df[vg_df["Period"] == "Normal"]
for _, row in sub_n.iterrows():
    print(f"{row['Model']:<16} {row['RMSE lag=-2']:>12.4f} {row['RMSE lag=-1']:>12.4f} "
          f"{row['RMSE lag=0 (nowcast)']:>12.4f} {row['Total gain (%)']:>11.1f}%")

print("\nSaved: vintage_gain_table.csv")

# =============================================================================
# I. CRPS BY VINTAGE
# =============================================================================
# Extends density evaluation to all 3 lags — shows if density forecast also
# improves as more information arrives within the quarter.

print("\n" + "=" * 70)
print("I. CRPS BY VINTAGE")
print("=" * 70)

crps_vint_rows = []
for model in loaded_models:
    for lag in LAGS:
        yhat = all_preds[model][lag]
        # Estimate sigma from expanding residuals (same method as crps_final.py)
        sigmas = np.array([rolling_sigma(actuals, yhat, t) for t in range(len(actuals))])

        for period_label, mask in MASKS.items():
            y_sub    = actuals[mask]
            mu_sub   = yhat[mask]
            sig_sub  = sigmas[mask]
            crps_arr = np.array([gaussian_crps(y_sub[i], mu_sub[i], sig_sub[i])
                                 for i in range(len(y_sub))])
            valid = ~np.isnan(crps_arr)
            mean_crps = float(crps_arr[valid].mean()) if valid.sum() > 0 else np.nan

            crps_vint_rows.append({
                "Model":  model,
                "Lag":    lag,
                "Period": period_label,
                "CRPS":   round(mean_crps, 6),
                "n":      int(valid.sum()),
            })

crps_vint_df = pd.DataFrame(crps_vint_rows)

# Compute relative CRPS vs ARMA for each lag and period
arma_crps = crps_vint_df[crps_vint_df["Model"] == "ARMA"].set_index(["Lag", "Period"])["CRPS"]
crps_vint_df["rel_CRPS"] = crps_vint_df.apply(
    lambda r: round(r["CRPS"] / arma_crps.get((r["Lag"], r["Period"]), np.nan), 4)
    if not pd.isna(arma_crps.get((r["Lag"], r["Period"]), np.nan)) else np.nan,
    axis=1
)
crps_vint_df.to_csv(OUT_DIR / "crps_by_vintage.csv", index=False)

print("\nRelative CRPS by vintage — Crisis period:")
print("-" * 55)
pivot_crps_c = crps_vint_df[crps_vint_df["Period"] == "Crisis"].pivot(
    index="Model", columns="Lag", values="rel_CRPS"
).reindex(loaded_models)
pivot_crps_c.columns = [f"lag={c}" for c in pivot_crps_c.columns]
print(pivot_crps_c.round(3).to_string())

print("\nRelative CRPS by vintage — Normal period:")
print("-" * 55)
pivot_crps_n = crps_vint_df[crps_vint_df["Period"] == "Normal"].pivot(
    index="Model", columns="Lag", values="rel_CRPS"
).reindex(loaded_models)
pivot_crps_n.columns = [f"lag={c}" for c in pivot_crps_n.columns]
print(pivot_crps_n.round(3).to_string())
print("\nSaved: crps_by_vintage.csv")

# =============================================================================
# F. INTERVAL WIDTH TABLE BY PERIOD
# =============================================================================

print("\n" + "=" * 70)
print("F. INTERVAL WIDTH TABLE")
print("=" * 70)

width_rows = []
for model in loaded_models:
    for period_label, mask in MASKS.items():
        row = {"Model": model, "Period": period_label}
        for lev in CONF_LEVELS:
            w_arr = all_ivals[model][lev]["width"][mask]
            w_arr = w_arr[~np.isnan(w_arr)]
            row[f"Avg width {int(lev*100)}%"] = round(float(w_arr.mean()), 5) if len(w_arr) > 0 else np.nan
        width_rows.append(row)

width_df = pd.DataFrame(width_rows)
width_df.to_csv(OUT_DIR / "interval_width_table.csv", index=False)

print("\nAverage 95% interval width by model and period:")
print("-" * 50)
pw = width_df.pivot(index="Model", columns="Period", values="Avg width 95%")[["Full","Crisis","Normal"]]
print(pw.round(4).to_string())
print("\nSaved: interval_width_table.csv")

# =============================================================================
# E. CLEAN DM TABLE FOR PAPER
# =============================================================================

print("\n" + "=" * 70)
print("E. DIEBOLD-MARIANO CLEAN TABLE")
print("=" * 70)

if df_dm is not None:
    # Extract vs ARMA, lag=0
    vs_arma = df_dm[
        (df_dm["benchmark"] == "ARMA") &
        (df_dm["lag"]       == 0)
    ].copy()

    def stars(p):
        if pd.isna(p):
            return ""
        if p < 0.01: return "***"
        if p < 0.05: return "**"
        if p < 0.10: return "*"
        return ""

    dm_clean_rows = []
    for model in [m for m in loaded_models if m != "ARMA"]:
        for period_label in ["Full", "Crisis", "Normal"]:
            sub = vs_arma[(vs_arma["model"] == model) & (vs_arma["period"] == period_label)]
            if len(sub) == 0:
                dm_clean_rows.append({
                    "Model": model, "Period": period_label,
                    "DM (RMSE)": np.nan, "Stars (RMSE)": "",
                    "p-val (RMSE)": np.nan,
                    "DM (MAE)": np.nan,  "Stars (MAE)": "",
                    "T": np.nan,
                })
                continue
            row = sub.iloc[0]
            dm_clean_rows.append({
                "Model":      model,
                "Period":     period_label,
                "T":          int(row["T"]) if "T" in row else np.nan,
                "DM (RMSE)":  round(row.get("dm_rmse",  np.nan), 3),
                "Stars (RMSE)": stars(row.get("pval_rmse", np.nan)),
                "p-val (RMSE)": round(row.get("pval_rmse", np.nan), 4),
                "DM (MAE)":   round(row.get("dm_mae",   np.nan), 3),
                "Stars (MAE)":  stars(row.get("pval_mae",  np.nan)),
            })

    dm_clean_df = pd.DataFrame(dm_clean_rows)
    dm_clean_df.to_csv(OUT_DIR / "dm_clean_table.csv", index=False)

    # Compact pivot for paper
    dm_pivot = dm_clean_df.copy()
    dm_pivot["DM_RMSE_str"] = dm_pivot.apply(
        lambda r: f"{r['DM (RMSE)']:.3f}{r['Stars (RMSE)']}" if not pd.isna(r["DM (RMSE)"]) else "—",
        axis=1
    )
    dm_pivot_wide = dm_pivot.pivot(index="Model", columns="Period", values="DM_RMSE_str")
    dm_pivot_wide = dm_pivot_wide.reindex(
        [m for m in loaded_models if m != "ARMA"],
        columns=["Full", "Crisis", "Normal"]
    )
    print("\nDM statistics vs ARMA, RMSE loss (negative = model better):")
    print("-" * 60)
    print(f"{'Model':<16} {'Full (T=22)':>14} {'Crisis (T=10)':>14} {'Normal (T=10)':>14}")
    print("-" * 60)
    for model, row in dm_pivot_wide.iterrows():
        print(f"{model:<16} {str(row.get('Full','—')):>14} {str(row.get('Crisis','—')):>14} {str(row.get('Normal','—')):>14}")
    print("\nNote: * p<0.10  ** p<0.05  *** p<0.01  (HLN small-sample correction)")
    print("Saved: dm_clean_table.csv")
else:
    print("  DM file not found — run diebold_mariano_final.py first")
    dm_clean_df = None

# =============================================================================
# H. NAIVE BENCHMARK COMPARISON
# =============================================================================

print("\n" + "=" * 70)
print("H. NAIVE BENCHMARK ROBUSTNESS CHECK")
print("=" * 70)

naive_rows = []
for model in ["ARMA", "RandomWalk", "HistMean"] + [m for m in loaded_models if m != "ARMA"]:
    for period_label, mask in MASKS.items():
        y    = actuals[mask]
        yhat = all_preds[model][LAG_NOWCAST][mask]
        naive_rows.append({
            "Model":  model,
            "Period": period_label,
            "RMSE":   round(rmse(y, yhat), 6),
            "MAE":    round(mae(y,  yhat), 6),
            "n":      int((~np.isnan(y) & ~np.isnan(yhat)).sum()),
        })

naive_df = pd.DataFrame(naive_rows)

# Relative to random walk (alternative denominator)
rw_rmse = naive_df[naive_df["Model"] == "RandomWalk"].set_index("Period")["RMSE"]
naive_df["Rel RMSE vs RW"] = naive_df.apply(
    lambda r: round(r["RMSE"] / rw_rmse.get(r["Period"], np.nan), 4), axis=1
)
naive_df.to_csv(OUT_DIR / "robustness_naive.csv", index=False)

print("\nNaive benchmarks vs models — full period:")
sub_f = naive_df[naive_df["Period"] == "Full"][["Model", "RMSE", "MAE", "Rel RMSE vs RW"]]
print(sub_f.to_string(index=False))
print("\nSaved: robustness_naive.csv")

# =============================================================================
# B. MASTER INTEGRATED TABLE
# =============================================================================
# One row per model x period with all key metrics.

print("\n" + "=" * 70)
print("B. MASTER INTEGRATED TABLE")
print("=" * 70)

# ARMA RMSE at lag=0 for relative computation
arma_rmse_dict = {}
arma_mae_dict  = {}
for period_label, mask in MASKS.items():
    y = actuals[mask]
    if "ARMA" in loaded_models:
        yhat_arma = all_preds["ARMA"][LAG_NOWCAST][mask]
        arma_rmse_dict[period_label] = rmse(y, yhat_arma)
        arma_mae_dict[period_label]  = mae(y,  yhat_arma)
    else:
        arma_rmse_dict[period_label] = np.nan
        arma_mae_dict[period_label]  = np.nan

master_rows = []
for model in loaded_models:
    for period_label, mask in MASKS.items():
        y    = actuals[mask]
        yhat = all_preds[model][LAG_NOWCAST][mask]
        n    = int((~np.isnan(y) & ~np.isnan(yhat)).sum())

        # ---- Point metrics
        rmse_val = rmse(y, yhat)
        mae_val  = mae(y,  yhat)
        rel_rmse = round(rmse_val / arma_rmse_dict[period_label], 4) if arma_rmse_dict[period_label] else np.nan
        rel_mae  = round(mae_val  / arma_mae_dict[period_label],  4) if arma_mae_dict[period_label]  else np.nan

        # ---- CRPS (Gaussian, rolling sigma)
        if model in all_ivals:
            sigmas = np.array([rolling_sigma(actuals, all_preds[model][LAG_NOWCAST], t)
                               for t in range(len(actuals))])
            crps_arr = np.array([gaussian_crps(y[i], yhat[i], sigmas[mask][i])
                                 for i in range(len(y))])
            valid_crps = ~np.isnan(crps_arr)
            crps_val = float(crps_arr[valid_crps].mean()) if valid_crps.sum() > 0 else np.nan
        else:
            crps_val = np.nan

        # ARMA CRPS for relative
        arma_yhat_mask = all_preds["ARMA"][LAG_NOWCAST][mask] if "ARMA" in loaded_models else np.full(len(y), np.nan)
        arma_sig_mask  = np.array([rolling_sigma(actuals, all_preds["ARMA"][LAG_NOWCAST], t)
                                   for t in range(len(actuals))])[mask] if "ARMA" in loaded_models else np.full(len(y), np.nan)
        arma_crps_arr  = np.array([gaussian_crps(y[i], arma_yhat_mask[i], arma_sig_mask[i])
                                   for i in range(len(y))])
        arma_crps_val  = float(arma_crps_arr[~np.isnan(arma_crps_arr)].mean()) if (~np.isnan(arma_crps_arr)).sum() > 0 else np.nan
        rel_crps = round(crps_val / arma_crps_val, 4) if arma_crps_val and not np.isnan(arma_crps_val) else np.nan

        # ---- Coverage at 95%
        if model in all_ivals:
            cov_95, _ = empirical_coverage(actuals, all_ivals[model], mask, 0.95)
            cov_dev   = round(abs(cov_95 * 100 - 95), 1) if not np.isnan(cov_95) else np.nan
        else:
            cov_95 = np.nan
            cov_dev = np.nan

        # ---- Interval width at 95%
        if model in all_ivals:
            w_arr = all_ivals[model][0.95]["width"][mask]
            avg_w = float(w_arr[~np.isnan(w_arr)].mean()) if (~np.isnan(w_arr)).sum() > 0 else np.nan
        else:
            avg_w = np.nan

        # ---- DM stat vs ARMA (from pre-loaded table if available)
        dm_stat = dm_stars = np.nan
        if dm_clean_df is not None and model != "ARMA":
            sub_dm = dm_clean_df[(dm_clean_df["Model"] == model) &
                                  (dm_clean_df["Period"] == period_label)]
            if len(sub_dm) > 0:
                dm_stat  = sub_dm.iloc[0]["DM (RMSE)"]
                dm_stars = sub_dm.iloc[0]["Stars (RMSE)"]

        master_rows.append({
            "Model":                 model,
            "Period":                period_label,
            "n":                     n,
            "RMSE":                  round(rmse_val, 5),
            "MAE":                   round(mae_val,  5),
            "Rel RMSE (vs ARMA)":    rel_rmse,
            "Rel MAE (vs ARMA)":     rel_mae,
            "CRPS":                  round(crps_val, 5) if not np.isnan(crps_val) else np.nan,
            "Rel CRPS (vs ARMA)":    rel_crps,
            "95% coverage":          round(cov_95 * 100, 1) if not np.isnan(cov_95) else np.nan,
            "Coverage deviation (pp)":cov_dev,
            "Avg 95% width":         round(avg_w, 5) if not np.isnan(avg_w) else np.nan,
            "DM stat (vs ARMA)":     dm_stat,
            "DM sig (vs ARMA)":      dm_stars,
        })

master_df = pd.DataFrame(master_rows)

# Add ranking columns within each period
for period_label in ["Full", "Crisis", "Normal"]:
    sub = master_df[master_df["Period"] == period_label].copy()
    sub["Rank RMSE"]  = sub["Rel RMSE (vs ARMA)"].rank()
    sub["Rank CRPS"]  = sub["Rel CRPS (vs ARMA)"].rank()
    sub["Rank Cov"]   = sub["Coverage deviation (pp)"].rank()
    sub["Overall rank"] = (sub["Rank RMSE"] + sub["Rank CRPS"] + sub["Rank Cov"]).rank().astype(int)
    for col in ["Rank RMSE", "Rank CRPS", "Rank Cov", "Overall rank"]:
        master_df.loc[master_df["Period"] == period_label, col] = sub[col].values

master_df.to_csv(OUT_DIR / "master_table.csv", index=False)
print("\nMaster table saved: master_table.csv")
print("\nPreview — Crisis period, key metrics:")
sub_view = master_df[master_df["Period"] == "Crisis"][
    ["Model", "Rel RMSE (vs ARMA)", "Rel CRPS (vs ARMA)",
     "95% coverage", "Coverage deviation (pp)", "Avg 95% width", "Overall rank"]
].sort_values("Rel RMSE (vs ARMA)")
print(sub_view.to_string(index=False))

# =============================================================================
# C. MODEL DECISION FRAMEWORK
# =============================================================================

print("\n" + "=" * 70)
print("C. MODEL DECISION FRAMEWORK")
print("=" * 70)

# Classify based on master table for the nowcast vintage
decision_rows = []
for model in loaded_models:
    row_crisis = master_df[(master_df["Model"] == model) & (master_df["Period"] == "Crisis")].squeeze()
    row_normal = master_df[(master_df["Model"] == model) & (master_df["Period"] == "Normal")].squeeze()
    row_full   = master_df[(master_df["Model"] == model) & (master_df["Period"] == "Full")].squeeze()

    # Scores (lower rel RMSE = better)
    rel_crisis  = row_crisis.get("Rel RMSE (vs ARMA)", np.nan) if hasattr(row_crisis, "get") else np.nan
    rel_normal  = row_normal.get("Rel RMSE (vs ARMA)", np.nan) if hasattr(row_normal, "get") else np.nan
    rel_crps_c  = row_crisis.get("Rel CRPS (vs ARMA)", np.nan) if hasattr(row_crisis, "get") else np.nan
    cov_dev_f   = row_full.get("Coverage deviation (pp)",  np.nan) if hasattr(row_full,   "get") else np.nan
    range_perf  = abs(rel_crisis - rel_normal) if not (np.isnan(rel_crisis) or np.isnan(rel_normal)) else np.nan

    decision_rows.append({
        "Model":                       model,
        "Best point forecast — crisis": "YES" if not np.isnan(rel_crisis) and rel_crisis == min(
            [r for r in [master_df[(master_df["Model"]==m) & (master_df["Period"]=="Crisis")]["Rel RMSE (vs ARMA)"].values[0]
                          for m in loaded_models] if not np.isnan(r)]) else "",
        "Best point forecast — normal": "YES" if not np.isnan(rel_normal) and rel_normal == min(
            [r for r in [master_df[(master_df["Model"]==m) & (master_df["Period"]=="Normal")]["Rel RMSE (vs ARMA)"].values[0]
                          for m in loaded_models] if not np.isnan(r)]) else "",
        "Rel RMSE — crisis":   round(rel_crisis, 3) if not np.isnan(rel_crisis) else np.nan,
        "Rel RMSE — normal":   round(rel_normal, 3) if not np.isnan(rel_normal) else np.nan,
        "Rel CRPS — crisis":   round(rel_crps_c, 3) if not np.isnan(rel_crps_c) else np.nan,
        "Coverage dev — full": round(cov_dev_f,  1) if not np.isnan(cov_dev_f) else np.nan,
        "Regime swing":        round(range_perf, 3) if not np.isnan(range_perf) else np.nan,
        "Most robust?":        "YES" if not np.isnan(range_perf) and range_perf == min(
            [abs(
                master_df[(master_df["Model"]==m) & (master_df["Period"]=="Crisis")]["Rel RMSE (vs ARMA)"].values[0] -
                master_df[(master_df["Model"]==m) & (master_df["Period"]=="Normal")]["Rel RMSE (vs ARMA)"].values[0]
             ) for m in loaded_models
             if not np.isnan(master_df[(master_df["Model"]==m) & (master_df["Period"]=="Crisis")]["Rel RMSE (vs ARMA)"].values[0])
             and not np.isnan(master_df[(master_df["Model"]==m) & (master_df["Period"]=="Normal")]["Rel RMSE (vs ARMA)"].values[0])
            ]) else "",
    })

decision_df = pd.DataFrame(decision_rows)
decision_df.to_csv(OUT_DIR / "decision_framework.csv", index=False)

print("\nModel Decision Framework:")
print("-" * 70)
for _, r in decision_df.iterrows():
    tags = []
    if r["Best point forecast — crisis"] == "YES": tags.append("BEST POINT/CRISIS")
    if r["Best point forecast — normal"] == "YES": tags.append("BEST POINT/NORMAL")
    if r["Most robust?"] == "YES":                 tags.append("MOST ROBUST")
    tag_str = " · ".join(tags) if tags else "—"
    print(f"  {r['Model']:<16}  crisis={r['Rel RMSE — crisis']:.3f}  "
          f"normal={r['Rel RMSE — normal']:.3f}  swing={r['Regime swing']:.3f}  [{tag_str}]")

print("\nSaved: decision_framework.csv")

# =============================================================================
# G. CONSISTENCY TABLE — point vs density vs calibration
# =============================================================================

print("\n" + "=" * 70)
print("G. CONSISTENCY TABLE")
print("=" * 70)

cons_rows = []
for model in loaded_models:
    for period_label in ["Crisis", "Normal"]:
        sub = master_df[(master_df["Model"] == model) & (master_df["Period"] == period_label)]
        if len(sub) == 0:
            continue
        row = sub.squeeze()
        rel_r  = row["Rel RMSE (vs ARMA)"] if "Rel RMSE (vs ARMA)" in row else np.nan
        rel_c  = row["Rel CRPS (vs ARMA)"] if "Rel CRPS (vs ARMA)" in row else np.nan
        cov    = row["95% coverage"]        if "95% coverage" in row else np.nan
        cov_d  = row["Coverage deviation (pp)"] if "Coverage deviation (pp)" in row else np.nan
        avg_w  = row["Avg 95% width"]       if "Avg 95% width" in row else np.nan
        dm_s   = row["DM sig (vs ARMA)"]   if "DM sig (vs ARMA)" in row else ""

        # Interpretation
        point_perf  = "Better" if not np.isnan(rel_r) and rel_r < 1.0 else "Worse/tied"
        dense_perf  = "Better" if not np.isnan(rel_c) and rel_c < 1.0 else "Worse/tied"
        calibration = ("Good"      if not np.isnan(cov_d) and cov_d < 10
                       else "Moderate" if not np.isnan(cov_d) and cov_d < 20
                       else "Poor")
        significant = "Yes" if isinstance(dm_s, str) and dm_s != "" else "No"

        cons_rows.append({
            "Model":                    model,
            "Period":                   period_label,
            "Point forecast perf":      point_perf,
            "Density forecast perf":    dense_perf,
            "Calibration (cov dev)":    f"{cov_d:.1f}pp" if not np.isnan(cov_d) else "—",
            "Sharpness (95% width)":    f"{avg_w:.4f}" if not np.isnan(avg_w) else "—",
            "Sig. better than ARMA?":   significant,
            "Interpretation": (
                "Strong: wins on point and density, calibrated"
                if point_perf == "Better" and dense_perf == "Better" and calibration == "Good"
                else "Good: wins on point, density moderate"
                if point_perf == "Better" and dense_perf == "Better"
                else "Mixed: point wins but density/calibration weaker"
                if point_perf == "Better"
                else "Weak: no consistent improvement over ARMA"
            ),
        })

cons_df = pd.DataFrame(cons_rows)
cons_df.to_csv(OUT_DIR / "consistency_table.csv", index=False)
print(cons_df[["Model","Period","Point forecast perf","Density forecast perf",
               "Calibration (cov dev)","Sig. better than ARMA?","Interpretation"]].to_string(index=False))
print("\nSaved: consistency_table.csv")

# =============================================================================
# FIGURES
# =============================================================================

print("\n" + "=" * 70)
print("GENERATING FIGURES")
print("=" * 70)

# ---- FIGURE 1: Coverage heatmap (all models x periods x levels) ----------
fig, axes = plt.subplots(1, 2, figsize=(13, 4.5))
for ax, period_label, title in zip(
    axes,
    ["Crisis", "Normal"],
    ["Crisis (2019Q1–2021Q2)", "Non-crisis (2022Q1–2024Q2)"]
):
    sub = cov_df[cov_df["Period"] == period_label].copy()
    piv = sub.pivot(index="Model", columns="Nominal level", values="Empirical (pct)")
    piv = piv.reindex([m for m in loaded_models if m in piv.index])

    im = ax.imshow(piv.values, aspect="auto",
                   cmap="RdYlGn", vmin=40, vmax=110)
    ax.set_xticks(range(len(piv.columns)))
    ax.set_xticklabels(piv.columns, fontsize=9)
    ax.set_yticks(range(len(piv.index)))
    ax.set_yticklabels(piv.index, fontsize=9)
    ax.set_title(f"Coverage (%) — {title}", fontsize=10, fontweight="bold", pad=8)
    ax.set_xlabel("Nominal confidence level", fontsize=9)

    # Annotate cells
    for i in range(piv.shape[0]):
        for j in range(piv.shape[1]):
            val = piv.values[i, j]
            if not np.isnan(val):
                ax.text(j, i, f"{val:.0f}%", ha="center", va="center",
                        fontsize=8, color="black" if 60 < val < 100 else "white")

# Nominal reference line indicator
for ax in axes:
    ax.axhline(-0.5, color="gray", lw=0.3)

plt.colorbar(im, ax=axes, label="Empirical coverage (%)", shrink=0.8, pad=0.02)
fig.suptitle("Empirical coverage analysis — bootstrap prediction intervals\n"
             "Target: diagonal cells should equal nominal level (100% = perfect calibration)",
             fontsize=10)
plt.tight_layout()
fig.savefig(OUT_DIR / "fig1_coverage_heatmap.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: fig1_coverage_heatmap.png")

# ---- FIGURE 2: Vintage gain chart (crisis and normal side by side) --------
fig, axes = plt.subplots(1, 2, figsize=(13, 5), sharey=False)
for ax, period_label, title in zip(
    axes, ["Crisis", "Normal"],
    ["Crisis (2019Q1–2021Q2)", "Non-crisis (2022Q1–2024Q2)"]
):
    sub = vg_df[vg_df["Period"] == period_label]
    for _, row in sub.iterrows():
        model = row["Model"]
        vals  = [row["RMSE lag=-2"], row["RMSE lag=-1"], row["RMSE lag=0 (nowcast)"]]
        col   = MODEL_COLORS.get(model, "#888")
        lw    = 2.5 if model in ["DFM", "Ridge"] else 1.8
        ls    = "-" if model in ["DFM", "Ridge", "MIDAS"] else "--"
        ax.plot([-2, -1, 0], vals, color=col, lw=lw, ls=ls,
                marker="o", markersize=5, label=model)

    ax.set_xticks([-2, -1, 0])
    ax.set_xticklabels(["lag=−2\n(2 months before)", "lag=−1\n(1 month before)", "lag=0\n(nowcast)"],
                        fontsize=8)
    ax.set_xlabel("Vintage", fontsize=9)
    ax.set_ylabel("RMSE", fontsize=9)
    ax.set_title(title, fontsize=10, fontweight="bold")
    ax.legend(fontsize=7.5, loc="upper right", framealpha=0.7)
    ax.annotate("← lower = better", xy=(0.02, 0.04), xycoords="axes fraction",
                fontsize=7.5, color="gray")

fig.suptitle("RMSE by vintage — improvement as intra-quarter information accumulates",
             fontsize=10, fontweight="bold")
plt.tight_layout()
fig.savefig(OUT_DIR / "fig2_vintage_gain.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: fig2_vintage_gain.png")

# ---- FIGURE 3: CRPS by vintage (crisis and normal) -----------------------
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
for ax, period_label, title in zip(
    axes, ["Crisis", "Normal"],
    ["Crisis (2019Q1–2021Q2)", "Non-crisis (2022Q1–2024Q2)"]
):
    sub = crps_vint_df[crps_vint_df["Period"] == period_label]
    for model in loaded_models:
        m_sub = sub[sub["Model"] == model].sort_values("Lag")
        if m_sub["rel_CRPS"].isna().all():
            continue
        col = MODEL_COLORS.get(model, "#888")
        lw  = 2.5 if model in ["DFM", "Ridge"] else 1.8
        ls  = "-" if model in ["DFM", "Ridge", "MIDAS"] else "--"
        ax.plot(m_sub["Lag"].values, m_sub["rel_CRPS"].values,
                color=col, lw=lw, ls=ls, marker="o", markersize=5, label=model)

    ax.axhline(1.0, color="#e63946", lw=1, ls=":", alpha=0.7)
    ax.set_xticks([-2, -1, 0])
    ax.set_xticklabels(["lag=−2", "lag=−1", "lag=0\n(nowcast)"], fontsize=8)
    ax.set_xlabel("Vintage", fontsize=9)
    ax.set_ylabel("Relative CRPS (ARMA = 1.0)", fontsize=9)
    ax.set_title(title, fontsize=10, fontweight="bold")
    ax.legend(fontsize=7.5, loc="upper right", framealpha=0.7)
    ax.annotate("← better than ARMA", xy=(0.02, 0.04), xycoords="axes fraction",
                fontsize=7.5, color="gray")

fig.suptitle("Relative CRPS by vintage — density forecast evaluation across vintages",
             fontsize=10, fontweight="bold")
plt.tight_layout()
fig.savefig(OUT_DIR / "fig3_crps_by_vintage.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: fig3_crps_by_vintage.png")

# ---- FIGURE 4: Master comparison bar chart (rel RMSE + rel CRPS) ---------
fig, axes = plt.subplots(2, 2, figsize=(13, 9))
titles = [
    ("Rel. RMSE — crisis",   "Crisis",  "Rel RMSE (vs ARMA)"),
    ("Rel. RMSE — normal",   "Normal",  "Rel RMSE (vs ARMA)"),
    ("Rel. CRPS — crisis",   "Crisis",  "Rel CRPS (vs ARMA)"),
    ("Rel. CRPS — normal",   "Normal",  "Rel CRPS (vs ARMA)"),
]
for ax, (title, period, metric) in zip(axes.flatten(), titles):
    sub = master_df[master_df["Period"] == period].sort_values(metric)
    colors = [MODEL_COLORS.get(m, "#888") for m in sub["Model"]]
    bars = ax.bar(sub["Model"], sub[metric], color=colors,
                  edgecolor="white", linewidth=0.5, alpha=0.85)
    ax.axhline(1.0, color="#e63946", lw=1.2, ls="--", alpha=0.8, label="ARMA = 1.0")
    for bar, val in zip(bars, sub[metric]):
        if not np.isnan(val):
            ax.text(bar.get_x() + bar.get_width() / 2,
                    val + 0.01, f"{val:.3f}",
                    ha="center", va="bottom", fontsize=7.5)
    ax.set_title(title, fontsize=9, fontweight="bold")
    ax.set_ylabel("Ratio vs ARMA", fontsize=8)
    ax.set_ylim(0, max(1.3, sub[metric].max() + 0.15) if sub[metric].max() > 1 else 1.15)
    ax.tick_params(axis="x", labelsize=8, rotation=30)
    ax.legend(fontsize=7.5)

fig.suptitle("Point forecast (RMSE) vs density forecast (CRPS) — relative to ARMA = 1.0\n"
             "Values below 1.0 indicate improvement over ARMA",
             fontsize=10, fontweight="bold")
plt.tight_layout()
fig.savefig(OUT_DIR / "fig4_master_comparison.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved: fig4_master_comparison.png")

# =============================================================================
# FINAL SUMMARY
# =============================================================================

print("\n" + "=" * 70)
print("MASTER EVALUATION COMPLETE")
print("=" * 70)
print(f"Output folder: {OUT_DIR}")
print()
print("CSVs:")
print("  master_table.csv         — full integrated table (all models x periods x metrics)")
print("  coverage_analysis.csv    — empirical coverage at 68/80/95% (full/crisis/normal)")
print("  vintage_gain_table.csv   — RMSE improvement lag-2 → lag-1 → lag-0")
print("  interval_width_table.csv — average interval width by model and period")
print("  dm_clean_table.csv       — clean DM table for paper (stars + p-values)")
print("  crps_by_vintage.csv      — CRPS at all 3 lags (density evaluation by vintage)")
print("  decision_framework.csv   — model classification (best point/density/robust)")
print("  consistency_table.csv    — point vs density vs calibration consistency")
print("  robustness_naive.csv     — naive benchmarks (random walk, historical mean)")
print()
print("Figures:")
print("  fig1_coverage_heatmap.png — coverage at 68/80/95% by model and period")
print("  fig2_vintage_gain.png     — RMSE improvement across vintages (crisis/normal)")
print("  fig3_crps_by_vintage.png  — relative CRPS across vintages (density evaluation)")
print("  fig4_master_comparison.png— RMSE vs CRPS side-by-side bar charts")
```
